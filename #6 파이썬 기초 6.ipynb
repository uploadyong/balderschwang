# 정렬 알고리즘(Sorting)

파이썬에는 다양한 효율적인 정렬 알고리즘이 내장되어있습니다.

이 알고리즘들이 어떻게 동작하는지를 이해하기 위해서 우리는 몇 가지 가장 유명한 정렬 알고리즘을 살펴보려고 합니다.

이러한 정렬 알고리즘들은 적합한 응용 상황이 다 다르며, 이에 따라 자기 자신들만의 차별화된 성질을 가지고 있습니다.


그럼에도 일반적으로 정렬 알고리즘이 좋고 나쁘고를 판단할 때, 우리는 다음을 중요하게 생각합니다.

- 런타임(Runtime) : 정렬 알고리즘이 완료되기까지 필요한 시간(적으면 좋음)
- 저장 공간 : 정렬 알고리즘을 사용하며 추가로 필요한 저장 공간(적으면 좋음)
- 정확성 : 항상 올바른 정렬 결과를 내놓는가?
- 안정성 : 같은 값들 간의 선후 관계를 유지하는가?
- 기타 등등

이렇게 다양한 판단 기준이 있으나, 이 중 특히나 중요한건 런타임입니다.

이 런타임을 측정하는 방법으로 우리는 흔히 **란다우 기호**(Landau Symbol)라고 불리는 표기법을 사용하는데요,

아마 이미 모두 알고 계시는 $O(n)$, $O(n\log(n))$ 이런 표기법입니다.

여기서 $n$은 입력 데이터 수(정렬의 대상이 되는 놈들의 수)를 의미하며, 결국 란다우 표기법은 이런 입력 데이터 수에 비례해서

얼마나 많은 단계를 거쳐서 정렬이 완료되는지를 표현하기 위한 도구입니다.

란다우 기호는 원래 수학에서 등장한 개념으로, 어떤 함수의 점근적인 성질을 분석하고 묘사하기 위한 도구입니다.

수학적으로 란다우 기호는 작은 o 기호, 큰 O 기호, $\Theta$ 기호, $\Omega$ 기호로 총 4가지의 종류가 있는데요,

컴퓨터 공학에서는 알고리즘의 런타임을 표현하는 도구로서 이 중 큰 O 기호를 주로 사용합니다.

각 기호에 대한 좀 더 자세한 설명은 [이 포스팅](https://blog.naver.com/balderschwang/222445651994)에서 확인하실 수 있습니다.

우리는 이번엔 큰 O 기호만을 짧게 보고 넘어가려고 합니다.

### 란다우 기호
우리는 함수 $f(n)$에 대해 다음의 집합을 정의합니다.
$$
\mathcal{O}(f(n)) = \{ g(n) \mid \exists c > 0, n_0 \in \mathbb{N} : g(n) \leq c \cdot f(n)\;  \forall n \geq n_0 \} .
$$

간단합니다. $\mathcal{O}(f(n))$은 다음을 만족하는 함수 $g(n)$을 모두 모은 집합이에요.
$$\exists c>0\;\exists n_0\in\mathbb{N} : \forall n\geq n_0 : g(n)\leq c\cdot f(n) $$
한 마디로, 특정 $c$가 존재하여, 특정 $n_0$ 이후부터로는 $g(n)$이 $c\cdot f(n)$을 넘지 않는 함수들의 모임이라는겁니다.

이는 우리가 $g$라는 함수를 $f$라는 함수로, 특정 시점 이후, 즉 점근적으로 위로 근사할 수 있음을 의미합니다.

점근적으로(충분히 큰 $n$에 대해,) $g(n)$의 함수값은 커봐야 $f(n)$에다가 고정된 상수 $c$를 곱한 형태를 넘지는 않는다 이거죠.

마치 우리가 실수에서 어떤 집합의 상한을 생각하는 것과 비슷한 과정을 함수에 대해서 하는겁니다.

표기법을 좀 더 단순화하기 위해 $g(n) \in \mathcal{O}(f(n))$인 경우, 대신 $g(n) = \mathcal{O}(f(n))$라는 표기도 자주 사용합니다.


우리는 두 함수 $f,g$가 점근적으로 같다는 개념을 다음이 만족됨으로 정의합니다.
$$\begin{flalign*}\exists c>0\;\exists n_0\in\mathbb{N} : &\forall n\geq n_0 : g(n)\leq c\cdot f(n) \\
&\text{and also}\\
\exists c'>0\;\exists n_0'\in\mathbb{N} : &\forall n\geq n_0' : f(n)\leq c'\cdot g(n) \end{flalign*}$$

이게 만족될 시, 정의에 의해  $\mathcal{O}(f(n)) = \mathcal{O}(g(n))$가 만족됩니다.

그렇다고 $g(n) =\mathcal{O}(f(n))$라고 $\mathcal{O}(g(n)) = \mathcal{O}(f(n))$인건 당연히 아니에요.

설명한대로 등호 표기는 말 그대로 표기를 더 짧게 단순화하기 위한 방법이라서, 헷갈리시면 안됩니다.

저는 원래 수학에서 이 기호를 사용할 때, 이러한 점 떄문에 정석대로 $\in$을 주로 사용하는데, 컴공에서는 등호를 더 표준으로 사용하는거 같아서 여기선 등호를 사용할게요


### 큰 O 표기법의 알고리즘에서의 사용

만약 $g$가 우리의 알고리즘이 완료되기까지의 단계를 입력 변수의 개수 $n$에 대해서 표현하는 함수라고 하고,

우리에게 이미 알려진 함수 $f(n)$에 대해 $g(n) \in \mathcal{O}(f(n))$인 경우, 위에서 설명한 의미대로

우리는 이 알고리즘이 (점근적으로는) 못해도 $f(n)$과 비슷하거나 그 이하의 단계가 필요하다고 할 수 있습니다.

란다우 기호로 표현하고 싶은건 알고리즘의 런타임인데요, 이 런타임은 n에 따라서, 또 데이터에 따라서 다 달라질겁니다.

알고리즘별로 정렬을 하기 위해 어떻게 접근할 지가 모두 다를테니까요. 접근 방식에 따라 유리한 경우도, 불리한 경우도 있겠죠.

따라서 우리는 알고리즘의 성능 측정을 위해 런타임을 고려하려면, 진짜 최악의 상황을 생각하더라도 대충 점근적인 형태는 이러이러하다라고 할 수 있어야합니다.

이를 위해서 점근적인 상한(점근적으로 최악의 경우 필요한 단계)을 보여주는 큰 O 표기법을 사용하는거에요.

큰 O 표기법을 사용하면 우리는 함수(알고리즘)들을 비교하고, 함수들의 성장률(계산 비용의 증가 양상)을 분석할 수 있습니다.

이 표기법은 무엇보다도 $n$이라는 입력 변수만을 변수로 보며, 정확한 비용보다는 점근적인 양상을 보여주기 위한 표기법이라 상수는 고려되지 않습니다.

큰 O 표기법을 사용하면 다음의 계산 법칙들은 꽤나 유용할 수 있습니다.

- $\mathcal{O}(f(n) + g(n)) = \mathcal{O}(\max(f(n), g(n)))$,
- $\mathcal{O}(C \cdot f(n)) = \mathcal{O}(f(n))$ for $C > 0$.

많은 알고리즘의 런타임은 $f(n) + g(n)$의 덧셈이 여러 번 등장하는 형태로 표현되는데요, 이 때는 그들 중 가장 큰 놈만 생각해도 됩니다.

이건 우리가 수학에서 수열을 다루면서 했던 것과 비슷한 직관이에요. 수열에서 $n^2 + n$이 있으면, 이 수열이 증가하는 양상은

결국 $n^2$가 결정하고, $n$은 거의 무시해도 될 정도로 크게 영향을 미치지 않았었는데요, 여기서도 마찬가지입니다. 우리는 점근적인 성질을 분석하는거니까요.

따라서 만약 $f(n) = n^2$, $g(n) = n$이라고 한다면, $\mathcal{O}(f(n) + g(n)) = \mathcal{O}(n^2+n)$ 대신,  $\mathcal{O}(n^2)$으로 표기하게 됩니다.

대부분의 알고리즘은 상수 $k$에 대해, $\mathcal{O}(n^k)$의 형태의 런타임을 가집니다.

이 경우, 우리는 해당 알고리즘을 다항 런타임(시간 복잡도)을 가지는 알고리즘이라고 표현하기도 합니다.

특히 $k = 1$인 경우, 우리는 이 알고리즘이 선형 런타임(시간 복잡도)를 가진다고 하며, $k = 2$인 경우엔 이차 런타임(시간 복잡도)라고 합니다.

큰 O 안에 들어오는 놈이 로그의 형태면 로그 런타임, 지수의 형태면 지수 런타임, 상수면 상수 런타임(이 경우, $\mathcal{O}(1)$이라 표기)이라고도 하구요.

결국 큰 O 표기법은 함수의 증가하는 양상을 표현하므로, 당연히 안에 들어있는 놈의 점근적인 성장세가 작으면 작을수록 좋습니다.

입력 변수의 개수에 구애받지 않는 상수 런타임이 가장 빠르고, 그 다음이 선형, 그 다음이 로그, 이차 ,... 이렇게 가는거죠.

이제 직접 정렬 알고리즘을 보면서 그들의 런타임도 생각해보도록 하겠습니다.

## 다양한 정렬 알고리즘

우리는 오름차순으로 리스트를 정렬하는 상황을 기본으로 생각하여 이야기를 진행합니다.

### 버블 소트(Bubble sort)

버블 소트는 매우 단순한 정렬 알고리즘으로, 이름의 버블처럼, 버블이 순서대로 터지듯이 정렬하는 알고리즘입니다.

버블 소트는 리스트의 성분을 순서대로 타고가면서, 인접한 성분과 자기 자신의 순서가 잘못된 순서면 두 성분을 바꾸는걸 반복하는 알고리즘이에요.

버블 소트는 다음의 코드로 실현할 수 있으며, 시간 복잡도는 $\mathcal{O}(n^2)$입니다.


```python

def bubblesort(lst):
    n = len(lst) # n : 총 입력 변수의 개수
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j] # 다중 할당
    return lst
```

시간 복잡도가 이차의 형태인건 알고리즘의 안에 두 개의 `for` 반복문이 두 개 중첩된 형태로부터 바로 눈으로도 확인이 가능합니다.

좀 더 정확하게 생각하자면

```python
 for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j] # 다중 할당
```
이 반복문에서 `i = 0,...,n-1`에 대해서, 

안에 있는 반복문은 `j=0,...,n-i-2`에 대해 반복을 하게 되니, 각 `i`번째 단계에서 안쪽의 반복문은 총 `n-i-1`번의 과정을 반복합니다.

첫번쨰(`i = 0`)에서는 $j = 0,...,n-2$이니 총 `n-1`번, 두번째(`i = 1`)에서는 `n-2`번,... (실질적으로) 마지막(`i = n-2`)에서는 단 한 번이 반복됩니다.

즉, 모든 반복되는 과정들은 $ (n-1) + (n-2) + ... + 1 = \frac{n(n-1)}{2} = \frac{n^2-n}{2}= \mathcal{O}(n^2)$로 표현할 수 있습니다.

각 반복 단계의 내부에 있는 명령어(`if`절과 리스트 값의 다중 할당을 통한 변경)은 $n$에 대해 의존적이지 않은 상수라 무시하는거에요.

그럼 이제 위 알고리즘을 실제로 사용해볼게요.

````python
def bubblesort_print(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j] > data[j+1]:
                data[j], data[j+1] = data[j+1], data[j]
            print(data)  # 현재 데이터의 정렬 상태를 출력(정렬 과정을 보여드리기 위한 조치로, 원래는 필요 없습니다.)
    return data

# 예시
data = [90, 25, 12, 34, 64, 11]
sorted_data = bubblesort_print(data)
print("Sorted array is:", sorted_data)
````

설명한대로입니다. 우리는 첫번째 성분부터 출발하여, 이 놈 이후에 오는 서로 순서가 다른 놈들을 순차적으로 옳은 순서로 바꿔줍니다.

그 다음에는 두번째 성분으로부터 출발하여 같은 과정을 반복하고, 이를 마지막까지하는거에요.

그러면 결국 우리는 위와 같이 정렬된 리스트를 얻을 수 있습니다.

여기서 유의할 점은 

```python
def bubblesort(lst):
    n = len(lst) # n : 총 입력 변수의 개수
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j] # 다중 할당
    return lst
```

`return lst`를 안해도 된다는 점입니다. 파이썬의 구동 방식에 의해 위 `bubblesort()`로 인해서 변경되는 리스트의 값은 그대로 유지되거든요.

따라서 굳이 `return lst`를 안하더라도, 해당 함수에 입력된 리스트는 함수가 끝난 시점에 변경되어 있습니다.

````python
def bubblesort(lst):
    n = len(lst) # n : 총 입력 변수의 개수
    for i in range(n):
        for j in range(0, n-i-1):
            if lst[j] > lst[j+1]:
                lst[j], lst[j+1] = lst[j+1], lst[j] # 다중 할당

data = [90, 25, 12, 34, 64, 11]
print(data)
bubblesort(data)
print(data)
````

그럼에도 우리는 저렇게 `return`으로 결과물을 바로 반환하면 유용한 경우가 종종 있어서 저렇게 하는 경우도 많습니다.

특히나 기본으로 내장되어 있는 데이터 정렬 알고리즘은 일반적으로 반환을 하지 않아서, 그 기본 내장 함수와 구분을 위해서 저렇게 하기도 합니다.


### 삽입 정렬(Insertion sort)

삽입 정렬도 매우 단순한 정렬 알고리즘 중 하나로, 한 마디로 리스트를 정렬된 부분과 정렬되지 않은 부분으로 나누는 정렬 방식입니다.

아이디어는 정렬되지 않은 부분으로부터 원소를 가져와서, 정렬되어 있는 부분의 알맞은 자리에 삽입을 하는거에요.

```python
def insertionsort(lst):
    for i in range(1, len(lst)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < lst[j]:
            lst[j + 1] = lst[j]
            j -= 1
        lst[j + 1] = key
    return lst
```
간단합니다. 코드에 등장하는 `key`는 결국 지금 우리가 바라보고 있는 정렬되지 않은 부분의 원소인데요,

우리는 각 반복 단계에서 이 `key`를 기준으로 왼쪽의 모든 성분들은 오름차순으로 잘 정렬됨을 가정합니다.

따라서 이 `key`를 기준으로 왼쪽으로 움직여(이를 인덱스 `j`를 통해 타고가는 방식), `key`보다 더 작은 첫번쨰 숫자를 찾는 순간 `key`를 삽입합니다.

이 떄, 삽입을 하려면 우리가 찾은 `key`보다 큰 숫자는 모두 오른쪽으로 땡겨 만들어진 공간에 `key`를 넣어야 하는데요,

`while`문의 내부에 있는 과정은 이를 위한 과정이라고 할 수 있습니다.


이 친구도 버블 소트와 비슷한 형태로 반복하는 구조를 가지므로, $\mathcal{O}(n^2)$의 런타임의 알고리즘이 됩니다.

````python
def insertionsort_print(data):
    for i in range(1, len(data)):
        key = data[i]
        j = i - 1
        print("List:", data, "Key:", key) # 정렬 과정을 보여주기 위한 코드로, 실제 정렬에는 필요 없습니다.
        while j >= 0 and key < data[j]:
            data[j + 1] = data[j] # key보다 더 큰 숫자는 전부 오른쪽으로 한 칸씩 땡겨줌
            j -= 1
        data[j + 1] = key
    return data

data = [90, 25, 12, 34, 64, 11]
sorted_data = insertionsort_print(data)
print("Sorted array is:", sorted_data)
````


### 선택 정렬(Selection sort)

선택 정렬도 삽입 정렬과 비슷하게 리스트를 정렬된 부분, 정렬되지 않은 부분으로 나눠서 정렬을 하는 단순한 알고리즘입니다.

삽입 정렬에서는 정렬되지 않은 부분에서 정렬의 대상이 되는 성분(Key)를 그냥 리스트의 순서대로 뽑아서,

정렬된 부분에서 차례대로 하나 하나 비교하면서 위치를 찾아 그 성분을 삽입하는 방식이었는데요,

선택 정렬에서는 애초에 정렬되지 않은 부분의 최소값을 선택해서, 그걸 정렬되지 않은 곳의 첫번쨰 원소와 교환하여 정렬시킨다는 차이가 있습니다.

예를 들어 가장 첫 단계에서 우리의 리스트는 전체가 정렬되지 않은 형태라고 생각할 수 있겠죠? 그니까 여기서 최소값을 찾아서,

그 최소값을 가장 첫번째 리스트 원소와 교환시킵니다. 그러면 그 순간, 그 첫번쨰 원소를 제외한 나머지 부분은 정렬되지 않았겠죠?

그 정렬되지 않은 부분에서 다시 최소값을 찾아서, 두번쨰 원소 자리에 두고, 이걸 계속해서 반복하는거죠.

알고리즘은 다음과 같이 코딩할 수 있습니다.
```python
def selectionsort(arr):
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```
여기서도 마찬가지로 두 개의 반복문이 사용되는데요, 이걸로 직관적으로 얘도 $\mathcal{O}(n^2)$의 복잡도를 가짐을 알 수 있습니다.

자세한 분석은 생략합니다.

````python
def selectionsort_print(data):
    for i in range(len(data)):
        min_idx = i # 최소값의 위치를 저장하기 위한 변수
        for j in range(i + 1, len(data)): # 정렬되지 않은 부분을 탐색, j는 최소값을 탐색하기 위해 사용되는 인덱스
            if data[j] < data[min_idx]:   # 만약 j번째 데이터가 우리가 가진 최소값보다 작다면
                min_idx = j               # 최소값 인덱스를 j로 업데이트
        data[i], data[min_idx] = data[min_idx], data[i] # for문으로 모든 정렬되지 않은 부분을 탐색했다면, i번째와 최소값의 값을 변경!
        print("List:", data, "Min:", data[i]) # 정렬 과정을 보기 위한 코드로 실제 정렬에 영향 x
    return data
data = [90, 25, 12, 34, 64, 11]
sorted_data = selectionsort_print(data)
print("Sorted array is:", sorted_data)
````


### 퀵 소트(Quick sort)

위에서는 일단 가장 단순한 정렬 알고리즘들을 다뤘는데요, 이제는 그거보단 좀 더 복잡하지만 더 효율적인 정렬 알고리즘을 다뤄보려고 합니다.

퀵 소트는 매우 효율적인 정렬 알고리즘으로, 흔히 말하는 **분할 정복법**(Divide-and-Conquer)이라는 전략을 따르는 알고리즘입니다.

분할 정복법은 하나의 복잡하고 커다란 문제를 직접 해결하기 쉬운 작은 단위의 문제들로 쪼개서 해결하는 알고리즘의 전략이자 패러다임입니다.

분할 정복을 이용하는 알고리즘은 우선 문제를 분할하고, 이 분할된 부분 문제들을 재귀적인 방법으로 정복하여, 그 해결책을 다시 합쳐서 원래 문제를 풉니다.

퀵 소트에서 우리는 우선 주어진 리스트에서 하나의 원소를 골라, 이를 **피봇**(Pivot)이라 합니다. 이게 분할 정복법의 분할에 해당되는 단계입니다.

우리는 이 피봇을 기준으로, 피봇보다 작은 원소는 왼쪽으로, 큰 원소는 오른쪽으로 보냅니다. 이 단계를 파티션(Partition)이라 합니다.

그러면 그 피봇은 해당 리스트의 최종 정렬 상태에서 봤을 떄, 정확하게 자기 위치에 자리하게 됩니다. 이게 퀵 소트의 핵심이에요.

여튼 파티션을 마친 상황에서, 우리는 다시 피봇 기준의 왼쪽의 부분 리스트, 오른쪽 부분 리스트가 남게 되는데요,

이 두 부분 리스트에서 다시 재귀적으로 피봇을 골라 파티션을 하는걸 반복하는 게 퀵 소트가 되겠습니다.

코드는 다음과 같습니다.

```python
def quicksort(arr):
    if len(arr) <= 1: 
        return arr
    
    pivot = arr[len(arr) // 2]   # 리스트의 중앙을 피봇으로 잡습니다.
    left = [x for x in arr if x < pivot] # 피봇보다 작은 놈들을 left라는 리스트에 몰아 넣습니다.
    middle = [x for x in arr if x == pivot] # 피봇과 같이 동일한 놈들은 중앙에 놓습니다.
    right = [x for x in arr if x > pivot] #피봇보다 큰 친구들은 오른쪽에 놓습니다.
    return quicksort(left) + middle + quicksort(right) # 이러한 과정을 middle을 제외한 나머지 left, right에 반복하고, 이들을 이어 붙입니다.
```

설명한 그대로에요. 다만 설명에서 피봇과 같은 값들이 들어있을 경우는 상정하지 않았으나, 실전에서는 리스트에 피봇과 같은 친구들이 있을 수도 있으므로

저렇게 `middle`이라는 리스트에는 피봇과 같은 값을 가진 친구들은 모두 넣어서 저렇게 정렬을 해줍니다.

저렇게 하면 재귀적으로 우리가 정렬하고자 하는 리스트로부터, `left`, `right`에 또 다시 퀵소트를 반복하게 되면서 결국 정렬이 되게 됩니다.

이제 런타임을 생각해보죠. $n$이라는 리스트의 길이가 주어지면, 우리는 이 리스트를 `left`, `right`라는 두 부분 리스트로 파티션합니다.

이 과정은 총 $\mathcal{O}(n)$만큼의 런타임을 필요로 하는데요, 이걸 우리는 `left`, `right`라는 두 부분 리스트에 다시 재귀적으로 적용합니다.

그러면 우리는 이 재귀적으로 반복되는 횟수를 구해서, 그 횟수를 $\mathcal{O}(n)$에 곱해서 퀵 소트의 런타임을 생각해볼 수 있겠죠.

우선 가장 이상적인 경우를 생각해보죠. 재귀가 끝나는 시점은 결국 파티션으로 나눠진 `left`, `right`의 길이가 1이 되는 시점인데요,

만약 $n = 16$개의 성분으로 이뤄진 중복된 성분이 없는 리스트가 주어진다면 이 리스트의 첫번째 재귀에서의 `left`, `right`는 7~8개의 성분으로 구성되며, 

또 이 놈들의 `left`, `right` 는 3\~4개의 성분으로  구성되고, 그 다음 단계의 `left`, `right`에는 1~2개의 성분으로 구성되어 반복이 끝납니다.

즉, 대충 $16\rightarrow 8 \rightarrow 4\rightarrow 2\rightarrow1$의 형태로, 지수 함수 $2^n$의 역연산의 형태만큼 재귀가 진행되므로

파티셔닝하는 $\mathcal{O}(n)$의 과정을 $\log_2 (n)$번 반복하는 형태가 되어 $\mathcal{O}(n\log n)$만큼의 시간복잡도를 예상할 수 있습니다.

물론 이건 평균적인 이야기에요. 파티셔닝을 $\log_2(n)$번 반복하는거도, $16\rightarrow 8 \rightarrow 4\rightarrow 2\rightarrow1$와 같은 형태로

`left`, `right`의 크기가 줄어드는 경우에나 가능한 이야기지, 크기가 이상하게 줄어드는 경우에는 저렇게 간단하지 않을 수 있습니다.

예를 들어서, 16개의 숫자들 중에 피봇으로 고른 친구가 하필 가장 큰 놈이었다면, 그러면 `left`에는 다시 15개의 숫자들이 남겠죠?

그래서 15개의 숫자들 중에서 다시 파티셔닝을 해야하는데, 여기서도 피봇이 하필 15개들 중 가장 큰 친구였다면, 다시 14개에다가 파티셔닝을 해야합니다.

이렇게 다음 부분 리스트들의 크기가 매우 조금씩 줄어드는 경우에는 결국 $\mathcal{O}(n)$만큼의 런타임이 필요한 파티셔닝을 총 $n$번 하기 떄문에, 

이런 리스트들에 대해서 퀵 소트는 $\mathcal{O}(n^2)$라는 매우 불건전한 런타임이 생기게 됩니다.


그러면 어느 경우가 평균적이고 어느 경우가 안좋은지를 구분해서 생각할 필요가 있겠죠.

퀵 소트의 입장에서 평균적인 경우는, 주어진 리스트에서 우리가 피봇을 선택할 때, 그 피봇이 `left`, `right`를 거의 균등한 크기로 나눌 수 있을 경우입니다. 

완전 절반으로 나누는 $16\rightarrow 8 \rightarrow 4\rightarrow 2\rightarrow1$와 같은 경우는 평균을 넘어서 최고의 best case인거구요.

반대로 최악의 경우 worst case는 설명한대로 고르는 피봇이 항상 해당 리스트의 최대값/최소값(혹은 거의 그의 준하는)을 고르는 경우입니다.

이는 예를 들어, 주어진 리스트가 극단적으로 큰 값과 작은 값이 교차되는 형태로 나타나는 경우로 생각할 수 있겠죠.

어느 경우든 결국 중요한건 피봇으로 어떤 친구를 고를까이기 떄문에, 효율적인 퀵 소트에서는 피봇의 선택이 매우 중요해집니다.

따라서 현실에서는 퀵 소트를 사용할 때에는 우리처럼 단순하게 중간의 친구를 고르는 걸 넘어서,

다양한 전략을 이용하여 피봇을 고르는 형태로 퀵 소트를 구현하여 평균적인 런타임을 더 개선시켜 사용됩니다.


그런데 이렇게 생각하실 수 있어요. 왜 $\log_2 n$번의 파티셔닝을 반복하는 게 평균적인 경우냐? 최고의 경우가 아니냐?

여기엔 수학적 배경이 숨어있습니다. 앞서 다룬 버블 소트나 삽입 정렬, 선택 정렬은 결국 반복문이 사용되므로, 그 반복문의 구조를 통해서

간단히 시간 복잡도를 표현할 수 있었는데요, 퀵 소트에서는 이게 불가능합니다. 파티셔닝 자체는 시간 복잡도를 계산할 수 있으나,

파티셔닝 다음 단계는 재귀적으로 주어지니까요.

따라서 저기 있는 재귀식을 하나의 확률론을 통해 모델링하여, 기대값을 구하여 직접적으로 확률적으로 평균적인 비교횟수를 구할 수 있습니다.

그 결과물들이 모두 평균적으로 $\log_2 n$번의 재귀가 평균적임을 보여주기 떄문에, 이걸 average case로 얘기하는거에요.

"평균적이다"라는 의미도 그냥 운이 평균이다 라는 의미보다는, 기대값을 이용해서 생각했을 때, 즉 확률론적으로 평균이다는 의미인거죠.

````python
def quicksort_print(data):
    print("Data: ", data) # 과정을 확인하기 위한 단계, 없어도 정렬에 문제 없음
    if len(data) <= 1:
        return data
    pivot = data[len(data) // 2]  # 피봇 고르기
    left = [x for x in data if x < pivot] # 왼쪽은 작은 친구들
    middle = [x for x in data if x == pivot] # 중간은 같은 친구들
    right = [x for x in data if x > pivot]  # 오른쪽은 큰 친구들
    
    print("Pivot:", pivot, "right:", right, "left:", left,"\n") # 과정을 확인하기 위한 단계, 없어도 정렬에 문제 없음
    return quicksort_print(left) + middle + quicksort_print(right) # 재귀식

data = [90, 25, 12, 34, 64, 11, 11]
sorted_data = quicksort_print(data)
print("Sorted array is:", sorted_data)
````

### 합병 정렬(Merge sort)

머지 소트는 퀵 소트와 마찬가지로 분할 정복법을 활용하는 대표적인 효율적인 정렬 알고리즘입니다.

오히려 퀵 소트보다 divide and conquer라는 면에서는 진짜 교과서적으로 딱 들어맞는 알고리즘이에요.

머지 소트는 우선 주어진 리스트를 두 개로 divide해서, 각 divide된 부분 리스트를 재귀적으로 정렬합니다.

그렇게 재귀적인 과정을 이어가면, 결국 가장 마지막에는 정렬된 친구들만이 남을테니, 이를 합병(merge)하는 conquer 방식으로 정렬을 하자는거에요.

리스트의 크기가 만약 1이라면, 그 리스트는 이미 그 자체로 정렬된거잖아요? 합병 정렬은 이런 이미 정렫뢴 리스트들을 합병하는 정렬 알고리즘입니다.

구현은 다음과 같이 가능합니다


```python
def mergesort(arr):
    if len(arr) <= 1:            # 분할을 재귀적으로 반복하다가 크기가 1인 리스트가 되면, 더 이상 재귀를 할 핑요 없이 바로 그 리스트를 반환
        return arr
    mid = len(arr) // 2          # 리스트를 절반으로 나눔
    left = mergesort(arr[:mid])  # 나눈후 재귀적으로 분할
    right = mergesort(arr[mid:]) 
    return merge(left, right)    # 분할된 놈들을 합병하는 부분
    
def merge(left, right):          # 분할된 리스트를 다시 합병하기 위한 함수
    result = []                  # 합병된 리스트가 저장될 공간
    
    i = j = 0 # i는 left, j는 right를 탐색하기 위한 인덱스

    # 순서에 맞게  left와 right를 합병하기 위한 반복문(left, right는 이미 정려된 리스트들이라고 전제해도 됨)
    while i < len(left) and j < len(right): # left나 right 중 하나가 더 이상 남지 않을 때까지 반복함
        if left[i] < right[j]:             # left의 원소와 right의 원소를 비교하여, left의 원소가 작으면 그걸 result에 넣음.
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])        # right가 더 크면 right 원소를 result에 넣음
            j += 1

    # 이 시점에서 left, right 중 한 놈은 모두 result에 들어있음
    # 그러면 둘 중 아직 result에 모든 결과를 넣지 않은 친구들은 무조건 result의 성분들보다 크기가 큼.
    # 따라서 그냥 그대로 result에 남은 친구들을 붙여주면 됨.
    result = result + left[i:] + right[j:] # 이렇게 하면 예를 들어 left가 모두 result에 저장되었다면, 
                                           # i는 left의 인덱스 범위를 벗어나서 left[i:]는 빈 리스트이므로, 원하는대로 result + right[j:]가 가능.
                                           # 마찬가지로 right를 모두 result에 저장한 경우에도 result + left[i:]가 됨. 
    return result
```

과정들은 결국 리스트들을 계속해서 반으로 나눠서, 분할된 리스트들의 길이가 1이 되도록 만든 후, 그들을 다시 병합한다는 아이디어입니다.

근데 이게 재귀의 구조상 재귀에 익숙하지 않으시다면, 위 코드가 대체 어떤 순서로 진행되는지가 불확실할 수 있는데요,

리스트를 하나 예시로 잡고 저 과정들을 천천히 순서대로 밟아간다면 결국 쉽게 이해가 되실겁니다.

그냥 설명한대로 리스트는 우선 계속해서 크기가 1이 될 때까지 분할되고, 그 이후에 그 크기가 1인 리스트들을 두 개 씩 합병하고, 

그렇게 합병된 크기가 2인 리스트들끼리 다시 합병해서 크기가 4가 되도록 만들고... 이걸 반복하는겁니다.

좀 어려울 수 있을 만한 부분은 아래 코드 부분 같은데요,

```python
    while i < len(left) and j < len(right): # left나 right 중 하나가 더 이상 남지 않을 때까지 반복함
        if left[i] < right[j]:             # left의 원소와 right의 원소를 비교하여, left의 원소가 작으면 그걸 result에 넣음.
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])        # right가 더 크면 right 원소를 result에 넣음
            j += 1

    # 이 시점에서 left, right 중 한 놈은 모두 result에 들어있음
    # 그러면 둘 중 아직 result에 모든 결과를 넣지 않은 친구들은 무조건 result의 성분들보다 크기가 큼.
    # 따라서 그냥 그대로 result에 남은 친구들을 붙여주면 됨.
    result = result + left[i:] + right[j:] # 이렇게 하면 예를 들어 left가 모두 result에 저장되었다면, 
                                           # i는 left의 인덱스 범위를 벗어나서 left[i:]는 빈 리스트이므로, 원하는대로 result + right[j:]가 가능.
                                           # 마찬가지로 right를 모두 result에 저장한 경우에도 result + left[i:]가 됨. 
```

여기도 이해하면 쉽습니다. 우리의 `left`, `right`는 전제상 이미 정렬된 리스트입니다. 따라서 만약 이 둘을 합친다면,

그냥 차례때로 `left`의 첫번째 원소와 `right`의 첫번째 원소를 비교하여, 더 작은 친구를 `result`에 저장하는데요,

`left`의 첫번재 원소가 더 작았다고 가정한다면, 그 다음으로는 `left`의 두번쨰 원소와 `right`의 첫번쨰 원소를 비교하여,

둘 중 더 작은 놈을 다시 `result`에 저장합니다.

이걸 그냥 계속 `left`나 `right` 중 한 놈이 모두 소진되어 `result`에 저장될 때까지 반복하는거에요.

그러면 결국 result에는 `left`나 `right` 중 한 리스트는 모두 반영한 리스트가 됩니다.

`result`에 만약 `left`의 모든 원소들이 반영되었으나, `right`에는 아직 원소가 조금 남아있다고 한다면,

자명히 `result`의 모든 원소는 `right`의 원소들보다 작거나 같으며, `right`는 정렬된 리스트니까,

가장 마지막에 `result`에 반영된 `right`의 원소의 인덱스 `j-1`를 기준으로, `j` 이후의 모든 `right`의 원소를 `result`에 더해줍니다.

`result = result + right[j:]`와 같은 형태로요.

근데 이건 다시 이렇게 `result = result + left[i:] + right[j:]`로 적어도 문제가 없습니다. 

`i`는 `left`의 인덱스 범위를 벗어나서 `left[i:]`는 어차피 빈 리스트니까요.

게다가 `result = result + left[i:] + right[j:]`라고 사용한다면, 반대의 경우(`right`가 먼저 소진되는 경우)에도 

의도한대로 `result = result = left[i:]`가 됩니다.


합병 정렬도 마찬가지로 $\mathcal{O}(n\log(n))$의 런타임을 가지는데요, 분할하는 과정이 $\mathcal{O}(\log(n))$, 

합병 시에 $\mathcal{O}(n)$의 런타임을 가지기 떄문입니다. 

합병 시에는 어차피 모든 원소들이 다 고려되어야 하니까요. 

크기가 1인 리스트들을 합치는 과정도 결국 크기가 1인 리스트들을 짝지어서 합치는걸 모두 동시에 수행하며, 

크기가 2인 리스트들을 합치는 단계에서도 이 크기가 2인 리스트들을 짝지어서 합하는걸 모두 동시에 하는거처럼요.

````python
def mergesort_print(data, depth=0):
    if len(data) <= 1:
        print("Tiefe:", depth, "Data:", data)
        return data
    mid = len(data) // 2
    left = mergesort_print(data[:mid], depth + 1)
    right = mergesort_print(data[mid:], depth + 1)
    merged = merge(left, right)
    print("Tiefe:", depth, "Links:", left, "Rechts:", right, "Zusammengefuegt:", merged)
    return merged
def merge(left, right):
    merged = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            j += 1
    merged = merged + left[i:] + right[j:]
    return merged
# Beispielaufruf
data = [90, 25, 12, 34, 64, 11]
sorted_data = mergesort_print(data)
print("Sorted array is:", sorted_data)
````

### 보고 정렬(Bogo sort)

보고 정렬은 멍청한 정렬(Dumb sort)이라는 다른 이름으로도 불리는 정렬 알고리즘입니다.

왜 멍청하냐, 그 이유는 진짜 멍청하기 떄문입니다. 정렬 알고리즘 중 가장 최악의 알고리즘이거든요.

왜 최악인지도 작동 원리를 생각하면 바로 보실 수 있을텐데요, 보고 정렬은 리스트를 그냥 계속해서 랜덤으로 정렬된 형태가 될 떄까지 섞습니다.

진짜 운으로 정렬이 완료된 최종 형태가 나올 떄까지 섞는걸 반복하는거에요 ㅋㅋ

이렇게 보고 정렬은 우연에 의존적인 정렬 방식이다보니, 보통 시간 복잡도를 이야기할 때 확률 변수를 가지고 분석을 합니다.

물론 알고리즘도 확률 변수(혹은 수도 확률 변수라는 친구)를 컴퓨터에 구현한 방식으로 작성되구요.

그 과정은 여기서 설명하진 않구요, 어쨌든 결과적으로 $\mathcal{O}(n! (n-1))$ 정도의 시간 복잡도를 생각하시면 됩니다.

보고 정렬은 걍 이런 친구도 있다 정도인거고, 그냥 간단히 보고 넘어가시면 됩니다. 코드는 다음과 같습니다.

```python
from random import shuffle

def bogosort(data):
    while not all(a <= b for a, b in zip(data, data[1:])):
        shuffle(data)
    return data
```

## 정렬 알고리즘의 성능을 판별하기 위한 또 다른 성질들

위에서도 설명한대로, 런타임은 정렬 알고리즘의 성능을 판가름하기에 좋은 요인이나, 그 외에 다른 요인들도 성능에 영향을 주었습니다.

런타임을 제외한 친구들 중 좀 중요하다 싶은 애들은 저장 공간과 안정성의 요인들인데요, 이들을 짧게나마 좀 다뤄보고 넘어가려고 합니다.

### 저장 공간

알고리즘을 위한 저장 공간, 한 단어로 **저장 공간 복잡도**(Space complexity)는 결국 알고리즘을 수행하기 위해 얼마나 많은 추가 공간이 필요한지를 보여줍니다.

당연히 우리의 입장에서는 저장 공간 복잡도가 낮은 알고리즘이 더 좋은 알고리즘입니다.

따라서 추가 저장 공간이 필요 없는 알고리즘을 우리는 **인 플레이스**(In-Place) 알고리즘이라고 이름을 붙여 특별 취급합니다.

좀 더 정확하게 말하면, 인 플레이스가 아예 추가 저장 공간이 필요없는 친구들을 의미하는건 아니에요.

위에서 다룬 버블 소트, 삽입 정렬, 선택 정렬도 저장 공간이 필요합니다. 

근데 엄청 적게 필요하죠. 주어진 리스트 외에 필요한 추가 저장 공간은 예를 들어 삽입 정렬이라면 `key`(삽입의 대상이 되는 원소를 저장하는 변수) 정도니까요.

얘네는 $\mathcal{O}(1)$만큼의 저장 공간 복잡도를 가지고, 이렇게 입력 변수의 크기에 비의존적인 추가 저장 공간이 필요한 친구들을 인 플레이스라고 하는거에요.

그에 비해 위에서 다룬 퀵 소트는

```python
def quicksort_print(data):
    if len(data) <= 1:
        return data
    pivot = data[len(data) // 2]  # 피봇 고르기
    left = [x for x in data if x < pivot] # 왼쪽은 작은 친구들
    middle = [x for x in data if x == pivot] # 중간은 같은 친구들
    right = [x for x in data if x > pivot]  # 오른쪽은 큰 친구들
 
    return quicksort_print(left) + middle + quicksort_print(right) # 재귀식
```

`left`, `right`라는 총 `n = len(data)`개의 원소를 저장할 수 있는 리스트를 생성하므로,  $\mathcal{O}(n)$ 만큼의 추가 저장 공간이 필요합니다.

따라서 얘는 인 플레이스 알고리즘이 아닙니다.

그럼에도 퀵 소트는 인플레이스로 구현이 가능합니다. 근데 다만 더 복잡해요

```python
def quicksort_inplace(data, low=0, high=None):
    if high is None:
        high = len(data) - 1
    if low < high:
        pi = partition(data, low, high)
        quicksort_inplace(data, low, pi - 1)
        quicksort_inplace(data, pi + 1, high)
        
def partition(data, low, high):
    pivot = data[high]
    i = low - 1
    for j in range(low, high):
        if data[j] < pivot:
            i += 1
            data[i], data[j] = data[j], data[i]
    data[i + 1], data[high] = data[high], data[i + 1]
    return i + 1
```

이 인플레이스 퀵소트도 위의 인 플레이스가 아닌 퀵소트처럼 $O(n\log n)$의 평균 런타임을 보여줍니다.

그리고 합병 정렬도 알고리즘 상 분할된 리스트들을 합치는 과정에서 $\mathcal{O}(n)$의 저장 공간 복잡도가 필요하므로, 인 플레이스가 아닙니다.

### 안정성

안정성은 알고리즘에서 다양한 맥락에서 사용되는데요, 정렬 알고리즘의 안정성은 결국 리스트 성분들의 상대 순서를 변경하는지의 여부를 의미합니다.

안정적인 정렬 알고리즘들은 상대 순서를 변경하지 않는 알고리즘으로 정의되며, 이는 풀어서 말해 만약 두 성분이 서로 같다면(정렬 기준으로 동등하다면),

최종적으로 해당 알고리즘으로 정렬된 리스트에서도 처음부터 그 두 성분의 순서가 그대로 유지됨을 의미합니다.

예를 들어 우리가 스트링들의 리스트를 길이로 정렬하는데, `["ㅁㄴㅇ", "ㄷㅇㅁ", "ㅇ"]`라는 리스트가 주어졌을 때,

알고리즘의 결과로 `["ㅇ", "ㅁㄴㅇ", "ㄷㅇㅁ"]`가 된다면 해당 알고리즘은 안정적이며, `["ㅇ", "ㄷㅇㅁ", "ㅁㄴㅇ"]`가 된다면 불안정적인거죠.

이 안정성은 단순히 숫자들을 정렬하는 등, 정렬 대상들이 단순히 그 자체로 자기 자신을 표현하는 친구들이면 큰 문제가 없는데요,

방금 예로 든 스트링들의 리스트, 리스트들의 리스트, 딕셔너리들의 리스트 같이 성분들이 여러 속성으로 구성되는 복잡한 리스트를 정렬할 때는 중요합니다.

예를 들어, 우리가 학교의 학생들의 이름과 성별의 두 정보를 따로 모아 포함하는 객체들의 리스트를 가나다순으로 정렬한다고 해보죠.

만약 우리의 정렬 알고리즘이 안정적이라면, 우선 해당 리스트를 학생들 이름을 기준으로 정렬하고, 그 결과물을 다시 성을 기준으로 정렬하면

우리는 성과 이름이 모두 정렬된 형태의 리스트를 얻을 수 있습니다.

하지만 만약 안정적이지 못하다면, 이름을 기준으로 정렬된 리스트를 다시 성으로 정렬해봤자, 그 중간에 성은 같으나 이름이 다른 친구들의 순서가 바뀔 수 있습니다.

따라서 우리는 이름과 성이 모두 정렬된 형태의 리스트를 얻지 못합니다.

우리가 다룬 정렬 알고리즘들 중 버블 소트, 삽입 정렬, 합병 정렬은 안정적이지만, 퀵 소트나 선택 정렬은 안정적이지 않습니다.
(참고로 위에서 우리가 다룬 인 플레이스가 아닌 퀵 소트는 안정적입니다.)

따라서 만약 단순한 숫자가 아니라, 복잡한 객체들의 리스트를 정렬하는 경우, 퀵 소트/선택 정렬의 선택은 추천되지는 않습니다.

#### 복잡한 객체들의 정렬 꿀팁

여러 속성으로 이뤄진 객체가 주어진다면, 우리는 이 객체가 가진 특정 속성을 기준으로 정렬을 하게 될겁니다.

이 정렬의 기준이 되는 속성을 우리는 Key라고 합니다.

예를 들어서, 다음의 클래스가 주어졌다고 생각해보죠.

```python
class Customer() :
    def __init__(self, name, balance) :
        self.name = name
        self.balance = balance
```
엄청 간단한 객체 클래스로, 고객들의 이름과 잔고를 저장하는 객체입니다.

이 클래스의 인스턴스들이 들어있는 다음의 리스트가 주어졌다고 해보죠.

```python
name_list = ["김찬호", "박종우", "손인욱"]
balance_list = [2400, 1557, 88848]

lst = [Customer(name,balance) for name,balance in zip(name_list,balance_list)]
```
그러면 `lst`는 각각 김찬호와 2400, 박종우와 1557, 손인욱과 88848이 묶인 `Customer` 객체들이 들어있습니다.

여기서 우리는 해당 리스트의 Key로 잔고 `balance`를 고른다고 해볼게요. 잔고를 기준으로 정렬하는거죠.

이 떄, 우리는 각 객체와 그 객체의 Key를 모아 투플로 만들 수 있습니다.

```python
lst_for_sort = [(c.balance,c) for c in lst]
```
그러면 `(Key, Object)`의 순서로 묶인 리스트가 생성됩니다.

우리는 이제 이 투플들을 `Key`(예시의 balance)를 고려하여 정렬하고, 그 결과물로부터 다시 해당 투플 옆에 끼어있는 `Object`(Customer)를 넣어주는 방식으로

객체들을 Key를 기준으로 정렬한 객체들의 리스트를 얻을 수 있습니다.

예를 들어서 버블 소트를 다음과 같이 구현한다면,

```python
def bubblesort(data):
```

````python
class Customer() :
    def __init__(self, name, balance) :
        self.name = name
        self.balance = balance

    def as_string(self) :
        return str(self.name) + " : " + str(self.balance)

def bubblesort(data):
    n = len(data)
    for i in range(n):
        for j in range(0, n-i-1):
            if data[j][0] > data[j+1][0]: # 첫번째 성분, 즉 Key간의 대소 관계만을 비교
                data[j], data[j+1] = data[j+1], data[j]  # 그리고 그 대소를 기준으로 투플 자체의 위치를 바꿈
    return data


name_list = ["김찬호", "박종우", "손인욱"]
balance_list = [2400, 1557, 88848]

lst = [Customer(name,balance) for name,balance in zip(name_list,balance_list)]
 # 정렬 전 객체들의 리스트
print(" | ".join([c.as_string() for c in lst]))
lst_for_sort = [(c.balance,c) for c in lst]

bubblesort(lst_for_sort)

lst = [tupel[1] for tupel in lst_for_sort]
print(" | ".join([c.as_string() for c in lst]))
````

## 파이썬에서 런타임을 벤치마킹하기

우리가 계속 다룬 $\mathcal{O}(n^2)$ 이런 런타임은 결국 이론적으로 평가하는 런타임입니다.

이 이론적인 런타임은 결국 기본 연산이라고도 부르는 사칙연산, 값의 비교, 값의 할당 등의 과정을 총 몇 번하는지를 입력 변수에 의존적인 형태로 평가한겁니다.

하지만 실제로 알고리즘을 사용할 때에는, 이 이론적인 구조 뿐 아니라 다른 요소들도 영향을 미치기 떄문에

실제로 런타임이 어떤지를 알 수 없는데요, 이를 위해 직접 특정 입력 변수들을 가지고 알고리즘을 실행하여 런타임을 분석하는걸 **벤치마킹**이라 합니다.

한 마디로 그냥 이상한 란다우 기호인가 뭔가 그런거 없이, 실제 실행해서 시간을 잰다는거에요.

파이썬에서는 이를 위해 `time` 혹은 `timeit`라는 모둘을 사용할 수 있습니다.

`time` 모둘은 실제 시간을 측정하기 위한 일반적인 모둘입니다. 이 모둘에는 실제 시간을 얻어내고, 특정 시점 간의 차이를 계산하는 함수가 들어있습니다.

`timeit` 모둘은 그에 비해 특정 부분 코드의 런타임을 측정하기 위한 특수한 모둘입니다. 이 모둘은 파이썬의 특정 부분에서의 런타임 측정을 최적화하며,

그 코드들의 런타임을 측정하기에 적합한 기능들을 포함하고 있습니다. 특히나 `timeit` 모둘은 코드를 여러 번 실행하여, 그 결과물을 자기가 평균내서

최대한 정확한 결과를 얻어볼 수 있도록 구성되어 있어서 간단하게 사용이 가능합니다.

`time` 모둘에는 다음의 유용한 함수들이 있습니다 :
- `time.sleep(secs)`: 프로그램을 `secs`초만큼 멈춤.
- `time.time()`: 1970년 1월 1일 00\:00\:00 UTC부터 현 시점(UTC 기준)까지 흐른 시간을 초 단위로 반환
- `time.localtime()`: 사용자의 현 시점 현지 시간 정보(연,월,일,시,분,초 등등)를 담은 객체를 반환함
- `time.localtime(secs)` : UTC 1970년 1월 1일 00\:00\:00 기준으로 `secs`초 후의 시간 정보를 사용자의 현지 시간으로 변환하여 담은 객체를 반환
- `time.gmtime()`: UTC의 현 시점 시간 정보를 담은 객체를 반환함.  
- `time.gmtime(secs)`: UTC 1970년 1월 1일 00\:00\:00 기준으로 `secs`초 후의 시간 정보를 UTC 기준 그대로 반환 
- `time.asctime()`: 사용자의 현재 시간 정보를 우리가 읽을 수 있는 형태로 반환함
- `time.asctime(t)` : 시간 정보를 담은 객체 `t`를 입력하면, 그 시간 정보 객체를 우리가 읽을 수 있는 형태로 반환함

위에서 말한 특정 시점 간의 차이는 `time.time()` 함수를 두 시점에 배치하여, 그들의 차이를 계산하여 얻어낼 수 있습니다.

다만 이 경우에는 항상 초 단위의 차이밖에 나타나지 않는데요, 만약 일, 시, 분, 초의 차이를 얻고 싶다면

함수 `time.localtime()`을 이용하여 두 시점을 측정 후, 각 시간 정보 객체간의 차이를 생각하는 방법을 사용할 수 있습니다.

````python
import time
start_time = time.time()
end_time = time.time()
execution_time = end_time - start_time
print(f"런타임 : {execution_time} 초")
````

살펴본대로 `time` 모둘은 벤치마킹에 활용될 수도 있으나, 그건 하나의 용례일 뿐, 그냥 일반적으로 시간을 다루기 위한 도구들이 든 모둘입니다.

실제로 우리가 벤치마킹 그 자체만을 위해서 활용하는 모둘은 `timeit`입니다.

이 모둘로 코드의 런타임을 측정하기 위해서 우리는 `timeit.timeit()`함수를 사용합니다.

```python
timeit.timeit(stmt, n)
```


이 함수는 하나의 코드 `stmt`와 `int`형 숫자 `n`을 받아서,  코드 `stmt`를 `n`번 수행하여, 시간을 측정한 후 총 소요 시간을 초 단위로 반환합니다.

만약 `n`이 지정되지 않는 경우에는 딱 한 번 수행하여 그 측정 결과를 바로 반환합니다.

파이썬에는 코드를 의미하는 `stmt`라는 객체가 있는데요, 이 객체는 그냥 우리가 흔히 사용하던 코드를 스트링으로 담은 형태로 정의됩니다.

우리는 코드를 스트링 형태로 직접 담아서 `stmt`의 자리에 집어넣을 수도 있는데요, 요즘은 그건 거의 하지 않습니다.

현대에는 직접적으로 `stmt`를 스트링으로 만드는 대신에, 특정 코드를 함수로 감싸서, 그 함수를 저기 집어넣습니다.

```python
timeit.timeit(func, n)
```

여기서 함수를 넣을 때 주의할 부분이 있습니다. 함수는 인력 매개변수를 필요로 하는 친구도 있고, 아닌 친구도 있어요.

매개변수가 필요 없는 경우에는 걍 다음과 같이 바로 집어넣으면 문제 없습니다.

```python
import timeit
def my_function():
    # 이 안에 실행하고 싶은 코드를 작성

# execution_time 과 같은 변수에 timeit()를 사용하여 시간을 측정하여 저장
execution_time = timeit.timeit(my_function, 1000)
print(f"런타임: {execution_time} 초") # 결과를 출력
```

근데 만약 이 함수에 입력 매개변수가 필요한 경우, 이 떄는 익명 함수 `lambda` 명령어를 활용합니다.

```python
import timeit
def my_function(arg1, arg2):
    # 이 안에 실행하고 싶은 코드를 작성

arg1 ,arg 2 = par1,par2 # 벤치마크 시에 사용할 함수의 입력 매개변수를 골라서 arg1, arg2에 저장

execution_time = timeit.timeit(lambda: my_function(arg1, arg2), number=1000) # lambda로 익명 함수의 형태로 my_function에 arg1,arg2를 넣는걸 구현
print(f"런타임: {execution_time} 초")
```
원래 익명 함수 `lambda`는 다음의 문법으로 사용되었습니다. 왼쪽에는 함수를 구성하는 변수, 오른쪽에는 그 변수를 이용한 함수의 모양이 왔죠.

```python
lambda x1,x2,...,xn : x_1,...,x_n을 활용한 함수식
```

그런데 이번에는 다음과 같이, 함수를 구성하는 함수가 왼쪽에 하나도 없습니다.

```python
lambda : my_function(arg1,arg2)
```
이는 결국 이 `lambda`로 우리가 `my_function()`에 `arg1,arg2`라는 고정적인 매개변수를 입력하는 그 과정 자체를 정의하는걸로 이해하실 수 있습니다.

원래 통상적인 `lambda`는 입력을 받아서 가공하여 뭔가를 반환하는 함수를 표현하는 도구이지만, 

여기서는 `my_function()`에 `arg1,arg2`라는 고정적인 매개변수를 입력하는 그 과정 자체를 실행하기 위한 방아쇠(재생 버튼)를 정의하는거에요.

````python
import time 
import timeit

# 1부터 n까지의 제곱수를 모두 더하는 함수
def sum_quad(n):
    return sum((i+1)**2 for i in range(n))

start_time = time.time()
sum_quad(10000)
end_time = time.time()
execution_time = end_time - start_time
print(f"런타임: {execution_time} 초")

execution_time = timeit.timeit(lambda : sum_quad(10000), number=10000)

print(f"런타임: {execution_time/10000} 초(10000번 반복한 평균)")
````

당연히 우리가 지금 하는 벤치마킹은 절대적인 게 아닙니다. 런타임에는 가용 저장 공간, 프로세서 등 매우 많은 요인들이 영향을 미쳐서,

그 때 그 때 런타임의 결과가 모두 달라집니다. 따라서 `timeit.timeit()`같은 함수로 충분히 많이 코드를 실행한 후,

그걸로 평균을 내서 런타임을 측정하는 게 정확한 결과를 얻을 수 있으며 바람직합니다. 

특히나 우리는 `timeit.timeit()`로 정렬 알고리즘들의 런타임도 측정할 수 있습니다.

예를 들어서 다음과 같이 선택 정렬과 머지 소트를 비교해보시죠.

````python
# 선택 정렬
def selection_sort(data):
    for i in range(len(data)):
        min_idx = i
        for j in range(i + 1, len(data)):
            if data[j] < data[min_idx]:
                min_idx = j
        data[i], data[min_idx] = data[min_idx], data[i]
    return data

# 머지 소트
def merge_sort(data):
    if len(data) <= 1:
        return data
    mid = len(data) // 2
    left = merge_sort(data[:mid])
    right = merge_sort(data[mid:])
    return merge(left, right)
    
def merge(left, right):
    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result = result + left[i:] + right[j:]
    return result

# 크기가 n인 역순 리스트를 선택 정렬하는 함수
def test_selection_sort(n):
    data = list(range(n, 0, -1))
    sorted_data = selection_sort(data)

# 크기가 n인 역순 리스트를 머지 소트하는 함수
def test_merge_sort(n):
    data = list(range(n, 0, -1))
    sorted_data = merge_sort(data)

for n in [1000, 2000, 4000, 8000]:
    time_selection = timeit.timeit(lambda: test_selection_sort(n), number=10)
    time_merge = timeit.timeit(lambda: test_merge_sort(n), number=10)
    print("리스트 크기:", n, "선택 정렬:", time_selection/10, "머지 소트:", time_merge/10,"\n")

````

매우 의도적으로 위 코드에서는 리스트 크기를 자꾸 2배씩 증가시키고 있는데요, 그러면 선택 정렬은 $\mathcal{O}(n^2)$의 시간 복잡도이니,

런타임이 $2^2 = 4$배 정도 증가해야할텐데요, 실제로도 그와 매우 비슷하게 증가됨을 알 수 있습니다.

또한 머지 소트는 $\mathcal{O}(n\log n)$의 시간 복잡도이므로, 약 2배씩 증가하는 게 예상되고, 실제로도 그런식으로 되는걸 확인할 수 있습니다.
