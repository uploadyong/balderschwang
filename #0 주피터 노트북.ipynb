# \#0 주피터 노트북

주피터 노트북은 일종의 파이썬 프로그래밍을 위한 대화형 공책으로, 프로그래밍, 데이터 분석 등을 매우 접근하기 쉽고 시각적인 방식으로 배우고 적용할 수 있게 해주는 좋은 도구입니다. 주피터 노트북은 특히나 처음 프로그래밍을 배우는 학생들이 직접적으로 상호 작용하여 복잡한 개념들을 이해하는대에 좋습니다.

## 주피터 노트북의 장점

- **즉각적인 피드백**: 코드를 셀(주피터 노트북의 단위)에 작성한다면, 그 코드를 즉시 실행 가능하며, 셀의 아래에 결과가 표시됩니다.
- **여러가지 기능성**: 코드 뿐 아니라 설명, 주석, 수학적인 공식들도 표시가 가능합니다. 물론 이는 형식에 맞게 작성되어야 합니다.
- **여러 언어를 지원함**: 보통은 파이썬이 주로 사용되는데요, 말고도 다른 여러 언어를 사용하실 수 있습니다.
- **간단한 공유 방식**: 그저 링크를 하나 보내주는 방법으로 다른 사용자와 공유가 가능합니다.

## 주피터 노트북의 구성

- **셀**: 주피터 노트북은 셀이라는 기본 단위로 구성됩니다. 셀들은 텍스트나 코드 등을 포함할 수 있으며, 모든 셀에는 다른 내용물을 넣어줄 수 있습니다.
- **코드 셀**: 코드 셀에는 프로그래밍 코드를 작성할 수 있으며, 이를 실행할 수 있습니다. 그리고 위에서 설명한대로 그래픽이나 계산 결과 등 코드 실행의 결과를 바로 받아볼 수 있습니다.
- **텍스트 셀**: 텍스트 셀에는 코드의 밑바탕이 되는 아이디어를 설명하기 위한 여러 설명들을 텍스트, LaTeX문법에 따른 수식 등을 작성할 수 있습니다.

## 주피터 노트북의 설치

주피터 노트북도 로컬(사용자 컴퓨터)에 설치가 가능합니다. 물론 온라인 IDE 환경을 제공하므로, 설치 없이 그저 웹사이트에서도 사용할 수 있습니다.

아무래도 온라인으로 사용하는 경우, 할당된 서버 리소스에 의존하여 사용 가능하므로, 서버가 안좋으면 사용에 지장이 있을 수 있고, 설치 시 이런 문제는 발생하지 않습니다. 또한 오프라인 사용이 가능하며, 보안상 더 안전하기도 합니다.

하지만 로컬 설치 시, 파이썬 환경도 설치하셔야 합니다. 파이썬 설치는 [파이썬 공식 홈페이지(영어)](https://packaging.python.org/en/latest/tutorials/installing-packages)를 참고해주세요.
    
근데 뭐 어차피 그냥 공부용으로 보시는 분들은 굳이 설치까지 하실 필요는 없어보입니다. 설치 과정은 아래에 작성할테니, 선택적으로 설치하시면 되겠습니다 :

1. **설치**: 설치 파일은 [공식 홈페이지](https://jupyter.org/install)의 설명대로, 파이썬 설치 후 명령 프롬프트에 `pip install notebook`와 같이, 설치하려는 인터페이스를 골라서 입력하시면, 자동으로 진행됩니다.
2. **시작하기**: 설치 후 명령 프롬프트에 `jupyter notebook`과 같이, 설치하신 인터페이스의 이름을 입력한다면 실행됩니다.
3. **코드 및 텍스트 작성**: 이제 새로운 Python Notebook(.ipynb)파일을 생성 후, 셀을 생성하셔서 텍스트 혹은 코드를 작성하시면 됩니다. 코드 실행은 `Ctrl + Enter`을 누르시면 됩니다.
5. **명령어들**: 코드 및 텍스트를 작성하시면 된다고는 했으나, 처음 파이썬 및 주피터를 사용하신다면 파이썬 명령어는 커녕 주피터 노트북의 명령어도 잘 모르실텐데요, 파이썬 코드는 다음 포스팅부터 본격적으로 다뤄지며, 주피터 노트북의 명령어들은 아래 셀에서 자세히 설명됩니다.

# 주피터의 명령어들

우선 다른거보다 실 사용을 위한 단축키 형식의 명령어를 먼저 다뤄보려고 합니다.

주피터에는 작성 모드(Edit Mode)와 명령 모드(Command Mode)라는 두 가지 모드가 있습니다. 만약 셀을 더블 클릭하신다면, 자동으로 작성 모드로 들어가서, 말 그대로 해당 셀의 내용을 작성하실 수 있습니다. 그리고 작성이 완료되었다면, 텍스트 셀이라도 위에 있는 재생 키 혹은 `Ctrl + Enter`를 눌러 실행하셔야 합니다.

작성 모드에 있다면, `Esc`를 눌러 명령 모드로 전환이 가능합니다.

## 명령 모드의 명령어들 :

- `Enter`: 명령 모드에서 다시 수정 모드(셀에 내용물을 작성하는 모드)로 전환됩니다.
- `P`: 명령 팔레트를 엽니다.
- `Shift + Enter`: 현재 셀을 실행하고, 다음 셀로 넘어갑니다.
- `Ctrl + Enter`: 현재 셀을 실행하고, 현재 셀에 그대로 머무릅니다.
- `Alt + Enter`: 현재 셀을 실행하고, 새로운 다음 셀을 생성합니다.
- `Y`: 셀을 코드 셀로 변환합니다.
- `M`: 셀을 텍스트 셀로 변환합니다.
- `A`: 현재 셀을 위에 새로운 셀을 생성합니다. 
- `B`: 현재 셀 아래에 새로운 셀을 생성합니다.
- `DD` (d를 연속 두 번): 현재 셀을 삭제합니다.
- `Z`: 이전 셀 삭제를 다시 복구합니다.
- `S`: 노트북을 저장합니다.
- `Shift + 위/아래 방향키`: 현재 셀을 기준으로, 위/아래 셀을 동시에 선택합니다.
- `Shift+M`: 선택된 셀을 병합합니다.
- 
## 작성 모드의 명령어들 (위에서 설명한대로, 명령 모드에서 `Enter`를 눌러 전환 가능):

- `Esc`: 다시 명령 모드로 전환합니다.
- `Tab`: 코드 셀에서는 자동 완성(변수 이름, 함수 이름 등 입력 시) 및 들여쓰기로 사용되고, 텍스트 셀에서는 들여쓰기로만 사용됩니다.
- `Shift + Tab`: 툴팁을 표시합니다.
- `Ctrl + ]`: 현재 행 전체를 들여씁니다.
- `Ctrl + [`: 현재 행의 들여쓰기를 다시 한 칸 왼쪽으로 옮겨줍니다. 
- `Ctrl + A`: 셀의 모든 내용물들을 선택합니다.
- `Ctrl + Z`: 이전에 작성된 내용을 복구합니다.
- `Ctrl + Shift + Z`: `Ctrl + Z`로 복구된 내용을 다시 원상복구 시킵니다. 복구의 복구
- `Ctrl + Shift + -`: 커서를 기준으로, 새로운 셀을 생성하여 기존의 셀과 나눕니다.
- `Ctrl + Enter`: 현재 셀을 실행합니다.

당연히 단축키는 더 많구요, 단축키 리스트는 상단 바에서 `Help > Show Keyboard Shortcuts`를 눌러 확인하실 수 있습니다.
