# 플롯(Plot)

수치 데이터나 함수의 시각화를 우리는 플롯(Plot), 그리고 이러한 플롯을 구성하는 과정을 플롯팅(Plotting)이라 합니다.

매트랩, R 등 다양한 수학/통계 관련 프로그래밍 언어는 주어진 데이터를 토대로 그들을 잘 시각화하는 함수를 그리는 기능들이 내장되어있습니다.

마찬가지로 파이썬에서도 이런 기능이 있는데요, 파이썬에서는 `matplotlib`라는 모둘을 통해서 플롯팅이 가능합니다.

이 모둘은 매우 강력하며, 여러 형태로 데이터들을 시각화할 수 있습니다. 우리는 이번 포스팅에서 이 모둘을 사용하기 위한 기초적인 사용법을 다뤄보려고 합니다.

## matplotlib를 import하기

`matplotlib`도 모둘이므로, 사용하려면 우선 이 모둘을 불러와야합니다. 이는 모둘 관련 포스팅에서도 다뤘듯이 다음과 같이 가능합니다.

```python
import matplotlib.pyplot as plt
```

근데 여기서 우리는 `matplotlib` 대신에 `matplotlib.pyplot`를 불러오는데요, 그 이유는 `matplotlib`를 그대로 사용하기엔 복잡하기 떄문입니다.

`matplotlib`가 플로팅을 위한 도서관이라고 한다면, `pyplot`는 그 도서관에 있는 책을 찾아주는 검색용 컴퓨터입니다.

직접 도서관에서 책을 찾아도 되겠으나, 우리가 원하는 특정 책이 있다면 그 책을 미리 검색해서 위치를 알고 찾는게 빠르겠죠?

이런 역핳을 대신해주는 게 `pyplot`입니다. 

`mathplotlib.pyplot`를 사용할 때, 이 이름이 너무 길어서 우리는 통상적으로 위와 같은 형태로 `plt`로 이름을 붙여서 사용합니다.

이렇게 로드한다면, 이제 `mathplotlib.pyplot`라는 긴 이름 대신, `plt`를 이용하여 `mathplotlib.pyplot`의 함수/클래스/변수 등을 이용할 수 있습니다.

## 선 그래프(Line Diagram)

여튼 우리가 가장 처음으로 다뤄볼 시각화 도구는 가장 단순한 형태인 선 그래프입니다.

선 그래프는 $(x_1,y_1),...,(x_n,y_n)$ 형태의 데이터들간의 관계를 시각화하는 도구로, 각 $x$ 지점에서 어떤 $y$값이 찍히는지를 2차원 평면에 나타내는 그래프입니다.

이를 위해서 우리는 $x_1,...,x_n$값들을 저장하는 리스트, 그리고 $y_1,...,y_n$값을 저장하는 리스트가 필요합니다.

물론 이 때, 두 리스트들의 길이는 동일해야합니다. 길이가 다르다면 그래프를 잘 그려낼 수 없을테니까요.

이 두개를 정의한다면, 우리는 함수 `plt.plot()`을 이용하여 $(x_1,y_1),...,(x_n,y_n)$의 선 그래프를 그릴 수 있습니다 : 

````python
import matplotlib.pyplot as plt 

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.plot(x, y)
````

지금 우리는 주피터 노트북을 사용하기 떄문에, 위 코드만 실행해도 그래프가 보일겁니다.

하지만 일반적인 파이썬 개발 환경에서는 저렇게만 하면 플롯이 보여지지 않는데요, 이를 위해서 `plt.show()`함수를 사용할 수 있습니다 : 

````python
plt.show()
````

그러면 파이썬은 사용자가 가장 최근 사용하던 도화지(그래프 그리는 공간)의 플롯을 보여줍니다.

이 선 그래프는 입력된 $(x_1,y_1),...,(x_n,y_n)$의 점들을 그래프 상에 찍어서, 순서에 맞게 서로 인접한 점들끼리의 직선을 그은 형태가 됩니다.

이 선 그래프는 우리가 흔히 생각하는 함수의 그래프를 그리기에 매우 적합한 도구입니다.

결국 $(x_1,y_1),...,(x_n,y_n)$의 점을 찍어서 이들을 잇는거다보니, 만약 함수 $f$의 정의역의 점 $x_1,...,x_n$에 대해
$$y_k = f(x_k)\quad, \forall k =1,...,n$$
형태가 되도록 $y_1,...,y_n$을 고른다면, 우리는 이걸로 만들어진 $(x_1,y_1),...,(x_n,y_n)$의 점을 이어서 $f$의 그래프를 그릴 수 있습니다.

예를 들어 $y = x^2$라는 함수는 다음과 같이 그려볼 수 있습니다.

````python
import matplotlib.pyplot as plt

step = 1               
x= range(-10, 11, step) # 구간 [-10,10]을 step의 보폭으로 쪼개서 x에 저장
y = [i**2 for i in x]   # 그 쪼개진 구간 위에서의 함수값을 y에 저장

plt.plot(x, y)
plt.show()
````

완전 이차 곡선은 아니고 자세히 보시면 뭔가 오돌토돌한데요, 그럼에도 우리는 꽤나 그럴듯한 이차 곡선 형태의 그래프를 그려낼 수 있습니다.


당연히 우리는 이렇게 함수를 그리고 나면, 그 위에 또 다른 그래프를 추가해서 겹칠 수도 있습니다.

````python
x = range(-10, 10, 1)
y1 = [i**2 for i in x]
y2 = [i for i in x]
plt.plot(x, y1, label='y=x^2')
plt.plot(x, y2, label='y=x')
plt.legend()
plt.show()
````

간단합니다. `plt.plot()`을 여러 번 사용하시면 됩니다.

우리가 도화지(지금 그래프가 그려지는 공간)를 초기화 하지 않는 이상, 파이썬은 `plt.plot()`을 같은 도화지에 계속 겹치게 그립니다.

위에서는 `plt.plot()`의 파라미터로 `label`이라는 친구를 사용했는데요, 이 파라미터는 해당 플롯에 어떤 이름을 붙일지를 지정합니다.

이렇게 `label`로 이름을 지정한 후 `plt.legend()` 함수를 실행한다면, 우리의 도화지는 오른쪽 위에 어떤 그래프가 어떤 `label`인지를 보여줍니다.

참고로 `plt.show()`를 한다면, 현재 도화지를 출력한 후, 이를 초기화합니다.(진짜 초기화는 사실 아니고, 현재 도화지의 위치를 잊는다는 게 맞습니다)

따라서 이 이후에 플롯되는 그래프들은 새로운 도화지에 그려지게 됩니다.

기본적으로 `plt.plot()`함수는 위와 같이, 선을 그은 형태의 그래프를 그려내는데요,

당연히 우리는 이 스타일을 변경할 수도 있습니다 : 

````python
x = range(-10, 11, 1)
y1 = [i**2 for i in x]
y2 = [i for i in x]
plt.plot(x, y1, 'r--', label='y=x^2')
plt.plot(x, y2, '-o', label='y=x', color='green')
plt.legend()
plt.show()
````

위와 같은 그래프인데, 그래프의 스타일을 $y = x^2$는 빨간색에 좀 길죽한 형태의 점선, $y = x$는 초록색에 길죽한 친구와 점이 이어지는 형태로 지정했죠.

이는 두 가지 방법으로 지정이 가능하며, 첫번째 방법은 $y = x^2$에 지정한 것과 같은 방식을 이용하는겁니다 :

```python
plt.plot(x, y1, 'r--', label='y=x^2')
```
여기서 우리는 $x$, $y1$의 이후 `r--`이란 파라미터를 추가했데요, `r`은 색깔을 지정하며, `--`는 저런 모양으로 그래프를 그릴 것을 지정합니다.

이런식으로 색깔을 지정한다면 다음의 선택지가 있습니다.

-  `b` 파란색(blue)
-  `g` 초록색(green)
-  `r` 빨간색(red)
-  `c` 시안색(cyan), 청록색
-  `m` 마젠타(magenta), 자홍색
-  `y` 노란색(yellow)
-  `k` 검정색(검정색은 key라고도 하는데, 여기서 앞 글자를 따옴)
-  `w` 흰색(white)
- `#FF5733` 사용자가 지정한 색상(R, G, B 순서대로 두 자리씩 0부터 256의 숫자를 16진법으로 표기, RGB에 의한 삼원색)

만약 위의 코드에서 `r--`을 `b--`로 바꾼다면 색깔이 파란색으로 변할거에요.

그리고 그래프의 모양을 지정하는 `--` 대신에도 다음의 선택지가 있습니다.

-  `-`  실선(디폴트임)
-  `--` 좀 긴 형태의 점선(위 $y = x^2$의 그래프처럼)
-  `-.` 긴 형태의 실선과 점선이 이어지는 형태
-  `:`  완전한 점선 형태
-  `-o` 플롯되는 점들에 마커(동그란 점)를 찍되, 그 사이의 연결부는 모두 실선으로 그리기(위 $y = x$의 그래프처럼)
-  `--o`, `-.o`, `:o` 모두 위의 설명과 같으나, 마커를 찍은 형태

$y = x$의 그래프는 다음의 명령어로 색깔이 지정되었습니다.

```python
plt.plot(x, y2, '-o', label='y=x', color='green')
```
우리는 위와 같이 `r--`로 색깔을 지정할 수도 있으나, 저렇게 `color = green`으로 직접 매개변수로 색을 지정해줄 수도 있습니다.

그 외에도 우리는 다음과 같은 매개변수들을 `plt.plot()`의 괄호 안에 넣어서 모양을 정할 수 있습니다.

-  `linewidth` 혹은 `lw`  : 그래프 선의 너비
-  `markersize` 혹은 `ms` : 마커의 사이즈(`-o` 등을 이용했을 때, 찍을 점들의 크기)
-  `marker`               : 마커의 스타일(`-o`는 점으로 마커를 찍는데, 그 모양을 바꿀 수 있음)
-  `alpha`                : 선의 투명도($0\leq$ alpha $\leq 1$)

특히나 `-o`는 그냥 일반적인 디폴트 실선 `-`를 지정하고(즉, 매개변수로 아무거도 안넣어도 됩니다.), `marker =  'o'`의 옵션을 골라서도 그리실 수 있습니다.

```python
plt.plot(x, y2, label='y=x', color='green', marker = 'o')
```

`marker`의 옵션은 이 동그란 점 `o` 외에도, 사각형 `s`, 다이아몬드 `D`, 별 `*` 등을 이용하실 수 있습니다.

더 자세한 `plt.plot()`함수에 대한 정보는 아래의 도움 코드를 이용하면 확인 가능하십니다.

````python
help(plt.plot)
````

## 그래프의 축과 플롯의 타이틀

우리는 `plt.show()`를 했을 때, 도화지의 이름(title)과 도화지의 축들에 대한 설명도 덧붙여줄 수 있습니다.

이를 위한 명령어는 다음과 같습니다.

```python
plt.xlabel('x', fontsize=14, color = 'red')
plt.ylabel('y', fontsize=14, color='blue')
plt.title('graph example', fontsize=16, color='k')
```
`plt.xlabel()`은 가로축의 이름을 지정하는 명령어로, 위의 명령어는 이 x축의 이름을 어떻게 표시할지, 크기는 어떻게 하면 되는지, 색깔은 뭔지를 지정합니다.

`plt.ylabel()`도 마찬가지인데, 세로축을 지정합니다.

`plt.title()`은 이 도화지가 가질 이름인 타이틀을 설정하는 함수로, 도화지의 이름, 색깔을 설정하는 명령어가 들어있습니다.

이걸 이용하면 위에서 그린 그래프를 다시 다음과 같이 그릴 수 있습니다.

````python
x = range(-5, 6, 1)
y1 = [i**2 for i in x]
y2 = [i for i in x]
plt.plot(x, y1, 'r--', label='y=x^2', linewidth=2, markersize=10, marker='o', alpha=0.5)
plt.plot(x, y2, 'g-.', label='y=x', linewidth=3, markersize=5, marker='x', alpha=0.8)

plt.xlabel('x', fontsize=14, color='red')           # 새로
plt.ylabel('y', fontsize=14, color='blue')          # 추가돤
plt.title('graph example', fontsize=16, color='k')  # 명령어들

plt.legend()
plt.show()
````

그런데 계속 이렇게 플롯팅을 하다보면 뭔가 위화감이 있죠.

우리는 그래프를 특정한 범위에서만 그리고 있는데, 도화지는 이론상은 무한합니다.

즉, 위와 같이 그래프를 $[-5,5]$의 범위에서만 그리는 경우, 도화지는 당연히 이 범위 훨씬 클테니 

`plt.show()`가 보여주는 도화지에서는 일반적으로 우리의 그래프가 거의 안보이는 수준으로 그려지는 게 정상이겠죠?

그런데도 불구하고 귀신같이 파이썬은 우리가 그린 그래프가 최적으로 잘 보이도록 조절하여 도화지를 보여줍니다.

이는 `matplotlib`가 사용자의 별도의 지정이 없으면 그래프가 잘 보이도록 자동으로 스케일링 및 이런 저런 작업을 해주기 떄문입니다.

대부분의 경우엔 이 `matplotlib`의 축복이 매우 유용한데요, 사용하다보면 이게 오히려 불편하거나, 원하는 시각화 결과를 만들어주지 않을 떄도 있습니다.

그럴 때 우리는 도화지에서 보여야 할 가로축과 세로축의 범위를 수동으로 지정하여, 그래프의 스케일을 좀 바꿔줄 필요도 있습니다.

이를 위한 명령어는 다음과 같습니다.

```python
plt.xticks(range(-10, 11, 2), fontsize=12, color='green')
plt.yticks(range(0, 1001, 100), fontsize=12, color='green')
```

간단합니다. `plt.xticks()`는 가로축을 어떻게 나눌지를 지정하는 함수로, 위의 명령어는 가로축을 $[-10,10]$의 범위에서,

보폭은 2가 되도록(즉, 가로축은 2의 단위마다 점을 찍어 나누어 표시) 그리며, 축을 나누는 지점의 글씨의 크기와 색깔을 지정하고 있습니다.

이 때, 이 보폭을 보여주는 점인 눈금을 우리는 틱(Tick)이라고 합니다.

마찬가지로 `plt.yticks()`는 세로축에서 같은 역할을 하는 함수입니다.

위의 코드를 붙여서 기존의 그래프를 그린다면, 우리의 그래프는 x축은 $[-10,10]$의 범위, y축은 $[0,1000]$의 범위에서 그려져 보여집니다.

````python
import matplotlib.pyplot as plt 

x = range(-5, 6, 1)
y1 = [i**2 for i in x]
y2 = [i for i in x]
plt.plot(x, y1, 'r--', label='y=x^2', linewidth=2, markersize=10, marker='o', alpha=0.5)
plt.plot(x, y2, 'g-.', label='y=x', linewidth=3, markersize=5, marker='x', alpha=0.8)
plt.xlabel('x', fontsize=14, color='red')
plt.ylabel('y', fontsize=14, color='blue')
plt.title('graph example', fontsize=16, color='k')
plt.legend()

plt.xticks(range(-10, 11, 2), fontsize=12, color='green')
plt.yticks(range(0, 1001, 100), fontsize=12, color='green')

plt.show()
````

설명한대로 가로축은 $[-10,10]$의 범위에서, 보폭이 2가 되도록 표시되며, 색깔은 초록색으로 표시됩니다.

## range의 한계

해석학에서 우리는 공간 위의 기하학적 도형을 파라미터 $t$에 대한 함수(곡선)으로 표현하는 방법을 자주 사용합니다.

예를 들어서, 원의 방정식 $x^2+y^2 = 1$은 파라미터 $t$를 이용하여 다음의 함수를 사용해서 표현할 수 있죠.

$$f:[0,2\pi] \to \mathbb{R},\quad t\mapsto \big(\cos(t),\sin(t)\big)$$

이런 함수도 `plt.plot()`으로 플롯이 가능합니다.

다만 `range(start, end, stepsize)`의 인자들은 오로지 정수만 허용되기 떄문에, $[0,2\pi]$의 구간을 쪼개는건

조금 다른 방식을 사용해야합니다. 우리는 이 원 곡선의 주기성을 활용해서 다음과 같이 구간을 쪼개어 플롯을 할 수 있습니다. 

````python
import matplotlib.pyplot as plt
import math

t = [i/100 for i in range(0, int(2*math.pi*100))]
x = [math.sin(i) for i in t]
y = [math.cos(i) for i in t]
plt.plot(x, y)
plt.show()
````

이런 경우에 `range()`를 쓰는 게 좀 번거롭긴한데요, 나중에 다른 방식으로 이 구간을 쪼개는 방법이 더 다뤄집니다.

## 산점도(Scatterplots)

산점도도 선 그래프와 마찬가지로, $(x_1,y_1),...,(x_n,y_n)$라는 데이터간의 관계를 보여주는 시각화 도구입니다.

하지만 둘은 당연히 전혀 다른 쓰임새가 있습니다.

선 그래프는 데이터들간의 연속적인 관계 혹은 시간적인 순서가 있음을 전제하여 점들 사이를 보간(Interpolation)하여 보여주는 그래프지만

산점도는 데이터들이 연속적이지 않고, 독립적이라는 전제에서 그들간의 개별성을 강조하여 흩어진 정도를 보여주는 그래프입니다.

비유하자면 선 그래프는 데이터들의 점들이 만들어내는 별자리를 보기 위한 도구인 반면,

산점도는 별자리가 아닌, 별들이 흩어진 모습 그 자체를 보기 위한 도구입니다.

산점도는 보통 두 데이터간의 상관성(Correlation)을 확인하거나, 데이터의 분포를 분석하기 위한 도구로 사용됩니다.

결국 구조적으로는 $(x_1,y_1), ...,(x_n,y_n)$을 이용하여 그리는거기 떄문에, 문법은 비슷하지만 이번엔 `plt.scatter()`라는 함수를 사용합니다 : 

````python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
plt.scatter(x, y)
plt.show()
````

선 그래프랑 비슷한데, 선 그래프는 점들 사이를 보간하여 직선으로 잇는 반면, 산포도는 그대로 점들을 찍은 형태를 보여줍니다.

`plt.scatter()`에서는 `plt.plot()`에서 사용했던 선 모양을 지정하는 등의 매개변수를 제외한 모든 매개변수를 동일하게 사용할 수 있습니다.

마찬가지로 축이나 타이틀 지정도 똑같이 하실 수 있습니다.

위의 예시로는 이 산포도의 역할이 잘 보이지 않을 수 있는데요, 보통 다음과 같은 코드로 확률 분포 따위를 시뮬레이션하여, 그 분포를 확인하기 위해 사용합니다.

````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n_samples = 500
sigmas = [0.5, 1.0, 2.0]
colors = ['#ff5733', 'green', 'blue']

plt.figure(figsize=(12, 6))

for i, (s, c) in enumerate(zip(sigmas, colors)):
    data = np.random.normal(loc=0, scale=s, size=n_samples)
    
    y = np.full(n_samples, i) + np.random.uniform(-0.1, 0.1, n_samples)
    
    plt.scatter(data, y, alpha=0.5, color=c, s=10, label=f'sigma = {s}')

plt.title('Simulated Random Variables: Visualizing Dispersion', fontsize=14)
plt.xlabel('Value (Range of Real Numbers)')
plt.yticks([0, 1, 2], ['Low Sigma', 'Medium Sigma', 'High Sigma'])
plt.axvline(0, color='k', linestyle='--', alpha=0.3)
plt.grid

plt.show()
````

위의 코드는 아직 이해가 안되는 부분들이 훨씬 많으실텐데요, 앞으로 파이썬 포스팅에서 하나 하나 배워나갈 예정이니 지금은 걱정않으셔도 됩니다.

## 피규어

우리는 하나의 도화지에 여러 그래프를 겹칠 수 있음을 다뤘고, 이를 이제 좀 더 자세히 들어가서 살펴보려고 합니다.

위에서 우리는 `plt.plot()`을 두 번 사용해서 그래프를 겹쳤었는데요, 당연히 `plot()`들만이 아니더라도

서로 다른 함수로 생성된 그래프도 겹칠 수 있습니다. 예를 들어 다음과 같이 선 그래프와 산점도를 겹칠 수 있습니다.

````python
import matplotlib.pyplot as plt
import math

x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]
plt.plot(x, y, label='sin(x)')
plt.scatter([i/10 for i in range(-100, 101, 10)], [math.sin(i) for i in range(-100, 101, 10)], color='red')
plt.legend()
plt.show()
````

기본적으로 `matplotlib`은 그래프를 레이어 방식으로 그립니다. 하나를 그리고, 그 위에 또 다른 하나를 그리고 이걸 반복하는거죠.

근데 당연히 우리는 그래프들을 이렇게 겹치는 방식으로 그리고 싶을 수도 있으나,

때떄로 따로 따로 그래프를 그릴 필요가 있을 수 있는데요, 이를 위해서는 `plt.figure()`함수가 사용됩니다.

크롬같은 인터넷 브라우저에서 새 탭을 열어서 하나의 창에 여러 사이트를 이용하는걸 레이어로 겹쳐서 그리는 방식이라고 한다면,

따로 따로 서로 다른 그래프를 그리는건 새 창을 여는거라고 생각할 수 있는데요, `plt.figure()`은 이 새 창을 만드는 함수인겁니다.

그리고 이 "새 창"을 figure라고 합니다.

예를 들어 우리는 위에서 본 플롯과 산점도를 다음과 같이 두 개의 창에 따로 따로 생성할 수 있습니다.

````python
import matplotlib.pyplot as plt
import math

plt.figure(1)
x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]
plt.plot(x, y, label='sin(x)')

plt.figure(2)
plt.scatter([i/10 for i in range(-100, 101, 10)], [math.sin(i) for i in range(-100, 101, 10)], color='red')
plt.show()
````

코드를 보면 아시다시피 
```python
plt.figure(1)
x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]
plt.plot(x, y, label='sin(x)')

plt.figure(2)
plt.scatter([i/10 for i in range(-100, 101, 10)], [math.sin(i) for i in range(-100, 101, 10)], color='red')

```

`plt.figure(1)`은 `1`이라는 이름의 창을 하나 새로 만드는 명령어고,  그 아래의 `plt.plot()`는 이 `1`이라는 이름의 창에 그래프를 플롯합니다.

그러고 나서 `plt.figure(2)`은 `2`라는 이름의 창을 새로 하나 만들며, 산점도는 이 새 창에 생성되는거죠.

그런데 만약 이 시점에서 다시 `1`이라는 이름의 창으로 돌아가고 싶다면, 우리는 똑같이 `plt.figure(1)`의 명령어를 통해 돌아갈 수 있습니다.

한 마디로, `plt.figure(name)`이라는 함수는 `name`이라는 창이 없다면 그 이름의 새 창을 생성, 있다면 기존의 그 창으로 돌아가는 역할을 합니다.

추가로 `plt.figure`는 창(figure)의 크기를 조정하는 역할도 할 수 있습니다.

```python

plt.figure(figsize = (a,b)) # figure의 사이즈를 가로 a인치, 세로 b인치로 조정
```

`plt.show()` 함수는 해당 함수가 사용된 시점에서 그려진 모든 창(모든 도화지)의 모든 탭(겹쳐져있는 형태의 그래프)을 동시에 출력해서 보여주는 함수입니다.

이 설명을 이해하셨다면 아래의 코드가 어떻게 플롯을 그리는지 상상하실 수 있을겁니다.

````python
import matplotlib.pyplot as plt
import math

plt.figure(1)
x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]
plt.plot(x, y, label='sin(x)')

plt.figure(2, figsize=(10, 5))
plt.scatter([i/10 for i in range(-100, 101, 10)], [math.sin(i) for i in range(-100, 101, 10)], color='red')

plt.figure(1)
y = [math.cos(i) for i in x]
plt.plot(x, y, label='cos(x)')
plt.legend()
plt.show()
````

## 부분 플롯(Subplot)

위에서 우리는 하나의 figure에 하나의 플롯을 그리는걸 다뤘었는데요,

당연히 많은 경우에 플롯들을 비교하기 위해서 하나의 figure에 여러 플롯을 그릴 필요가 있을 수 있습니다.

이렇게 하나의 figure에 그려지는 여러 플롯들을 부분 플롯(Subplot)이라 하며, 이는 `subplot()`함수를 이용하여 생성할 수 있습니다.

다만 오해하시면 안되는게요, `subplot()`함수는 이러한 부분 플롯을 직접적으로 생성하는 함수는 아니고,

우리의 플롯(위에서 다룬 `plot()`, `scatter()`, `hist()` 등)을 어떻게 하나의 figure에 나눌 것인가를 지정하는 함수입니다. 바로 보시죠 :

````python
import matplotlib.pyplot as plt
import math

x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]

plt.subplot(2, 1, 1) # figure을 2행 1열로 나눠서, 아래에 오는 친구들을 첫번째 칸에 플롯하라(이제부터 2행 1열로 나눈 칸에서 첫번째 칸을 쓰겠다)

plt.plot(x, y, label='sin(x)')
plt.legend()

plt.subplot(2, 1, 2) # figure을 2행 1열로 나눠서, 아래에 오는 친구들을 두번째 칸에 플롯하라(이제 2행 1열로 나눈 칸에서 두번째 칸을 쓰겠다)
plt.scatter([i/10 for i in range(-100, 101, 10)], [math.sin(i) for i in range(-100, 101, 10)], color='red',label="red dots")
plt.legend()
plt.show()
````

간단합니다. 함수 `subplot()`은 다음의 구조로 매개변수를 받습니다.

```python
plt.subplot(nrow, ncol, index)
```

`subplot()`으로 우리가 결국 하고싶은건, 하나의 figure을 어떻게 나눌지를 지정하는거기 떄문에, 첫번째 두 인수 `nrow`, `ncol`은

각각 figure을 몇 행 몇 열로 나눌 것인지를 정하는 놈이 됩니다.

예를 들어 3행 4열로 나눈다면, `nrow = 3`, `ncol = 4`을 지정해야 하는데요,

우리는 이렇게 나눠진 칸에다가 플롯을 넣어야하잖아요? 그래서 지금 우리가 이 3행 4열의 어떤 칸에 있는지를 알려주는 놈이 `index`가 되겠습니다.

`index`는 왼쪽에서 오른쪽으로, 위에서 아래로의 순서로 커지는 형태로 번호가 매겨집니다.

따라서 `plt.subplot(3,4,7)`인 경우, 우리는 3행 4열로 나눠진 칸(가로 4 세로 3)에서, 가로 2행, 세로 3열에 있는 칸에 위치하게 됩니다.

그 위치에다가 기존의 플롯을 위한 명령어를 작성한다면, 딱 그 위치에 해당 부분 플롯이 들어가게 됩니다.

## 그래프가 그리는 면적에 색칠하기

위에서 `plt.plot()` 등을 이용한 그래프가 그리는 면적 부분에 색칠하는 기능도 `mathplotlib`에 포함되어 있습니다.

이를 가능케하는 함수는 `plt.fill_between()`입니다.

````python
import matplotlib.pyplot as plt
import math

x = [i/10 for i in range(-100, 101)]
y = [math.sin(i) for i in x]
plt.plot(x, y, label='sin(x)') # 참고로 이 코드를 뺴더라도 색칠은 정상적으로 됩니다. 이 코드는 이 색칠이 되는 함수 테두리를 생성하기 위해 필요합니다.
plt.fill_between(x, y, 0, where=[yi>0 for yi in y], color='green', alpha=0.3)
plt.fill_between(x, y, 0, where=[yi<0 for yi in y], color='red', alpha=0.3)
plt.show()
````

사용법은 간단해요. 우선 `plt.plot()`으로 함수를 그리고, 다음과 같은 명령어를 해주면 됩니다.

```python
plt.fill_between(x, y, 0, where=[yi>0 for yi in y], color='green', alpha=0.3)
```

`x`, `y`는 색칠할 그래프의 형태를 지정하는 역할을 하며, 그 이후에 오는 `0`은 색칠을 시작할 선의 기준을 정합니다.

`0`은 그래프 상 `y = 0`인 직선을 기준으로 색칠할 범위를 정하겠다는 의미가 됩니다. 

이 `0` 대신에 만약 다른 함수의 점을 넣어준다면, 우리는 그 함수와 `x,y`가 그리는 함수 사이를 색칠하게 됩니다.(이 설명은 아래 코드에서 이어짐)

그리고 `where = [yi>0 for yi in y]`은 그래서 구체적으로 어디를 색칠할지를 정합니다. 

`where`에는 True/False로 이뤄진 리스트가 들어오는데요, 우리의 점들을 순서대로 나열하여, `where`에 들어오는 리스트의 같은 순서의 성분이

`True`인 경우, 그 부분은 색칠하고, `false`인 경우엔 색칠하지 않게 됩니다.

다만 `True`에서 `false`로 급격히 바뀌는 부분, 예를 들어 위의 사진에서 양수에서 음수가 되어버리는 부분은 `matplotlib`가 좀 적당히 조절해서,

가능한 가장 자연스러운 형태가 되도록 색칠을 합니다. 우리한테는 좋은 점인거죠.

그 이후에는 `color`, `alpha`, `label` 같은 매개변수들이 오는데요, 얘네들은 기존의 `plt.plot()`과 동일하게 사용하시면 됩니다.

이 `plt.fill_between`을 사용하면 여러 함수의 그래프들이 만들어내는 영역을 칠할 수도 있습니다. 


````python
import matplotlib.pyplot as plt
import math

x = [i/10 for i in range(-100, 101)]
y1 = [math.sin(i) for i in x]
y2 = [math.cos(i) for i in x]

plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')

plt.fill_between(x, y1, y2, where=([y1i>y2i for y1i, y2i in zip(y1, y2)]), color='green', alpha=0.3, label='sin(x)>cos(x)')

plt.legend()

plt.show()
````

간단합니다. 모두 같은데요, 그전엔 걍 `0`을 넣었던 부분에 이번에는 코사인 함수의 `x`점들 위에서의 함수값을 모은 `y2`를 넣습니다.

그리고 이 `y2`를 이용하여

```python
plt.fill_between(x, y1, y2, where=([y1i>y2i for y1i, y2i in zip(y1, y2)]), color='green', alpha=0.3, label='sin(x)>cos(x)')
```
를 실행한다면, 우리는 위와 같이 사인 함수가 코사인 함수보다 큰 경우에 색을 칠한 모습을 나타낼 수 있습니다.

## 도형 그리기

`matplotlib`는 위에서 계속 다뤘던 플롯들 뿐 아니라, 기하학적인 플롯 또한 생성시킬 수 있습니다.

결국 `plt.plot()`의 원리는 점을 찍고, 그 점을 잇는거다보니 기하학적인 도형들의 꼭짓점을 찍고 그들을 잇게 만들어 도형을 그릴 수 있는거죠.

````python
import matplotlib.pyplot as plt
import math

# 삼각형 그리기
x = [0, 1, 0, 0]  # 삼각형의
y = [0, 0, 1, 0]  # 꼭짓점을 표현하기 위한 리스트들
plt.plot(x, y, label='Triangle') # 그리기

# 원 그리기
t = [i/100 for i in range(0, 628)]
x = [0.3*math.cos(i)+0.1 for i in t]
y = [0.3*math.sin(i)+0.2 for i in t]
plt.plot(x, y, label='Circle')

# 색칠놀이
plt.fill(x, y, color='red', alpha=0.3)
plt.fill([0, 1, 0], [0, 0, 1], color='blue', alpha=0.3)

#축 끄기
plt.axis('off')
plt.legend()
plt.show()
````

사실상 설명한대로 원리는 모두 똑같아요.

다만 기하학적인 도형을 축 위에 그리는건 좀 이상하니까, 위 코드에서처럼 `plt.axis('off')`를 이용하여 축을 없애줍니다.

`plt.axis()` 함수는 이름부터 알 수 있듯, 축을 설정하는 함수입니다.

이 함수에 사용되는 유용한 매개변수들은 예를 들어 다음과 같은 친구들이 있습니다.

- `equal` 두 축의 눈금을 모두 같은 스케일로 스케일링(데이터를 모두 포함할 수 있도록 더 큰 스케일에 맞춤)
- `tight` 축을 데이터에 딱 맞게 스케일링(여백이 없도록 스케일)
- `scaled` `equal`, `tight`를 동시에 한다고 생각하면 됨.
- `on` 혹은 `off` 축을 그래프에 보여줄지, 말지의 여부를 결정. 말 그대로 온/오프


위와 같은 도형을 그린다면, 우리는 보통 변이나 꼭짓점 등에 이름을 붙이기도 하는데요, 이는 다음과 같이 가능합니다.

````python
import matplotlib.pyplot as plt
import math
x = [0, 1, 0, 0]
y = [0, 0, 1, 0]
plt.plot(x, y, label='Triangle', marker='o')

t = [i/100 for i in range(0, 628)]
x = [0.3*math.cos(i)+0.1 for i in t]
y = [0.3*math.sin(i)+0.2 for i in t]
plt.plot(x, y, label='Circle')

plt.fill(x, y, color='red', alpha=0.3)
plt.fill([0, 1, 0], [0, 0, 1], color='blue', alpha=0.3)
plt.axis('equal')

# 특정 위치에 텍스트 추가하기
plt.text(0.5, 0.1, 'c', fontsize=12)
plt.text(0.05, 0.5, 'b', fontsize=12)
plt.text(0.55, 0.55, 'a', fontsize=12)
plt.text(0.05, 0.05, 'A', fontsize=12, ha='center')
plt.text(0.85, 0.05, 'B', fontsize=12, ha='center')
plt.text(0.05, 0.85, 'C' , fontsize=12, ha='center')
plt.axis('off')
plt.legend()
plt.show()
````

`text`함수는 입력된 좌표에 텍스트를 적는데요, 우리는 이 텍스트를 그 좌표를 기준으로 정확히 어디에 적을지 인자 전달로 설정할 수 있습니다.

위 코드에서도 사용된 `ha`의 경우, horizontal allignment, 즉 수평 정렬을 어떻게 할지를 정하는 매개변수입니다.

우리는 `ha = 'center'`를 입력했으므로, 수평 방향에서는 입력된 좌표를 중심으로 텍스트를 적게 됩니다.

`center` 외에는 좌표를 기준으로 왼쪽에 적는 `left`, 오른쪽에 적는 `right` 지정이 가능합니다.

그 외의 다른 매개변수들은 다음과 같습니다.

- `va` 수직 정렬(vertical allignment)은 좌표를 기준으로`center` 위 `top`, 아래 `bottom` 등 지정 가능 
- `color` 색깔
- `fontsize` 폰트 크기
- `rotation` 텍스트 회전(0~360 혹은 'vertical'(90도), 'horizontal'(0도) 등)
- `alpha` 텍스트 투명도
- `backgroundcolor` 텍스트 배경 색깔(마치 형광펜 칠하는거처럼)
- `clip_on` `True`로 설정될 시, 글자가 그래프를 벗어나면 자름. `False`면 걍 둠.
- `bbox` 텍스트를 박스 안에 적을지의 여부(얘는 좀 복잡합니다.)
