# 수치적 방법 : 영점의 계산

수치해석학의 알파이자 오메가는 영점을 계산하는겁니다. 영점은 함수가 0이 되게하는 정의역의 지점을 의미합니다.

영점을 구하는 게 중요한 이유는 너무 많은데요, 지금 우리 시점에서 간단하게 생각할 수 있는건, 영점을 푼다는건 방정식을 푼다는 맥락에 사용할 수 있기 떄문입니다.

방정식은 결국 어떤 함수 $f(x)$에 대해서, $f(x) = 0$을 만족하는 $x$를 찾는거고, 이건 함수 $f$의 영점을 찾는거랑 같으니까요.

방정식이 간단하면 그냥 쉽게 풀 수 있겠으나, 대부분의 유용한 방정식들은 해석적으로 풀 수 없는 형태가 많은데요,

이 때 우리는 수치적인 방법을 이용하여 방정식의 해를 근사하여 계산하는 방법을 이용합니다.

심지어 해석적으로 풀 수 있는 경우에도 그 과정들이 복잡할 수 있는데요, 수치적 방법을 이용하면 그 과정을 상당히 간단하게 해결할 수도 있습니다.

## 예시 : 3차 방정식의 해를 위한 카르다노의 공식

이차방정식은 고등학교에서 배우는 근의 공식으로 매우 간단하게 해를 구할 수 있습니다.

그에 비해서 3차 방정식에는 근의 공식이 알려져있으나, 그 과정이 조금은 복잡한데요, 이번엔 이걸 좀 간단하게 다뤄보려고 합니다.

카르다노의 공식은 3차 방정식의 해를 구하는 공식입니다. 일반적으로 3차 방정식은 다음과 같은 형태로 표현할 수 있습니다.

$$
\alpha x^3 + \beta x^2 + \gamma x + \delta = 0,\quad \alpha \neq 0
$$

이는 다시 양 변을 $\alpha$로 나누면 다음과 동치입니다.
$$
x^3 + \frac{\beta}{\alpha}x^2 + \frac{\gamma}{\alpha} x + \frac{\delta}{\gamma} =0
$$

여기서 이차항 이하의 모든 항들의 계수를 그냥 간단히 $a,\;b,\;c$라고 표현한다면, 우리는 이 방정식을 다시
$$
P(x) = x^3 + a x^2 + b x + c
$$
라는 함수에 대해서, $P(x) = 0$이라는, $P$의 영점을 구하는 문제와 동일하게 생각할 수 있습니다.

이 $P$에 우리는 $x = z - \frac{a}{3}$을 넣으면
$$
P\left(z-\frac{a}{3}\right) = \left(z-\frac{a}{3}\right)^3 + a\left(z-\frac{a}{3}\right)^2 + b\left(z-\frac{a}{3}\right) + c
$$
계산을 생략하지만, 이차항이 상쇄되어 사라져서, 결국 다음과 같이 $z$에 대한 함수를 얻어낼 수 있는데요,
$$
Q(z) = z^3 + pz + q
$$

이 $Q(z)$를 $P(x)$의 축약된 다항식이라 합니다. 3차 방정식에서 이차항을 없애버린 형태라는거죠.

여기서 $u$, $v$를 $3uv + p = 0$가 만족되도록 잡으면, $z = u+v$에 대해 다음이 유도됩니다.
$$
Q(u + v) = (u + v)^3 + p(u + v) + q = u^3 + v^3 + 3u^2v + 3uv^2 + p(u + v) + q = u^3 + v^3 + (3uv + p)(u + v) + q = u^3 + v^3 + q
$$
그러면 방정식 $Q(u + v) = 0$은 다음의 형태가 됩니다.
$$
u^3 + v^3 + q = 0 \quad \cdots(*)
$$
여기서 $3uv + p = 0$가 성립하므로, $v^3u^3 = -\frac{p^3}{27}$가 유도됩니다. 이제 위 $(*)$의 양 변에 $u^3$를 곱하면 다음의 형태가 나타납니다.

$$
u^6 + u^3v^3 + qu^3 = 0\quad \Leftrightarrow\quad u^6 + qu^3 - \frac{p^3}{27} = 0
$$
이제 $t=u^3$를 잡아주면, 해당 등식은 다시 다음과 같은 이차방정식이 됩니다.
$$
t^2 + qt - \frac{p^3}{27} = 0
$$
참고로 $u^3 + v^3 + q = 0$ 와 $3uv + p = 0$에서 $u$, $v$는 대칭적이라서, 만약 $(*)$에 $v^3$을 곱하여 $t = v^3$를 잡더라도 방정식은 동일합니다.

그리고 이차방정식 $t^2 + qt - \frac{p^3}{27} = 0$은 축약 다항식 $Q$의 이차 레졸벤트(Quadratic resolvent)라고도 합니다.

이 이차방정식의 두 해를 $t_\pm$라고 한다면, 이 해들은 다음을 만족합니다.
$$
u^3 = t_\pm, \quad v^3 = -q - u^3 = -q - t_\pm
$$
무엇보다도 이차 방정식의 계수 정리(비에트의 정리)에 의해서, 다음이 성립됩니다.
$$
t_+ + t_- = -q
$$
이를 $u,v$의 대칭성을 활용하여 생각하면 다음의 결과를 얻을 수 있습니다.
$$
u^3 = t_+, \quad v^3 = t_-
$$
이 때, 대칭성을 생각한다는건 쉽게 생각해서 바로 위 등식에서 $u,v$를 바꿔도 큰 문제가 없다는겁니다.

특히나 $z = u+v$를 골랐었는데요, $u$, $v$는 이러한 대칭 관계이므로, 둘을 바꾸더라도 결국 $z$에는 어떠한 영향도 없습니다.

따라서 $u^3 = t_- $ und $v^3 = t_+$ 가 되는 경우는 아예 생각해주지 않아도 됩니다.

여튼 우리는 
$$
u^3 = t_+, \quad v^3 = t_-
$$
이 방정식을 생각해야하는데요, 이 두 방정식의 $u,v$의 범위를 복소수까지 허용해준다면, 각 방정식에는 정확히 3개의 (복소) 근이 존재합니다.

이 근들을 각각 $u_1,u_2,u_3$, $v_1,v_2,v_3$이라 해보죠. 그러면 $Q(u+v) = 0$이라는 관계에 의해서,

$P$의 축약된 다항식 $Q$의 (복소수를 포함한) 영점들의 후보 집합은 다음과 같이 나타낼 수 있습니다.

$$
\{u_i+v_j : i,j \in \{1,2,3\}\}
$$
이 집합은 그저 "후보"들의 집합일 뿐입니다. 그 이유는 결국 $Q(u+v) = 0$이 되기 위한 조건으로 우리는 추가적으로 $3uv + p = 0$라는 조건을 이용했는데요,

따라서 실제 영점들의 집합은 해당 집합에서 $3uv + p = 0$라는 조건을 추가해 준 다음 형태의 집합이 됩니다.

$$
\{u_i+v_j : i,j \in \{1,2,3\},3u_iv_j + p = 0\}
$$
이 집합은 3개 이하의 원소를 가지는데요(그 뒤의 수학적 배경은 생략), 

만약 $u_i+v_j$가 실수일 경우, 이 놈은 $Q(u_i+v_j) = 0$을 만드므로, $x_{ij} = u_i+v_j- \frac{a}{3}$는 $P(x_{ij}) = 0$을 만들게 됩니다.

이런식으로 3차 방정식의 해를 찾는 방법을 카르다노의 방법, 이걸 공식화한 다음의 공식을
$$x = \left( \sqrt[3]{-\frac{q}{2} + \sqrt{\frac{q^2}{4} + \frac{p^3}{27}}} + \sqrt[3]{-\frac{q}{2} - \sqrt{\frac{q^2}{4} + \frac{p^3}{27}}} \right) - \frac{a}{3}$$
카르다노 공식이라고 합니다. 근데 이 공식은 그리 중요하지 않습니다. 이 공식은 이차 방정식의 근의 공식과 비슷하게

삼차 방정식도 대수적으로 풀고, 방정식의 계수들로 공식화 할 수 있음을 보여주는 사실상 기념비적인 공식일 뿐이에요.

실제로 이 카르다노 방법으로 삼차 방정식을 풀기 위해서는 저 과정을 이해하는 게 중요하지, 공식은 큰 의미가 없습니다.

보면 아시다시피 우리의 근은 최대 3개가 될 수 있는데, 카르다노 공식은 오로지 $p,q$라는 우리가 알고 있는 계수들로

근을 표현하는 것에 집중한 공식이라서, 결국 하나의 근 밖에 계산하지 못하고, 다른 근들은 위에서 봤던 $u_i+v_j - \frac{a}{3}$에서

집합 $\{u_i+v_j : i,j \in \{1,2,3\},3u_iv_j + p = 0\}$의 원소 $u_i, v_j$를 뽑아서 생각해야하거든요.

실제로 카르다노 방법을 구현하는 경우에도 저 공식은 사용하지 않습니다.

### 카르다노 방법의 구현

결국 위의 과정들을 거꾸로 보신다면, 카르다노의 방법은 결국 특정 형태의 이차 방정식을 풀어서 그걸로 얻어진 수치적 정보를 이용하여

역으로 축약된 방정식을 푼 이후, 그걸로 원래의 3차 방정식을 풀어가는 과정으로 생각할 수 있습니다.

따라서 카르다노 방법은 결국 위에서 다룬 형태의 2차 방정식을 푸는 함수를 정의하고, 그걸로 축약된 방정식을 풀고, 

이를 통해 3차 방정식의 해를 구하는 과정을 거쳐서 간단히 구현할 수 있습니다.

다음 코드에는 `numpy`라는 매우 유용한 파이썬 모둘을 사용하는데요, 이 이야기는 다음 포스팅에서 더 다뤄집니다.

````python
import numpy as np

def solve_quadratic(p,q):
    """x^2 + px + q = 0 형태의 이차방정식을 품"""
    # 결국 근의 공식을 이용하는건데, cancellation을 방지하기 위한 과정이 수행됨.
    d = (p**2 - 4*q)**0.5
    if p < 0: 
        x1 = (-p + d) / 2
        x2 = q / x1
    else:
        x1 = (-p - d) / 2 
        x2 = q / x1
    return {x1, x2} # 리턴은 집합 형태로 

def cbrts(t): 
    """u^3 = t_+ 형태의 방정식을 풀기 위한 함수"""
    rt = t**(1/3)
    return {rt, rt * np.exp(2j*np.pi/3), rt * np.exp(4j*np.pi/3)} # 저 방정식의 복소수 근이 모두 포함된 집합이 리턴됨.

def solve_reduced_cubic(p,q):
    """
    축약된 방정식 x^3 + px + q = 0을 풀기 위한 함수
    """
    t_p, t_m = solve_quadratic(q, -p**3/27)
    return {ui+vj for ui in cbrts(t_p) for vj in cbrts(t_m) if abs(3*ui*vj + p) < 1e-10}

def solve_monic_cubic(a,b,c):
    """
    최고차항이 1인 3차 방정식 x^3 + ax^2 + bx + c = 0을 풀기 위한 함수
    """
    p = b - a**2/3
    q = c + 2*a**3/27 - a*b/3
    return {zi - a/3 for zi in solve_reduced_cubic(p,q)}

def solve_cubic(a,b,c,d): 
    """"
    원래 풀고자 하는 3차 방정식 ax^3 + bx^2 + cx + d = 0을 풀기 위한 함수
    """
    return { xi for xi in solve_monic_cubic(b/a, c/a, d/a)}
````

간단합니다. 세부적인 설명은 그냥 주석만 보셔도 아실거구요,

우리는 결국 위의 코드들을 이용하면 다음의 과정으로 3차 방정식을 풀게 됩니다.

1. $a x^3 + b x^2 + cx + d = 0$형태의 방정식을 풀기 위해, `solve_cubic(a,b,c,d)`를 호출
2. 그러면 `solve_cubic(a,b,c,d)`는 $x^3 + \frac{b}{a}x^2 + \frac{c}{a}x + \frac{d}{a} = 0$의 해를 `solve_monic_cubic(b/a,c/a,d/a)`를 호출하여 반환
3. `solve_monic_cubic(b/a,c/a,d/a)`는 x^3 + px + q = 0 형태의 축약된 방정식을 풀기 위한 $p,q$를 계산 후 `solve_reduced_cubic(p,q)`가 푼 값을 반환
4. `solve_reduced_cubic(p,q)`는 결국 $t^2 + qt - \frac{1}{27}p^3$을 풀기 위해 `solve_quadratic(p,-p**3/27)`의 결과로 나온 값을 구해서,
   이 값에 해당되는 $t = u^3$를 구한 후, `cbrts()`함수로 $u^3 = t_+$의 복소 근을 계산, 그들 중 $3uv +p = 0$을 만족하는 놈들만 추림.
5. 그렇게 추려진 값이 결국 `solve_cubic(a,b,c,d)`의 반환값으로 리턴됨.

근데 여기서 주의할 점이 있어요. 우리는 결국 복소근도 고려하여 생각해서, 그들 중 실수 근을 생각하려는거잖아요?

문제는 우리가 수치적 과정을 다루기 떄문에, 해석적으로 허수 부분이 0이 되는 친구들도 수치적 계산 과정 중에 완전히 사라지지 않을 수 있습니다.

따라서 결과적으로 계산되어 담겨있는 놈들 중에서는 실제 해석적으로는 실수근이나, 허수 파트가 살아있을 가능성이 있습니다.

그 경우에는 허수 파트가 무시할 정도로 적을 경우, 걍 무시하는 전략을 택하시면 됩니다. 이 "무시할 정도"가 어느 정도인지는 지금은 다루지 않겠습니다.


실제로 우리는 $10x^3 -6x^2 + 11x -6 = 0$이라는 방정식을 다음과 같이 풀 수 있습니다.

````python
a = 10
b = -6
c = 11
d = -6
s = solve_cubic(a,b,c,d)

# 해를 출력
print([f"{xi:.15f}" for xi in s]) # 값은 복소수 형태가 됩니다.
p = lambda x: a*x**3 + b*x**2 + c*x + d

# 그 해를 다시 우리의 3차 다항식에 넣은 값으로 검산, 출력
print([f"{p(xi):.15f}" for xi in s]) 
````

# 반복법

위의 3차 방정식의 카르다노 방법은 좀 특수하게 3차 다항 방정식을 위한 방법이구요, 우리는 이제 좀 더 일반적인 문제 상황을 다뤄보려고 합니다.

우리는 $x$라는 알려지지 않은 수, 벡터 등을 구해야 합니다.

이를 위한 방법들 중 가장 널리 알려진 반복법이라는 친구는 연쇄적으로 $x$를 근사하는 $x_1, x_2,...$ 열을 구성합니다.

이 때 만약 우리가 이렇게 구성되는 수열 $(x_n)_{n\in\mathbb{N}}$이 $x$로 수렴한다는 전제를 할 수 있는 경우에는 이 $x_1,x_2,..$를 충분히 많이 생성하여

특정 오차 지점에서 생성을 멈춘 후, 그걸 $x$의 근사로 생각할 수 있겠죠.

(물론 수렴성이 필수는 당연히 아닙니다. 수치적 방법은 수렴성을 모르는 상황에서도 일단 알고리즘을 돌려서 수렴 비슷한 무언가를 얻어내는 경우도 있어요.)

그러면 결국 이 $x_1, x_2,...$를 어떻게 구성시킬지가 관건입니다. 그 구성 방법에 따라서 해당 방법이 $x$로 수렴하는지의 여부가 정해질테니까요.

일반적인 관점에서 이런 류의 과정에서는 $x_1,...,x_n$까지의 숫자를 생성한 이후, 그 다음 근사인 $x_{n+1}$을 $x_n$을 비롯한 우리에게 주어진 정보로 계산합니다.

이를 다음과 같은 형태로 생각할 수 있습니다.
$$
x_{n+1} = F(x_n)
$$
여기서 $F$는 $x_n$ 및 $n$번째 단계에서 주어진 정보를 어떻게 활용해서 $x_{n+1}$을 계산할 것인지를 결정하는 알고리즘 역할을 하는 함수이구요.

물론 여기서 말하는 "이전 단계"는 바로 전 단계의 $x_{n}$만이 아니에요. $x_n$을 비롯한 $x_1,...,x_{n-1}$ 등 이전 단계의 어떤 친구들을 이용해도 상관 없습니다.

따라서 반복법은 더 일반적인 형태로 다음으로 표기하기도 합니다.

$$
x_{n+1} = F(x_n, x_{n-1}, \ldots, x_1).
$$

결국 정리하면 한마디로 우리는 단계 별로 점차 $x$를 근사하려고 노력하는 반복적인(iterative) 방법을 반복법이라고 하는겁니다.

하지만 당연히 우리는 $x_1, x_2,...$ 이걸 유한 개 구성시키는 방법으로 정확한 $x$를 얻어낼 수 없습니다.

컴퓨터 계산의 한계점 문제를 제쳐두더라도 애초에 해석적으로도 $x_1, x_2,..$ 얘네가 $x$로 수렴해도, 특정 $N$에 대해 $x_N = x$가 될지는 모르는거니까요.

그래서 우리는 우리가 정해둔 오차의 허용치(Tolerance) $\text{eps}>0$에 대해서, 다음을 만족하는 $N$을 구하여 $x$를 근사합니다.

$$
\Vert x_N - x\Vert\leq \text{eps}
$$

이러한 조건을 종료 조건(Termination condition)이라 합니다.

근데 애초에 $x$가 우리가 구하고자 하는 친구인데, $x$를 이용해야하는 위의 종료 조건은 말이 안됩니다. 경우에 따라서 가능하기도 하겠으나, 안되는 경우도 많겠죠.

따라서 대신에 우리는 다음과 같이, 근사된 친구들을 활용하여 종료 조건을 사용합니다.

$$
\Vert x_{n+1}-x_n\Vert \leq \text{eps}
$$

우리는 이번 포스팅의 종료 조건으로 이 근사를 이용한 종료 조건을 활용합니다.

근데 이거도 항상 통하는건 아니에요. 문제의 문맥에 따라 다릅니다. 경우에 따라서 종료 조건이 만족되더라도 $x_n$이 $x$로 수렴하지 않을 수도 있거든요.

반대로 $x_n$이 $x$로 수렴하더라도 종료 조건이 달성되지 않을수도 있구요.

따라서 수치적 과정에서는 언제나 종료 조건을 잘 선택하는 게 중요합니다. 물론 이번 포스팅에서는 그걸 깊게 고민해야할 이야기는 전혀 다뤄지지 않습니다.

그냥 일반론적인 이야기에요.

### 반복법의 구현

반복법은 "반복"이라는 단어가 들어가는만큼, 실제 구현에서 반복문, 특히나 `while`이 자주 사용됩니다. 

반복문의 조건으로 종료 조건을 달아놓고, 종료 조건이 만족될 떄까지 단계별로 계속 $x_n$을 계산하는거죠.

`for`의 경우 우리가 얼마나 반복해야 할지를 미리 알고 지정해야 하기 떄문에, 

그냥 종료 조건이 만족될 때까지 반복해야하는, 정확한 종료 시점을 모르는 상황에서는 일반적으로는 자주 사용되지 않습니다.

물론 경우에 따라서 $x_n$이 너무 $x$로 느리게 가까워져서 알고리즘이 끝없이 돌아가는 경우도 있는데요,

이를 피하기 위해서 반복문에 최대 반복 횟수를 지정하기도 합니다. 이 경우엔 `for`를 최대 반복 횟수를 제한하는 용도로 활용할 수도 있겠죠.

또한 반복법이 만들어내는 모든 숫자 $x_1,x_2,...$들을 그냥 다음 값을 만든 후 필요없으면 버리는 대신, 어디에 저장해놨다가

그걸 플롯하여 수렴이 어떤 형태로 진행되는지를 시각화해서 보기 위해 사용하기도 합니다. 결국에는 전부 사용자의 필요에 맞게 구현하면 됩니다.

간혹 좀 복잡한 문제가 주어지거나, 단순하더라도 문제 해결 과정에서의 유연성을 확보하기 위해서 `while` 대신에 제네레이터 함수를 사용하는 경우도 있습니다.

그냥 반복문으로 모든걸 해결하는 대신에, 제네레이터 함수를 이용하여 근사치들을 `yield`하여 만들어내는 형태로 구현해서,

종료 조건을 걸어둔 반복문에 그 제네레이터 함수를 넣어서, 종료 조건을 확인하는거죠.

이렇게 한다면 제네레이터 함수를 다른 컨텍스트에서도 사용하거나, 서로 다른 종료 조건에 대해 비교하는 것도 가능해집니다.


### 예시 : 이분법(Bisection method)

반복법의 가장 간단한 예시는 영점을 근사하는 이분법입니다. 이분법의 문제 상황은 $f$라는 함수의 $[a,b]$라는 구간 위에서의 영점을 찾는겁니다.

이를 위한 아이디어는 간단해요. 영점 정리에 의해서 만약 $f$가 연속이고, $f(a)f(b)<0$이라면 $(a,b)$에는 $f (x) = 0$인 $x$가 존재합니다.

우리는 이 $(a,b)$를 구간의 중점 $m:= \frac{a+b}{2}$에 대해 $(a,m)$, $(m,b)$로 말 그대로 이분 할 수 있는데요,

그러면 $f$의 영점은 $x = m$이거나, $(a,m)$에 있거나, $(m,b)$에 있을겁니다.

따라서 우리는 $f(a)(m)<0$인지, $f(m)f(b)<0$인지를 구분하여, 이렇게 함수값이 음수가 되는 이분된 구간으로 들어가서,

거기서 같은 과정을 계속해서 반복해서 영점의 위치를 찾는거에요.

이분법에는 종료 조건으로 $m$에 대해 $\vert f(m)\vert \leq \text{eps}$인지를 직접 함수값을 구해 판단하거나,

아니면 결국 계속 구간이 영점을 중심으로 작아지니, $b-a \leq \text{eps}$인지를 판단하는 방법을 사용합니다.

개인적으로는 함수를 호출할 필요가 없는 후자의 종료 조건이 더 좋아보이네요.

**주의할 점:** 보통 이분법에서 결국 우리가 영점의 후보로 생각하는건 각 단계에서 계산되는 구간의 중점 $m = \frac{b+a}{2}$입니다.

따라서 이분법을 마치 $m_{n+1} = F(m_n)$ 형태의 반복법으로 생각할 수 있는데, 이는 그렇게 정확하지는 않습니다.

위에서 다룬 $x_{n+1}=F(x_{n},...,x_1)$이라는 반복법의 형태는 정확히 말해서, 

$$
x_{n+1}\text{을 구하기 위해서 $x_1,...,x_n$을 사용할 수 있어야한다}
$$

라는 의미도 내포하고 있습니다. 이는 각 단계를 상태 변화라고 생각한다면, 다음 상태는 이전 상태로부터 만들어져야한다는거죠.

근데 만약 $m_{n+1} = F(m_n)$의 형태로 이분법을 생각한다면, 개념적으로 오류가 발생합니다.

우리는 $m_{n+1}$을 구하기 위해서 $m_n$을 사용하지 않으니까요. 즉, 여기서는 다음 상태, 이전 상태의 관계를 찾아보기가 너무 모호해집니다.

이전 상태는 중점인 $m_n$이 아니라, 우리가 $n$단계에서 축소시킨 구간 구간의 양 끝 점인 $a_n,b_n$이 보여줍니다.

따라서 이분법은 $x_n = (a_n,b_n)^T$라는 벡터에 대해서, 다음과 같이 반복하는 방법이라고 생각하는 게 더 정확합니다.
$$
(a_{n+1},b_{n+1}) = F(a_n,b_n)
$$

이런 입장에서 정확히 말하면 이분법은 $a = b$이고 $f(a) = 0$가 되는 $(a,b)$를 근사해내는 방법으로 생각할 수 있습니다.

뭐 이건 좀 이론적인 이야기구요, 실제 구현에서는 크게 신경쓰지 않으셔도 되긴 합니다.

우리는 위에서 설명한 구현 방법에 따라서 이분법을 `while`, `for`, 제네레이터 함수를 이용하여 모두 구현해보겠습니다.

우선 `while`입니다.

````python
def bisec_while(f,a,b, tol=1e-6):
    if f(a)*f(b) > 0  : # 만약 f(a)f(b)>0이라면 우리는 이분법을 사용할 수 없음. f(a)f(b) == 0인 경우는 이미 영점을 아는거니 생각하지 않음.
        return None
        
    while (b-a)/2 > tol: # 종료 조건의 부정이 들어와야합니다. 그래야지 종료 조건이 만족될 때 반복문이 종료됨
        m = (a+b)/2      # 현재 단계의 중점을 구함
        if f(m) == 0:    # 만약 중점이 영점이면 이걸 바로 반환하고 종료
            return m
        elif f(a)*f(m) < 0: # 중점이 영점이 아니지만, f(a)f(m) < 0이 되면, (a,m) 구간에 영점이 있다는거니 b = m로 업데이트하여 다음 단계를 진행
            b = m
        else:              # 반대로 f(a)f(m)>=0인 경우, f(m)f(b)<0이라는거니, (m,b) 구간에서 조사를 이어감
            a = m
    return (a+b)/2

from math import cos, pi
f = lambda x: cos(x)
a = 1
b = 2
print(bisec_while(f,a,b)) # 코사인 함수의 구간 (1,2)에서의 이분법을 이용한 영점 근사
print(pi/2)               # 코사인 함수의 구간 (1,2)에서의 해석적 영점의 수치적 값
````

간단합니다. 그냥 설명한 그대로에요. 어차피 주석보면 다 이해되실거라 별로 설명할 부분은 없어보이네요.

이제 그러면 제네레이터 함수를 한 번 활용해볼게요.

````python
def bisec_gen(f,a,b):
    while True:
        m = (a+b)/2 # 중점을 계산
        if f(a)*f(m)<0: # 중점을 기준으로 영점이 왼쪽에 있다면
            b = m       # 왼쪽 구간을 탐색해야됨
        else:
            a = m       # 오른쪽이라면 오른쪽 구간을 탐색
        yield a,b       # 제너레이터는 계속해서 다음 탐색 구간을 뱉어내는 형태로 정의
````

제네레이터는 말 그대로 어떤 수치를 생성하기 위한 규칙을 의미하며, 제너레이터 함수는 이 규칙을 함수의 형태로 구성시킨 놈을 의미합니다.

우리의 이분법의 입장에서 제네레이터는 함수 `f`, 시작 구간 `a,b`를 받아서, 각 이분법 단계별로 구간을 반환하는 형태가 되어야하므로

위와 같이 중점의 위치를 조사한 후, 그를 통해 업데이트된 새로운 구간의 끝 점 `a,b`를 반환하게 됩니다.

제너레이터는 그냥 저 다음 구간을 계속 생성하는 친구이기 떄문에, 종료 조건 이런걸 신경쓸 필요 자체가 없습니다.

따라서 만약 여러 종료 조건을 테스트하고 싶다면, 저 제네레이터 함수로 제네레이터를 하나 정의해서, 서로 다른 종료 조건을 비교할 수 있습니다.

````python
def bisec_len(f,a,b,tol=1e-6):
    g = bisec_gen(f,a,b)
    while b-a > tol: # 구간의 길이이 충분히 작아지면 종료
        a,b = next(g) # 제네레이터로 계속 다음 이분 구간을 생성
        
    return (a + b)/2  

def bisec_val(f,a,b,tol=1e-6):
    g = bisec_gen(f,a,b)
    m = (a+b)/2
    while abs(f(m)) > tol:  # 직점 중점 m의 함수값을 비교하여 충분히 0에 가까운 시점에 종료
        a,b = next(g)
        m = (a+b)/2
    return m

from math import cos, pi
f = lambda x: cos(x)
a = 1
b = 2
print(bisec_len(f,a,b))   # 구간 종료 조건을 이용한 이분법 근사치
print(bisec_val(f,a,b))   # 값을 판단하는 조건을 이용한 이분법 근사치
print(pi/2)               # 코사인 함수의 구간 (1,2)에서의 해석적 영점의 수치적 값
````

다시 설명하자면 종료 조건은 별개로 하여 제네레이터 함수는 반복법이 진행되며 생기는 값들만을 고려하는거죠.
    
이 방법은 사용자에 의해 주어진 종료 조건으로 근사를 마무리하든가, 아니면 제네레이터 함수를 다른 곳에 이용하는 등

말 그대로 유연성을 확보할 수 있다는 장점이 있습니다.

또 위에서 설명한대로 우리는 최대 반복 횟수를 지정해놓고, 그 만큼만 계산이 반복되도록 설정할 수도 있습니다.

최대 반복 횟수도 함수의 입력 변수로 남겨서 사용자가 지정하게도 할 수 있지만, 

이번에 우리는 최대 반복 횟수를 해석적으로 구할 수 있으므로 `math` 모둘의 함수들을 빌려와서, 직접 최대 반복 횟수를 구하는 방법도 다음과 같이 구현 가능합니다.

````python
from math import log2, ceil

# 간단히 최대 반복 횟수를 직접적으로 입력 받는 코드
def bisec_iter(f,a,b,tol=1e-6, maxIt = 10000): 
    for n,(a,b) in zip(range(maxIt), bisec_gen(f,a,b)):
        if b-a<= tol :
            return (a+b)/2
    return (a+b)/2

# 해석적인 관계를 활용해서 직접 최대 반복 횟수를 지정하는 코드
def bisec_smart_iter(f,a,b,tol=1e-6):
    N = ceil(log2(b-a) - log2(tol)-1 ) # 반복 횟수를 직접 계산
    for n,(a,b) in zip(range(N), bisec_gen(f,a,b)):
        pass
    return (a+b)/2

from math import cos, pi
f = lambda x: cos(x)
a = -1
b = 2
print(bisec_iter(f,a,b))
print(bisec_smart_iter(f,a,b))
````

## 익셉션(Exception)

만약 충분히 많은 반복 과정을 거쳤음에도 반복법이 수렴하지 않는다거나, 종료 조건이 만족되지 않거나, 더 이상 반복법 진행이 불가능한 경우에는

오류가 발생했다는 메세지를 생성하도록 코딩하는 게 유용할 수 있습니다.

파이썬에는 오류/익셉션(파이썬은 이 둘을 그리 엄격히 구분하지 않습니다. 파이썬에서 익셉션은 그냥 오류메시지가 아니구요, 객체입니다.

여러분들이 코딩 중에 보던 오류 메세지들도 이 익셉션이 발생한걸로 이해하실 수 있어요.

물론 익셉션은 파이썬이 미리 지정한 익셉션 뿐 아니라, 사용자에 의해 지정된 익셉션도 존재합니다.

예를 들어 우리는 이분법의 맥락에서 다음과 같이 사용자 지정에 의한 익셉션을 정의할 수 있습니다.

```python
class BisecError(Exception):
    def __init__(self, message):
        super().__init__(message)
```

간단합니다. 기존의 클래스를 정의하던 방법과 유사하나, 저기 보시다시피 괄호 안에는 이 친구가 익셉션 클래스임을 알리기 위해 `Exception`이 들어옵니다.

정확히 말하자면 저기 괄호 안에 `Exception`은, 우리가 지금 정의하는 클래스가 `Exception`이라는 상위 클래스로부터 상속받는 하위 클래스임을 지정하는겁니다.

객체 지향 프로그래밍에서 클래스의 상속은 결국 상위 클래스에서 있던 함수, 속성 등을 하위 클래스에 그대로 물려받겠다는거에요.

예를 들어서, 차량 관리 프로그램을 만든다면, 우리는 차량이라는 최상위 객체를 정의할 수 있습니다.

차량은 차량 번호, 제조사, 등록일 등등의 속성이 있다고 한다면, 차량의 하위 객체인 승용차, 화물차, 이륜차 등도 똑같이 번호, 제조사, 등록일이 있을거잖아요?

위와 같이 상속을 이용하면 그걸 일일이 각 하위 객체의 클래스에 정의할 필요 없이, 자동으로 차량이라는 객체의 속성이 그 하위 객체에도 남아있습니다.

따라서 저렇게 `Exception`으로부터 상속받는 클래스를 지정하면, 다른 일반적인 익셉션에 사용할 수 있는 속성이나 함수 등을 그대로 사용할 수 있습니다.

상속되는 대상인 함수에는 메소드들 뿐 아니라 생성자 `__init__`도 포함되고, 따라서 우리의 `BisecError`의 생성자도 그냥 저렇게

`super().__init__`, 즉 상위 클래스인 `Exception`의 생성자 `__init__`을 그대로 빌려와서 사용할 수 있습니다. 

익셉션이 발생했을 때 오류 메세지를 내보내는걸 우리는 익셉션을 `raise`한다고 합니다. 이는 다음의 코드를 보시면 확실히 이해가 되실겁니다.

````python
from math import cos, pi

class BisecError(Exception):
    def __init__(self, message):
        super().__init__(message)

def bisec(f,a,b,tol=1e-6):
    # 만약 두 구간의 양 끝점의 곱이 양수이면 영점 정리 활용이 불가능하죠.
    if f(a)*f(b) > 0:
        # 따라서 우리는 BisecError를 생성하여, 이를 raise 합니다! 결과물은 코드 실행 시 확인 가능
        raise BisecError("양 끝점의 함수값 부호가 같아서 영점이 있는지 판단이 안됩니당") 
                                                       
    # 아래는 통상적인 while을 이용한 이분법 코드
    while (b-a)/2 > tol:
        m = (a+b)/2
        if f(m) == 0:
            return m
        elif f(a)*f(m) < 0:
            b = m
        else:
            a = m
    return (a+b)/2

f = lambda x: cos(x)
a = 0
b = 50
print(bisec(f,a,b))
````

위 코드를 실행 시, $\cos(0)\cdot \cos(50)>0$이므로, `bisec()`이 실행되는 도중 `BisecError`가 `raise`되는데요,

이건 결국 `BisecError`가 속성으로 가지고 있는 스트링 `message`가 저렇게 에러 메세지로 출력되는걸 의미합니다.

이 `message`에서 당연히 이게 어떤 문제인지를 명확히 알려줄 수 있는 메세지를 적어야합니다.

근데 위 코드에는

```python
 if f(a)*f(b) > 0:
    raise BisecError("양 끝점의 함수값 부호가 같아서 영점이 있는지 판단이 안됩니당") 
```

$f(a)f(b)>0$이 되는 그 즉시 에러를 잡아내는데요,

이런 식으로는 이를테면 $[a,b]$의 구간 내부 어딘가에 영점이 존재하긴 하나, 하필 $f(a),f(b)$의 부호가 같은 경우도 모두 알고리즘이 종료되어버립니다.

그 경우에는 영점이 있는데도 불구하고 중간 단계에서 운이 나빠서 영점을 근사하지 못하게 되어버리죠.

따라서 우리는 $f(a)f(b)>0$가 만족되어, 원칙적으로는 계속해서 진행이 불가함에도 불구하고, 계속해서 일단 알고리즘을 진행해보고싶을 수 있는데요,

이 경우에 우리는 `try-catch`라는 방법을 사용할 수 있습니다.

이건 자바에서도 사용되는 익셉션을 거르는 방식인데요, 말 그대로 일단 코드를 `try`해보고, 그 과정에서 익셉션이 발생하는지를 `catch`하여,

이 익셉션이 발생하는 경우를 위한 조치를 따로 하자는겁니다.

이러한 `try-catch`는 우리의 이분법 알고리즘에서 다음과 같이 적용할 수 있습니다.

````python
def bisec_try(f,a,b,tol=1e-6):
    try:
        return bisec(f,a,b,tol) # 우선 위에서 정의한 bisec(f,a,b,tol)을 그대로 실행하여, 정상적으로 동작한다면 그 값을 반환함.
    except BisecError as e:     # 만약 BisecError가 발생한다면, 다음의 명령을 수행
        N = 10                  # 아이디어는 (a,b) 구간에서 a, b 외에 다른 점을 잡는것. 총 N 개의 점을 테스트할거임.
        d = (b-a)/N             # 이를 위해서 구간 (a,b)를 총 N개의 점(a는 미포함, b를 포함)으로 나눌 예정
        for i in range(1, N):   # 그러면 a+d, a+2d, ... , a+(N-1)d의 점들이 생길텐데, i는 이 중 i번째 점을 호출하기 위한 인덱스
            b = a + i*d         # 우리는 f(a)f(b)<0 인지의 여부를 따지는 대신, i번째 점 a + id에 대해 f(a)f(a+id)<0인지의 여부를 따짐.
            try:                # 이를 다시 try함
                return bisec(f,a,b,tol) # 한 점이라도 f(a)f(a+id) <0라면, 정상적으로 다음 과정이 수행되어 값이 반환됨.
            except BisecError as e: # 만약 f(a)f(a+id)>0가 성립되는 경우,
                pass                 # 아무거도 안하고 다음 i에 대해서 f(a)f(a+id)<0인지를 가림

        # 만약 어떤 점에서라도 부호가 같았다면, 어쩔 수 없이 그냥 에러 메세지를 출력
        raise BisectError("함수값의 부호가 같아지는 이분 구간이 존재하는데, 대응책도 안먹힙니당")


#우리가 세운 대책(f(a)f(b)>일 경우, b 대신 (a,b)구간의 10개의 점을 생성해서, 그걸로 부호를 다시 판단)
f = lambda x: cos(x)
a = 0
b = 50
print(bisec_try(f,a,b))
````

복잡해보일 수 있는데, 그냥 말 그대로 `try`로 특정 명령어를 시도하고,  익셉션을 캐치하는데요,

자바에서는 이를 위해 그대로 `catch`라는 키워드가 사용되지만, 파이썬에선 `except`를 사용합니다.

말 그대로 `except BisecError as e`는 `try`의 코드를 `BisecError`가 발생하지 않는 한 실행하다가, 

발생하는 경우 그 이하의 코드를 실행하라는 의미입니다. 이 때, `e`는 혹여나 이 익셉션 인스턴스를 사용할 일이 있는 경우,

그를 호출하기 위한 변수 이름, 식별자로 이해할 수 있습니다. 위의 코드에서는 굳이 이 익셉션을 그대로 가져와서 사용할 일은 없어서 `e`는 사용되지 않습니다.

## 영점을 찾는 다양한 반복법들

반복법은 다양하게 있는데요, 우리는 이분법에 이어 이번엔 좀 간단한 레벨에서 다양한 반복법을 다뤄보려고 합니다.

### 뉴턴 방법(Newton's Method)

뉴턴 방법은 반복법의 가장 대표적인 친구로, 함수 $f$의 영점을 찾는 이분법보다 훨씬 더 빠르게 수렴하는 좋은 방법입니다.

이 방법의 핵심은 미분 가능성이에요. 함수 $f$가 어떤 점 $x_0$에서 미분 가능하다는건 \[해석학 1,2\]에서도 다룬대로, 

$x_0$점에서 $f$를 근사하는 아핀 선형 함수 $\tilde f(x) = a(x-x_0) + f(x_0)$가 존재함과 같은 의미였습니다. 

이 떄, $a$를 $f$의 $x_0$에서의 미분 계수라고 하고 $f'(x_0):= a$라고 표기했었죠.

즉 우리는 다음의 근사적인 관계를 알고 있습니다.

$$
f(x) \approx f(x_0) + f'(x_0)(x-x_0)
$$

이를 이용하여 이렇게 저렇게 오밀 조밀하면 우리는 다음의 형태로 $x_1,x_2,...$를 구성하여 함수 $f$의 영점을 근사해나갈 수 있습니다.

$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$

우리는 $\vert x_{n+1}-x_n\vert \leq \epsilon$이 되는 $n$을 찾거나, $f(x_{n+1})$이 0에 충분히 가까운 등의 종료 조건을 사용합니다.

뉴턴 방법은 다음과 같이 구현될 수 있습니다.

````python
def newton_gen(f,df,x0): # 제네레이터 함수로 구현합니다. f는 함수, df는 f의 도함수, x0은 시작점
    while True:
        x0 = x0 - f(x0)/df(x0)
        yield x0

def newton(f,df,x0,tol=1e-6,maxIt=100):
    g = newton_gen(f,df,x0)
    x = x0
    while abs(f(x)) > tol and maxIt > 0:
        maxIt -= 1
        x = next(g)
    return x


from math import sin, cos
f = lambda x: cos(x)
df = lambda x: -sin(x)
x0 = 1.5
print(newton(f,df,x0))

f = lambda x: sin(x)**3
df = lambda x: 3*sin(x)**2*cos(x)
x0 = 0.5
print(newton(f,df,x0))
````

### 할선법(Secant method)

할선법은 뉴턴 방법의 일종이라고도 할 수 있습니다. 뉴턴 방법은 다음의 방법으로 근사를 계속해나갔는데요,
$$
x_{n+1} = x_n - \frac{f(x_n)}{f'(x_n)}
$$
이 떄문에 위에서의 구현 과정에서도 보시다시피, 우리는 $f$의 도함수 $f'$를 알고 있어야 했습니다.

근데 이 전제 조건 자체가 생각보다 강합니다. 우리는 해석적으로 미분이 불가능한 경우와 마주칠 수도 있고, 계산 비용도 추가로 드니까요.

따라서 우리는 직접 도함수를 구하는 대신, 평균 변화율을 이용하여 순간 변화율(미분계수)를 근사할 수 있는데,

$$
f'(x_n)\approx \frac{f(x_n)-f(x_{n-1})}{x_n-x_{n-1}}
$$

이를 이용하여 다음과 같은 방법으로 영점을 근사하는 방법을 할선법이라 합니다.
$$
x_{n+1} = x_n - f(x_n) \frac{x_n - x_{n-1}}{f(x_n) - f(x_{n-1})}
$$

뉴턴 방법의 일종이므로, 종료 조건도 비슷하고 구현도 비슷하게 가능합니다.

### 가위치법(Regula Falsi)

가위치법은 이분법과 할선법을 섞었다고 볼 수 있는 반복법입니다.

아이디어는 이분법처럼 연속인 $f$에 대해 $f(a)f(b)<0$인 경우, $(a,b)$에 $f$의 영점이 존재한다는 것에서 출발하지만,

값을 업데이트하는 방식에서는 할선법의 방식을 이용합니다.

가위치법은 다음과 같은 계산 과정을 이용합니다.


$$
c = a - f(a) \frac{b - a}{f(b) - f(a)}
$$
$f(a) \cdot f(c) < 0$인 경우, 다음 단계는 $[a, c]$ 에서 진행되며, 반대의 경우엔 $[c, b]$에서 진행됩니다.

## 수렴 속도

수치적 방법에서 수렴 속도는 중요합니다. 수렴 속도는 말 그대로 수열 $(x_n)_{n\in\mathbb{N}}$이 $x$로 수렴하는 속도를 의미합니다.

수렴 속도가 느리면 다시 말해서 우리가 $x$를 충분히 잘 근사한다고 생각할 수 있는 $x_N\approx x$을 계산하는대에 많은 단계가 필요하다는 의미이므로,

당연히 수렴 속도는 빠르면 빠를수록 좋습니다.

우리는 반복법에서 $n$번째 단계의 근사인 $x_n$과 $x$의 오차인 $e_n := x_n - x$의 점근적인 양상으로 수렴 속도를 분석할 수 있습니다.

적당한 양수 $q\geq 1$에 대해서, 만약 다음을 만족하는 상수 $C>0$가 존재한다면,

$$
\limsup_{n \to \infty} \frac{|e_{n+1}|}{|e_n|^q} = C
$$
우리는 해당 반복법이 $q$차 수렴하는 방법이라고 합니다. 저건 다시 다음으로도 적어줄 수 있는데요,

$$
\exists N\in\mathbb{N} : \forall n\geq N : \vert e_{n+1}\vert \leq C\cdot \vert e_n\vert^q
$$

이건 의미상 특정 시점 $N$ 이후부터는 오차가 전 단계의 오차의 $q$승보다도 더 적어진다는 의미입니다.

이 오차는 보통 1보다 작은 매우 작은 수거든요? 따라서 저 관계가 성립할 떄, $q$가 크면 클수록 오차가 더 빨리 줄어들게 됩니다.

따라서 $q$가 높은, 고차의 반복법이 가장 수렴 속도가 빠른 방법으로 고려됩니다.

$q = 1$인 경우 우리는 선형 수렴, $q >1 $인 경우에는 $q-$차 수렴하는 반복법이라고 표현합니다.

이분법은 선형 수렴하는 반복법의 대표적인 예시입니다. 그 이유는 간단하죠. 모든 반복 과정에서 구간이 정확히 절반으로 나뉘니까요 :

$$
\vert e_{n+1}\vert \leq \frac{1}{2} \vert e_n\vert
$$

물론 이런 이야기는 모두 반복법에서 우리가 근사해야하는 $x$가 알려진 경우를 전제하고 있습니다.

수렴 속도를 계산하는대에도 수치적 방법이 사용될 수 있습니다. $x$가 알려진 전제에서 우리는 각 $n$번째 오차인 $e_n$을  플롯할 수 있고,

인접한 플롯의 점간에 선분을 이을 수 있는데요, 그렇게 만들어진 선분들의 기울기가 결국 $q$에 대한 수치적 근사가 됩니다.

여튼 위에서 우리는 결국 수렴 차수 $q$가 높으면 높을수록 좋은 반복법임을 다뤘었는데요,

사실 이 수렴 차수는 다음을 만족하는 반복법들간의 속도적 우열을 가리기 위한 개념입니다.

$$
\lim_{n \to \infty} \frac{|e_{n+1}|}{|e_n|} = 0.
$$

선형 수렴처럼 오차 비율이 수렴하는 $C>0$가 존재하는 대신, 완전히 0으로 수렴해버리는거죠. 이를 우리는 초선형(super linear) 수렴하는 반복법이라고 합니다.

이러한 초선형 수렴하는 반복법이 주어지는 경우, 우리는 기존의 $x$를 알고 있어야 선형 수렴 속도를 이야기할 수 있다는 전제로부터 탈피할 수 있습니다.

일련의 간단한 과정을 거치면 $\vert x_{n+1}-x_n\vert \approx \vert x_n - x\vert$가 되거든요.

따라서 우리는 $\vert x_n - x\vert$ 대신, $\vert x_{n+1} - x_n\vert$을 종료 조건으로 사용할 수 있다는 장점이 있습니다.

수렴 속도는 다음과 같이 플롯을 한다면 더 눈에 확 들어옵니다.

````python
def bisect_iterands(f,a,b,tol=1e-6):
    g = bisec_gen(f,a,b)
    A = [a] # 플롯을 위해서 이번에는 각 부분 구간들을 저장합니다.
    B = [b]
    while b-a > tol:
        a,b = next(g)
        A.append(a)
        B.append(b)
    return A,B

def newton_iterands(f,df,x0,tol=1e-6,N=100):
    g = newton_gen(f,df,x0)
    X = [x0]  # 마찬가지로 각 근사치들을 저장
    x = x0
    while abs(f(x)) > tol and N > 0:
        N -= 1
        x = next(g)
        X.append(x)
    return X

# 이 조건으로 이분법과 뉴턴 방법을 적용해서 비교할 예정
from math import sin, cos
f = lambda x: cos(x)
df = lambda x: -sin(x)
x0 = 2.2
a = 1
b = 2

# 플롯 코드
import matplotlib.pyplot as plt
from math import pi

ni = newton_iterands(f, df, x0)  # 뉴턴 근사치로 이뤄진 float들의 리스트
bi = bisect_iterands(f, a, b)    # 각 단계의 구간 양 끝점으로 이뤄진 리스트들의 투플

# 이분법의 중점 및 오차 계산
midpoints = [(bi + ai) / 2 for ai, bi in zip(bi[0], bi[1])]
bisect_errors = [abs(pi/2 - m) for m in midpoints]

# 뉴턴 오차 계산
newton_errors = [abs(pi/2 - n) for n in ni]

plt.scatter(range(len(bisect_errors)), bisect_errors, label='bisect')
plt.scatter(range(len(newton_errors)), newton_errors, label='newton')
plt.yscale('log')
plt.legend()
plt.show()
````

이분법은 설명한대로 선형 수렴하는데, 뉴턴 방법은 이차 수렴한다는 게 알려져있습니다.

우리의 이론대로면 뉴턴 방법이 훨씬 빠르게 수렴해야하는데, 실제로 위 플롯을 통해서 우리는 오차가 훨씬 빠르게 줄어든다는걸 알 수 있어요.

근데 그렇다고 이 수렴 속도가 절대적인건 아닙니다.

````python
# 이번엔 다른 함수에 각 반복법을 적용하여 비교해봄.
from math import sin, cos
f = lambda x: sin(x)**3
df = lambda x: 3*sin(x)**2*cos(x)
x0 = 0.5
a = -1
b = 2

import matplotlib.pyplot as plt

# 위에서 사용한 그 함수 그대로 사용함
ni = newton_iterands(f, df, x0)
bi = bisect_iterands(f, a, b)   

# 똑같이 플롯할 점들을 계산 및 저장
midpoints = [(ai + bi) / 2 for ai, bi in zip(bi[0], bi[1])]
bisect_errors = [abs(m) for m in midpoints]

newton_errors = [abs(n) for n in ni]

plt.scatter(range(len(bisect_errors)), bisect_errors, label='bisect')
plt.scatter(range(len(newton_errors)), newton_errors, label='newton')
plt.yscale('log')
plt.legend()
plt.show()
````

위 경우에는 오히려 뉴턴 방법이 이분법보다 훨씬 느리게 오차가 줄어드는걸 알 수 있는데요,

이는 우리의 근사 대상이 되는 함수 $f(x) = \big(\sin(x)\big)^3$가 $x = 0$의 근처에서 매우 작은 미분계수를 가지기 떄문입니다.

이걸 좀 더 정확하게 얘기하려면 더 복잡한 이야기가 필요해서, 그냥 이런 경우도 있다는 정도로만 알고 넘어가셔도 충분합니다.

## 그래디언트법(Gradient method)

위에서는 영점을 찾는 반복법들이 다뤄졌었는데요, 그래디언트법은 $f$라는 함수를 최소화하기 위한 반복법입니다.

기본적인 문제 상황은 $f: \mathbb{R}^n\to \mathbb{R}$이라는 함수의 최소값의 위치를 $x_n$을 구성하여 근사적으로 찾아보자는거에요.

그래디언트법은 이름 그대로, $n+1$단계의 근사 $x_{n+1}$를 구하기 위해 $f$의 그래디언트 $\nabla f$와 전단계의 $x_n$를 활용합니다.

$-\nabla f(x_n)$은 기하학적으로 그 $x_n$이라는 지점으로부터 가장 가파르게 줄어드는 방향을 가리키는 벡터가 되는데요,

이를 이용해서 다음과 같은 형태로

$$
x_{n+1} = x_n - \alpha\cdot \nabla f(x_n)
$$

현재의 점 $x_n$을 기준으로 $-\nabla f(x_n)$의 방향으로 $\alpha$라는 가중치/보폭만큼 움직인 점을 근사치로 잡아 업데이트를 반복하자는거죠.

이 방법은 $f$가 볼록한 함수이고, $\alpha$가 충분히 작은 경우에 성립됨이 알려져있습니다.

1차원의 경우에서 $\nabla f(x_n) = f'(x_n)$인데요, 만약 $\alpha := \frac{1}{f''(x_n)}$을 잡는다면, 그래디언트 방법은 다음의 형태로 업데이트됩니다.

$$
x_{n+1} = x_n -\frac{1}{f''(x_n)}\cdot f'(x_n)
$$

이거 어디서 많이 본 모양이죠. 만약 $g := f'$라고 한다면, 위 식은 다시 다음으로 바뀝니다.
$$
x_{n+1} = x_n - \frac{g(x_n)}{g'(x_n)}
$$

즉, 그래디언트 방법은 결국 $f$의 도함수 $g = f'$의 영점을 찾는 뉴턴 방법이랑 다를바가 없습니다.

이는 실제로 우리가 알고 있는 "극점에서는 미분계수가 0이 된다"라는 성질 또한 잘 보여주는 수치적인 결과라고 할 수 있습니다.

(물론 실제 최소값으로 가는지는 좀 더 체크가 필요합니다. 아이디어상 그렇다는거죠.)

그 밖에도 뉴턴 방법과 그래디언트 방법은 끈끈한 관계가 있는데요, 이에 관한 이야기는 \[수치해석학\], \[수치적 선형대수학\]에서 더 자세히 다뤄집니다.
