# 파이썬 기초 3 : if 조건문과 함수

## if 조건문

`if` 조건문은 파이썬을 비롯한 프로그래밍 언어의 가장 기본적인 컨트롤 구조입니다.

`if` 조건문은 특정 조건이 만족되었을 때, 특정 코드를 수행하라는 명령을 위해서 사용됩니다. 

파이썬에서의 일반적인 신택스는 다음과 같습니다 :

```python
if 조건 :
    # 조건이 True일 경우 수행 될 명령어들
    # ...
```

간단합니다. 만약 위 `조건`의 위치에 있는 부울 표현이 참이 되는 경우, 그 아래 코드 블록에 있는 명령이 수행되며, 만약 거짓일 경우에는 해당 블록이 무시됩니다.

만약 `조건`이 거짓일 경우 수행할 다른 명령을 설정하고 싶다면, 이는 `else`를 이용할 수 있습니다.

```python
if 조건 :
    # ...
    # ...
else :
    # ...
    # ...
```

설명한대로 `else`는 `조건`이 거짓일 경우에만 발동됩니다. 그리고 `조건`이 거짓일 떄에는 `if`의 블록은 당연히 무시됩니다.

또한 `elif`를 이용하면, `조건`이 거짓인 경우, 추가적인 조건을 이용하여 구분된 명령을 할 수 있습니다.


```python
if 조건 1:
    # ...
    # ...
elif 조건 2:
    # ...
    # ...
else:
    # ...
    # ...
```
간단합니다. `elif`는 "else if"의 준말로, 만약 `조건 1`이 거짓이고, `조건 2`는 참일 때 그 위치의 코드 블록을 작동시킵니다.

그리고 이 때 `else`는 만약 `조건 1`과 `조건 2`가 모두 거짓인 경우 발동됩니다.

이런식으로 복수의 조건을 거는 조건문도 사용할 수 있습니다.

## 들여쓰기(Indentation)

코드에서 들여쓰기는 엄청 많이 사용됩니다. 들여쓰기는 다음을 의미합니다.

```python
x = 10
    y = 20 # 들여쓰기 된 행

```

들여쓰기는 `Tab`키 혹은 공백을 4번 입력하는 방법을 사용하여 하실 수 있는데요, 보통 다른 프로그래밍 언어에서 이 들여쓰기는 단지 인간들이 그 코드를 볼 때 가독성을 높이기 위한 선택적인 조치입니다. 보통 코드 블록 안에서 이렇게 들여쓰기를 하는데요, C나 자바에서는 이런 들여쓰기를 하지 않아도 컴파일 시 어떠한 문제도 발생하지 않습니다.

이에 비해서 파이썬에서는 이 들여쓰기가 필수이며, 항상 코드 블록이 등장할 때는 들여쓰기를 해줘야합니다.

예를 들어서 위에서 사용한 `if` 조건문으로 다음 코드를 생각하시죠.

```python
x = 101
if x > 10:
    print("x > 10")
else:
    print("x <= 10")
```

`if`의 이후에는 조건 `x>10`가 만족되는 경우에 수행할 명령이 등장하는데요, 이 명령들에 한 번 들여쓰기를 해줘야합니다.

만약 이 블록 안에 또 하나의 블록이 들어오는 경우, 이 떄는 또 다시 한 번 더(총 두 번) 들여쓰기를 해야합니다.

이 이유는 간단합니다. 다른 프로그래밍 언어는 이렇게 코드 블록의 레벨을 나눌 때, 중괄호 `{ }`를 사용하는데요, 파이썬에서는 이를 사용하지 않습니다. 그 대신에 컨벤션이던 들여쓰기를 필수로 강제함을 통해서 코드 블록의 레벨을 나눕니다.

만약 위 코드를 `if`의 블록의 들여쓰기를 없애고 실행한다면, 오류가 발생합니다.

````python
x = 101

if x > 10:
print("x > 10")
else:
    print("x <= 10")
````

이러한 들여쓰기는 `if` 뿐 아니라 함수, 반복문 등 코드 블록이 등장하는 모든 부분에서 필수적으로 지켜져야합니다.

## 함수

함수는 부분 프로그램으로 생각할 수 있는데요, 하나의 커다란 본 프로그램을 작동시키는 과정에서 사용하는 특정 기능들을 모은 놈들이라고 할 수 있습니다. 함수는 프로그램의 작동 도중 제한없이 사용 가능하며 복잡한 계산이나 기능을 단순하게 사용하기 위한 역할을 합니다.

### 함수의 정의

함수는 키워드 `def`를 이용하여 정의할 수 있습니다. 다음과 같이요.

```python
def f(parameters) :
    # 함수가 수행할 명령들
```

이 때, `f`는 해당 함수를 호출할 때 사용할 이름, `parameters`에는 이 함수에 입력되는 입력 변수들을 적을 수 있습니다. 만약 입력 변수가 필요 없는 경우에는 그냥 저 `parameters`의 칸을 비워두면 됩니다.

여기서 주목할 점은, 매개변수의 목록에서 우리는 매개 변수의 타입을 굳이 지정하지 않는다는겁니다. 파이썬에서는 매개 변수의 타입을 지정하지 않으므로, 혹시나 잘못된 매개변수가 들어가는 경우, 오류를 발생시킬 수 있습니다 :

````python
def abso(x):
    if x < 0:
        return -x
    else:
        return x

print(abso(-1))
````

````python
print(abso("asd"))
````

그리고 반대로, 만약 우리가 생각치도 않은 입력 변수들이 입력되었으나, 그 함수의 코드를 실행하는 과정에서 문제가 없다면, 그 입력 변수에 대해서도 함수 사용이 가능합니다.

````python
def add_numbers(a,b) : # 누가 봐도 숫자를 더하는 함수
    return a+b
add_numbers("이거도 ", "더해진다고?")
````

만약 함수에 복수의 매개변수들이 들어오는 경우, 이 매개변수들을 `,`로 구분하여 적어주면 됩니다.

`if`에서와 마찬가지로 이 함수가 수행하는 명령어들도 들여쓰기를 하여 구분해줘야합니다.

### 함수의 반환값

함수는 값을 반환할 수 있습니다. 이를 위한 키워드는 `return`이고, 대충 다음과 같이 사용됩니다.

```python
def f(parameters) :
    # 함수가 수행할 명령들
    return returnvalue
```

`returnvalue`의 위치에는 저 함수의 결과로서 반환할 값을 지정합니다.

물론 값을 반환하지 않는 함수들도 있는데요, 이러한 함수들은 `returnvalue`칸을 그냥 공란으로 두거나, 그냥 애초부터 `return`이라는 키워드를 사용하지 않습니다. 정확히는 반환을 하지 않는다기보다는, `None`이라는 특수한 객체를 반환합니다.

이 반환이 없는 함수는 그럼 어따 써먹을까요?

### 함수의 효과

많은 함수들은 반환값보다, 그 함수를 소환함으로서 발생하는 효과들을 위해 사용됩니다. 예를 들어 우리가 계속 만만하면 사용하던 함수 `print()`의 경우, 반환을 하지 않는(`None`을 반환) 함수인데, 대신에 콘솔에 입력된 매개변수들을 출력하는 역할을 합니다.

마찬가지로, 특정 데이터를 불러온다거나, 사용자로부터 값을 받아오는 등의 함수들도 이러한 함수들에 해당됩니다.

물론 반환값이 있는 함수들도 이러한 효과가 있는 경우도 있습니다. 효과 유무가 반환값 유무는 아니에요.

파이썬에도 당연히 기본적으로 표준 라이브러리에 미리 정의되어 있는 함수들이 있습니다. 이 놈들은 따로 다른 조치 없이 사용이 가능한데요, 대부분 빠르게 특정 단순한 계산 등을 수행하기에 적합한 놈들이 많아서 유용합니다. 지금 얘기할만한 놈은 예를 들어 다음과 같습니다.

- 기타 등등

### 함수 이름의 규칙

함수의 이름도 변수와 비슷하게, 영문 대/소문자, 숫자, 언더스코어 `_`로 구성되며, 예약어는 사용이 불가능합니다. 변수와 달리 대문자 사용이 허용됩니다.

그건 파이썬이 정한 규칙이구요, 실제로 프로그래밍시에 함수들은 항상 그 역할을 확실히 보여주는 강력한 이름을 지어줘야 한다는 점과, 대문자도 가능하나, 일반적으로는 소문자로만 함수의 이름을 구성시킨다는 컨밴션이 있습니다.

함수에 관한 더 다양한 이야기는 우선 두 가지 파이썬에서 중요한 데이터 타입을 더 다루고 해볼게요.

## 문자열 (str)

문자열은 말 그대로 문자들의 열입니다. 문자열은 텍스트를 표현하기 위해 사용되며, 큰 따옴표 `"텍스트"`, 혹은 작은 따옴표 `'텍스트'`로 표현됩니다. 또한 문자열은 `+`와 `*`연산을 지원합니다.

예시부터 보시죠.

```python
x = "Hello" 
y = "Hello" + " " + "World"  # 순차
z = "Hello" * 3  # "Hello"를 3번 반복적으로 순차
```

간단합니다 `+`는 이미 아시는 문자열을 이어 붙이는 순차이구요, `*`는 이러한 순차를 여러 번 반복하는 연산입니다.

수학이랑 똑같죠. 수학에서 곱셈은 결국 덧셈을 여러 번 하는거니까요.

문자열은 우리가 사용하는 거의 모든 문자로 구성시킬 수 있습니다. 여기에는 줄 바꿈 등을 표현하는 포맷 문자도 포함됩니다.

```python
x = "Hello, World!"
y = "Hello\nWorld!" # Hello 후 새로운 행을 열어서 World가 오는 형태
z = "Hallo\tWorld!" # Hello 후 들여쓰기를 한 후 World가 오는 형태
```

문자열은 각 단일 문자들을 이어붙인 구조로 이해할 수 있습니다. 즉, 문자열 `x`의 `i`번째 문자는 `x[i]`의 형태로 불러올 수 있습니다.

```python
x = "Hello"
x[0]  # == "H"
x[2]  # == "l"
```

파이썬에서도 인덱스가 0부터 출발하므로, `i`번째 문자는 우리가 보는 기준으로 문자열의 `i+1`번째 문자가 됩니다.

저렇게 인덱스로 문자열의 단일 문자를 호출하는 경우, 그 결과물도 다시 문자열(하지만 길이가 1인)이 되므로, 이들에 다시 연산을 사용할 수도 있습니다 :

```python
x = "Hello"
x[0] + x[4]  # == "Ho"
x[2] == x[3]  # == True
```

또 저렇게 단일 문자만 뽑아오는 게 아니라, 부분 문자열(문자열을 이루는 놈들 중 문자열 형태인 놈들)을 추출할 수도 있습니다.

이를 슬라이싱(Slicing)이라고 하는데요, `x[i:j]` 문자열 `x`의 `i`번째부터 `j-1`번째 문자들로 구성된 부분 문자열을 반환합니다.

만약 `i`가 생략된다면, `x[:j]`는 처음부터 `j-1`까지의 부분 문자열이 반환되며, 마찬가지로 `j`를 생략한다면 `x[i:]`는 i번쨰부터 끝까지의 부분 문자열을 반환합니다.

````python
x = "Hello"
print(x[1:3] + "\n")
print(x[:3]+"\n")
print(x[3:]+"\n")
x[:]
````

또한 위와 같이 `x[:]`로 아무런 인덱스도 넣지 않는다면 전체 스트링이 반환됩니다.


파이썬은 C나 자바와 달리, **음수의 인덱스**도 허용을 해줍니다.

아이디어는 간단하게, 0번째 인덱스가 결국 첫번쨰 문자니까, 거기서 더 왼쪽으로 간다면, 다시 가장 오른쪽의 문자로 간다는겁니다.

예를 들어, `x[-i]`는 문자열 `x`를 가장 오른쪽을 기준으로 `i`번째 문자를 반환합니다.

````python
x[-3] # x의 오른쪽으로부터 3번째 숫자, 즉 l
````

슬라이싱도 똑같이 가능합니다. `x[-i:]`은 `x[-i]`로부터 출발하여, 가장 마지막까지의 문자들을 포함한 부분 문자열을 반환합니다.

```python
x = "Hello"
x[-1]  # == "o" 
x[-3:]  # == "llo", -3번째 숫자인 ㅣ(첫번째 l)부터 출발하여, 마지막까지의 문자들을 포함한 부분 문자열을 반환
```

나머지 슬라이싱도 마찬가지 방법을 이용하여 가능합니다.

여기는 규칙성이 좀 있는데요, 결국 위 `x = "Hello"`를 기준으로 `x[-1] == x[4]`의 관계가 성립함을 알 수 있습니다. 그래서 만약 마이너스 인덱스가 등장한다면, 스트링의 길이를 더해서 우리가 편한 양수 인덱스로 전환해서 생각하시면 간단합니다.


문자열을 구성하는 각각의 문자를 직접적으로 변경할 수는 없습니다. 예를 들어서

```python
x = "Hello"
x[0] = "h"
```

이런식으로 `"H"`를 `"h"`로 바꾸려고 시도한다면, 에러가 발생합니다. 한 마디로 문자열의 경우 `x[i]`의 호출은 읽기 전용입니다.

대신, 우리는 슬라이싱을 활용하면 다음의 방법으로 문자를 바꿔줄 수 있습니다 :

```python
x = "Hello"
y = "h" + x[1:]  # "h" + "ello"
```


## 문자열을 위한 함수와 메소드

문자열을 위한 함수와 메소드는 매우 많습니다.

이 때, 메소드라는 개념은 함수와 역할은 동일하나, 이 놈이 실행되는 방식이 구체적인 객체에 직접적으로 참조하는 방식이라는대에서 차이가 있는 그런 놈입니다. 함수는 `f(x,y)`이런식으로 함수에 객체 `x,y`를 직접 값을 넣는 형태인데요, 메소드는 대신에 `x.f()`이런 식으로, 자바에서 오브젝트 메소드를 사용하듯, 구체적인 객체 `x`를 참조하여 `f()`를  수행합니다. 

비유적으로 메소드 `x.f()`는 `x`가 주체적으로 `f()`를 수행하는겁니다.

문자열에 자주 사용되는 함수는 대표적으로 `len()`이 있습니다.

```python
x = " Hello "
len(x)  # 7이 반환됨.
```
이 함수 `len()`은 입력된 문자열의 길이를 반환합니다. 이 때, 길이는 포맷 기호, 띄어쓰기 등을 모두 포함하여 카운트됩니다.


그리고 메소드는 다음과 같은 놈들이 있습니다.

```python
x.upper()  # x의 모든 글자를 대문자로 변환하여 반환, 기존의 스트링은 변경하지 않음.
x.lower()  # x의 모든 문자를 소문자로 변환하여 반환, 기존의 스트링은 변경하지 않음.
x.strip()  # x의 양 끝의 공백 문자(white space, 줄바꿈 \n, 들여쓰기 \t, 띄어쓰기 등)을 제거하여 반환
x.replace("H", "h")  # x에 들어있는 모든 문자 "H"를 "h"로 바꿔서 반환
```

x = [1, 2, 3]

스트링이 들어있는 리스트는 `join()`메소드를 사용할 수 있습니다.

````python
x = ["Hello", "World","!"]
print(" ".join(x))  # "Hello World !"가 반환됨.
print("".join(x))   # "HelloWorld!"가 반환됩니다.
print("and".join(x))   # "HelloandWorldand!"가 반환됩니다.
````

간단합니다. `join()`메소드는 괄호 안에 들어오는 스트링으로 구성된 리스트를 받아서, 그 리스트에 들어 있는 스트링들을 이어 붙입니다.

이 때, 메소드를 위해 참조되는 스트링(위 예시의 `" "`, `""`와 `"and"`)은 리스트의 스트링들 사이에 들어오게 됩니다.


반대로 우리는 스트링을 리스트로 만들 수도 있습니다. 이는 `split()`메소드를 사용하시면 됩니다.

````python
y = "Hello World"
y.split()
````

`split()`메소드는 입력 매개변수(괄호 안에)로 스트링을 받는데요, 입력된 스트링을 기준으로 해당 스트링을 분리합니다. 그리고 입력된 스트링은 참조되는 스트링이 리스트로 변환되는 과정에서 사라집니다.

### 슬라이스(Slice)

슬라이스의 주된 용도는 리스트의 일부분을 추출하는 것입니다.

위에서 스트링을 다루며 봤던대로, `x[i:j]`라는 표현은 리스트 `x`의 `i`번째 성분부터 `j-1`번째 성분까지의 부분 리스트를 반환합니다.

`j`가 생략된 표현 `x[i:]`는 `i`번째 성분부터 끝까지의 성분을 포함한 부분 리스트를 반환하고,

`i`가 생략된 표현 `x[:j]`는 처음부터 `j-1`번쨰 성분까지의 부분 리스트를 반환합니다. 

만약 `i,j`가 모두 생략된다면, `x[:]`는 `x` 그 자체를 그대로 반환합니다.

```python
x = [1, 2, 3, 4, 5]
x[1:3]  # [2, 3] 
x[:3]  # [1, 2, 3]
x[3:]  # [4, 5] 
```

스트링에서와 마찬가지로 음수 인덱싱이 허용됩니다. 규칙도 똑같아요. `x[-i]`는 `x`의 마지막에서 `i`번째(인간이 보는 기준과 동일) 성분을 반환합니다.

특히나 리스트에서는 성분 출력의 간격이 중요할 경우가 많은데요, 만약 `x`의 `i`부터 `j-1`까지의 성분을, `k`라는 간격으로 뽑은 부분 리스트를 반환시키고 싶은 경우, `x[i:j:k]`로 슬라이싱이 가능합니다. 이는 스트링에서도 가능하지만, 보통 리스트에서 많이써서 리스트에 넣었어요. 이렇게 하면 `x[i:j:k]`는 대충 `x`의 `i`번째, `i+k`번째 , `i+2k`번째, ... 성분들을 `j-1`번째 이하의 성분까지 반환합니다.

```python
x = [1, 2, 3, 4, 5]
x[-1]  # 5가 반환됨.
x[-3:]  # [3,4,5]가 반환됨.
x[::2]  # [1,3,5]가 반환됨.
```

슬라이스는 리스트의 부분을 변경하거나, 부분을 삭제하기 위해서도 사용 가능합니다.

```python
x = [1, 2, 3, 4, 5]
x[1:3] = [10, 20]  # x는 [1, 10, 20, 4, 5]가 됨. 1:3, 즉 1번쨰, 2번째, 인간의 보는 기준 2번째 3번째 성분들이 각각 10과 20으로 변경됨.
x[1:3] = []  # x는 [1, 4, 5]가 됨. 마찬가지로 1번쨰, 2번째 성분들을 삭제함.
```

우리는 파이썬으로 수학적인 작업을 많이 할거라서, 정수로 이뤄진, 간격이 동일한 수열을 생성하는 경우가 많습니다.

이 때는 `range()`함수를 이용할 수 있습니다 : 

```python
x = list(range(5))  # [0, 1, 2, 3, 4]
y = list(range(1, 5))  # [1, 2, 3, 4]
z = list(range(1, 10, 2))  # [1, 3, 5, 7, 9]
```

`range()`는 입력 매개변수를 총 세 개 받습니다. 

첫번째는 수열의 시작점, 두번째는 끝점(-1), 마지막은 그 수열의 간격을 표현합니다.

따라서 `range(i,j,k)`는 `i`, `i+k`, `i+2k`,...의 숫자들을 `j-1`이하가 되는 범위에서 생성합니다.

만약 위와 같이 하나의 매개변수가 입력되는 경우, `range()`는 이를 끝점(-1)으로 알아먹고, 시작점을 자동으로 `0`으로 잡습니다.

참고로 `range`는 그저 함수가 아니라, 파이썬의 또 다른 데이터 타입 중 하나인데요, 이에 대한 자세한 이야기는 나중에 다시 다뤄집니다.

또 우리는 함수를 각 리스트의 성분에 적용시킬 수 있어요. 이를 위해서는 `map()`함수가 사용됩니다.

예를 들어, 숫자 리스트의 각 성분에 절대값 함수는 다음과 같이 적용시킬 수 있습니다.

```python
x = [-1, -2, -3]
y = list(map(abs, x))  # [1, 2, 3]
```

이 `map`도 파이썬의 객체 중 하나인데요, `map(abs,x)`는 그 자체로는 `map`이란 객체를 저장합니다.

따라서 우리는 `list()`를 이용하여 타입을 변환해야지 원하는 절대값을 씌운 리스트를 얻을 수 있습니다.

참고로 `abs()`함수는 수학의 `\vert x\vert`, 즉 절대값을 계산하는 함수입니다.

````python
x = [1, -2, -3]
y = map(abs, x)
y
````

````python
list(y)
````

## 다시 함수

### 선택적 매개변수

위에서 우리는 함수에 매개 변수를 지정해줄 수 있음을 다뤘는데요, 경우에 따라선 매개변수의 입력을 선택적으로 정의할 수 있습니다.

```python
def gop(x, y=5):
    return x * y
```

위 함수에서 `y`는 선택적 매개변수로, 만약 `gop`함수를 사용할 때 두번째 매개변수가 입력되지 않을 경우, 함수는 자동으로 기존에 정해진 `y = 5`라는 값을 사용하여 해당 함수에 적용시켜 그 결과를 반환합니다.

물론 `x`는 이렇게되면 필수적인 매개변수로, 사용 시 `x`에 해당되는 매개변수를 입력하지 않으면 오류가 반환됩니다 :

````python
def gop(x, y= 5) : 
    return x*y

print(gop(3))
print(gop(3,4))
print(gop())
````

위에서 함수를 다루며, 입력 변수의 개수가 다른 경우가 있는걸 알 수 있었는데, 그 이유가 이거에요.

### 함수 호출 시 매개변수를 직접적으로 지정하기

함수를 정의하는 경우 우리는 각 입력 매개변수에 이름을 지정합니다 :

```python
def gopsem(x, y = 5, z=2):
    return x * y * z
```

`gopsem()`함수는 위의 `gop()`함수에 선택적 매개변수 `z = 2`를 더 추가한 함수입니다.

여기서 만약 우리가 필수 매개변수인 `x`의 값은 입력하고, `y`는 그대로 두고, `z`만 값을 지정해주고싶을 수 있잖아요?

이는 다음과 같이 가능합니다 :
```python
gopsem(9)  # 90
gopsem(9, z=4)  # 180
```

````python
def gopsem(x, y = 5, z=2):
    return x * y * z

print(gopsem(9))
print(gopsem(9, z = 4))
````

### 익명 함수(Anonym function)

익명 함수는 우리가 기존의 `def`라는 키워드로 정의했던 함수 대신, 일회용으로 쓸 수 있는 함수를 의미합니다.

특히나 익명 함수는 수학의 함수를 잠깐 계산 과정을 위해서 사용하기 편리합니다 :

~~~python
lambda x1, x2, ... : x1,...를 사용한 표현
~~~

`x1`, `x2`,..는 우리가 익명 함수에서 사용할 입력 변수를 의미하구요, `:`뒤에 오는 `x1,...를 사용한 표현`은 이 입력 변수로 만들 수 있는 표현을 의미합니다. 이 표현의 값이 곧 해당 익명 함수의 반환값이 됩니다. `return`이 따로 필요 없어요. 이렇게 익명 함수에는 *표현*만이 등장하고, 다른 명령어는 올 수 없습니다.

예를 들어서 $f(x,y) = x + y$의 함수는 다음과 같이 익명 함수로 표현할 수 있습니다.

```python
f = lambda x, y: x + y
```

이렇게 정의한다면 다음과 같이 바로 함수를 사용할 수 있습니다.

````python
f = lambda x,y : x+y
f(1,2)
````

근데 이렇게 `f`를 직접 지정해주기보다 익명 함수는 진짜 그 자리에서 사용하기 위한 용도가 많습니다 :

````python
(lambda x,y: x+y)(2,3)
````

````python
x = [1, 2, 3]
x = list(map(lambda x:x**2,x))
x
````

### 매개변수의 참조

함수 사용시에 입력되는 매개변수는 변수에 값을 할당하는 것과 비슷하게 동작합니다.

만약 `x`라는 매개변수가 `f`로 전달되는 경우, `f(x)`를 사용 시, 함수는 기존의 `x`가 가리키고 있던 것과 동일한 객체를 참조하여, 이를 불러와서 함수 `f`를 사용하게 됩니다. 이를 *call-by-reference*라고 합니다.

이 때문에 리스트와 같은 가변 객체(나중에 다뤄짐)는 함수에 의해서 의도치 않게 값이 변경될 수 있습니다. 다음 코드를 보시죠.

````python
a = [3, 1, 2]

def sort_print(x):
    print(x)
    x.sort()
    print(x)

sort_print(a)
a
````

이 함수는 우선 리스트의 값들을 한 번 쫙 출력하고, `sort()`메소드를 적용시킨 후, 다시 리스트를 출력합니다.

이를 `a = [3, 1, 2]`에 적용하면, `sorted_print(a)`가 사용되는 시점에서 함수는 `a`가 가리키던 객체 `[3, 1, 2]`를 불러옵니다.

따라서 만약 함수 내부에서 이 `[3, 1, 2]`를 직접적으로 변경하는 작업을 한다면, 그건 그 함수가 사용된 이후에도 쭉 이어집니다.

함수는 `[3, 1, 2]`라는 객체를 변경했고, `a`는 그대로 이 함수로 인해서 변경된 객체를 가리키니까요.

이는 다른거보다 입력 매개변수가 가변 객체이기 떄문입니다. 가변이 아닌 `int`, `str` 등의 객체들은 함수에 값이 전달되는 경우, 그 값을 직접적으로 참조하는 대신, 그 값과 똑같은 새로운 객체를 생성해서, 그 놈에 함수를 적용합니다.

## 자동 완성

IDE는 보통 자동 완성 기능이 있습니다. 주피터 노트북의 경우 `Tab`으로 작동됩니다.

예를 들어, `pr`를 입력하시고 Tab`을 누르면, 주피터 노트북은 자동으로 `pr`로 시작하는 명령어 및 변수들을 보여주는데요, 이를 방향키를 이용하여 선택 후 엔터하시면 그 놈이 자동완성됩니다. 밑에 코드셀에서 해보시죠.

````python
pr
````

이 자동 완성 기능은 특히나 객체를 위한 메소드, 혹은 객체 고유의 속성(객체가 가지고 있는 변수)를 불러올 때 편합니다.

객체들의 메소드나 속성은 일일이 다 외울 수 없는 노릇이라서요.

예를 들어서 스트링의 메소드 및 속성은 다음 코드 셀에서 `a.`이후에 `Tab`을 누르시면 확인하실 수 있습니다.

````python
a = "Hello"
````

````python
a.
````
