# 문자열 (Strings)

이미 우리는 문자열을 다뤘었는데요, 그 때는 다른거보다 그저 포맷 지정에 관해서 그리 자세하게는 다루지 않고 넘어갔습니다.

이번에는 문자열의 포맷 지정에 관한, 좀 더 자세한 스트링 이야기가 다뤄집니다.

## 문자열의 포맷 지정

문자열은 결국 문자를 출력하는 역할을 하는 놈이다 보니까, 이 문자들이 어떤 형태로 보여야 되는지의 포맷을 지정할 필요가 있습니다.

포맷 지정은 여러 종류가 있는데요, 대표적인 포맷 지정은 스트링의 특정 부분에 변수의 값을 넣어주는겁니다.

이를 위해서는 `%-`포맷팅이 사용됩니다 : 

````python
name = "장지환"
name2 = "김찬호"
print("안녕하세요 저는 %s입니다. 그 쪽은 %s씨 맞나요?" % (name, name2)) 
````

그 방법은 기존의 C언어나 자바와 매우 비슷합니다. 스트링 변수의 값을 넣는다면 `%s`를 저렇게 적어주고, 정수를 적고 싶다면 `%d`를 넣어줍니다.

마찬가지로 부동소수는 `%f`를 사용할 수 있고, 부동 소수의 소수점 및 문자 길이는 `%4.5f`등과 같이 사용 가능합니다.

그리고 위 예시에는 파이썬에서 사용하는 포맷팅의 특별한 신택스를 보여줍니다.  

저런식으로 변수들의 값이 들어갈 위치는 C나 자바처럼 스트링의 내부에 지정한 후, 그 변수의 값이 어떤 변수의 값인지는

뒤에 `%()`를 이용하여, 차례대로 넣고 싶은 친구들을 넣어줍니다.

저 방법이 맘에 들지 않는다면, 우리는 `format()` 메소드를 사용할 수 있습니다.

````python
name = "장지환"
name2 = "김찬호"
x = "안녕하세요 저는 {}입니다. 그 쪽은 {}씨 맞나요?".format(name, name2)
x
````

`format()`메소드의 경우, 스트링에 등장하는 `{}`가 있는 위치에 차례대로 입력 변수들을 넣어 붙여줍니다. 

그러면 `format()`은 입력된 변수의 값들을 순서에 따라서 `{}`의 위치에 붙여 넣습니다.

만약 `format()`의 입력 변수들이 붙여지는 위치를 순서에 다르게 조절해주고 싶다면, 다음의 방법을 사용하시면 됩니다.

````python
name = "장지환"
name2 = "김찬호"
x = "안녕하세요 저는 {a}입니다. 그 쪽은 {b}씨 맞나요?".format(a = name, b= name2)
x
````

`a = name`은 결국 `a`라는 놈이 들어있는 `{a}`의 위치에 `name`의 값을 넣으라는 의미입니다.

더 많은 놈을 포맷 지정 하려면 더 많은 변수를 마찬가지로 `x = name`의 형태로 `format()`의 매개변수로 넣어주시면 됩니다.

`format()`함수를 사용하는 경우, 우리는 `:`-신택스라는걸 사용할 수 있습니다.

````python

name = "박종우"
born_in = 1993

print("안녕하세요 저는 {a}입니다. 저는 {b}년생 입니다.".format(a = name, b= born_in))
print("안녕하세요 저는 {a:s}입니다. 저는 {b:.2f}년생 입니다.".format(a = name, b= born_in)) # :s는 스트링 형식, :f를 붙이면 부동소수 형식, :d는 정수 형식으로 출력
print("안녕하세요 저는 {a:>5}입니다. 저는 {b:<10}년생 입니다.".format(a = name, b= born_in)) # >5는 해당 변수를 위해 5칸을 확보, 오른쪽으로 붙임. < 5는 반대 방향으로
````

간단합니다. 보시다시피 결국 기존에는 `%s`, `%d`, `%f` 등으로 사용할 수 있던걸 `:`를 이용하여 사용한다는 것 뿐이에요.

`:.2f`는 `f`, 즉 부동 소수형으로 소수점 2자리수까지 표현하라는거고, `b<10`은 `b`의 값이 올 자리로 총 10자리를 확보 후, `b`의 값은 왼쪽으로 정렬시키라는겁니다.

### 포맷 스트링(f-String)


사실 위의 놈들은 그냥 이렇게 하는 방법이 있습니다 정도구요, 포매팅을 위한 가장 최신의, 가장 좋은 도구는 `f-String`입니다.

````python
name = "장지환"
name2 = "김찬호"
x = f"안녕하세요 저는 {name}입니다. 그 쪽은 {name2}씨 맞나요?"
x
````

`f-String`은 위와 같이, `f""`의 형태로, 기존의 스트링 앞에 `f`가 붙은 놈입니다.

파이썬은 이러한 놈을 포맷 지정을 위한 특수 명령어가 포함된 스트링으로 인식하여, 위와 같이 `{ }`안에 오는 친구를 스트링에 붙여줍니다.

`f-String`은 이러한 포맷의 형식을 매우 직관적으로 보여준다는대에 장점이 있습니다.



심지어 이 괄호 `{ }` 안에는  변수 뿐 아니라, 표현들도 올 수 있습니다.


````python
x = f"10과 20을 더하면 {10 + 20}입니다." 
x
````

그러면 출력 시에는 그 표현의 결과를 스트링에 붙여줍니다.

마찬가지로 저런 단순한 표현 뿐 아니라, 함수들도 표현의 일종이라 함수들을 넣어줄 수도 있습니다.

이런 면에서 포맷 스트링은 기존의 `%()`나 `format()`함수를 이용하는 것보다 더 좋은 방법입니다. 가독성도 더 높구요.

위에서 사용했던 `:-`신택스도 `f-string`에서 사용 가능합니다.

## 이스케이프 시퀀스(Escape sequence)

문자열 내부에는 특수한 포맷을 지정하기 위한 문자들이 올 수도 있습니다. 이들은 예를 들어 다음과 같습니다 :

- `\n` - 행 바꿈(새로운 행으로 넘기기)
- `\t` - 탭(들여쓰기)
- `\\` - `\`가 이런 명령어의 접두어로 사용되므로, 문자열 내부에서 `\`를 표현하기 위한 문자. 
- `\"` - `"`가 문자열을 표현하는대에 사용되므로, 문자로서 `"`를 스트링에 넣고 싶다면 `\`를 붙여서 사용해야함.
- `\'` - `'`도 문자열을 표현하는대에 `"` 대신에 사용 가능하므로, 이렇게 `\'`를 붙여서 사용함.

이렇게 `\` 이후에 뭔가가 와서 표현되는 포맷 서식 지정자를 이스케이프 시퀀스(Escape sequence)라 합니다.

예를 들어 다음의 스트링은

````python
x = "안녕하세요\n\"반갑습니다\" \t! \'\\\'"
print(x)
x
````

출력 시 위와 같은 형태가 됩니다. 물론 출력 하기 전 단계에 `x`에는 `"안녕하세요\n\"반갑습니다\" \t! \'\\\'"`가 그대로 들어있습니다.

위 코드에서도 보시다시피, 이스케이프 시퀀스는 출력 시에만 반영됩니다.

이스케이프 시퀀스들은 모두 길이가 1로 인식됩니다. 즉, 만약 `str = "asd\n"`이라는 스트링이 주어진다면, 이놈의 길이는 4이고, 인덱싱을 하더라도 `str[3] == "\n"`이 됩니다.

스트링들 중에서는 이러한 이스케이프 시퀀스들을 포함하지 않는 스트링도 존재하는데요,

이스케이프를 포함하지 않는 스트링을 `Raw-String`이라 하며, 다음과 같이 사용 가능합니다.

````python
x = r"안녕하세요\n\"반갑습니다\" \t! \'\\\'"
print(x)
````

그냥 스트링의 앞에 `r`을 붙이면 됩니다. `f-String`과 비슷하죠.

보시다시피 `Raw-String`은 이스케이프 시퀀스들을 모두 그냥 보통의 문자로 인식합니다.

그리고 C나 자바와 달리 파이썬은 `"""`로도 스트링을 정의 가능합니다.

```python
x = """안녕하세요
반갑습니다."""
```

이 경우, 해당 스트링은 우리가 눈으로 볼 수 있는 행 바꿈과 들여쓰기를 실제 출력시의 서식으로 인식합니다.

````python
x = """안녕하세요
반갑습니다.
    들여쓰기도 인식됩니다."""
print(x)
````

이러한 이스케이프 시퀀스를 적절히 활용하면 다음과 같은 가독성 좋은 정보들도 출력할 수 있습니다.

````python
print("이름\t점수\t학점\n장지환\t95\tA+\n손인욱\t87\tB+")
````

## 독스트링(Docstrings)

독스트링은 함수 정의 시 가장 앞에 오는 스트링을 의미합니다.

예전에 다룬대로 파이썬은 `help` 혹은 `?`를 이용하여 기본적으로 제공되는 함수들이나, 사용자가 해당 파이썬 코드에서 정의한 함수들의 정보를 불러올 수 있는데요, 독스트링에 적은 글귀들은 함수들의 정보로서 불러와집니다.

독스트링으로는 그냥 함수 정의 시 바로 다음 문장의 스트링이 지정됩니다.

이 때, 독스트링은 `"""`를 이용한 삼중 스트링을 사용하는 컨벤션이 있습니다. 예를 들어 다음과 같이요. 

```python
def zegob(x):
    """
    이 함수는 입력 매개변수 x의 제곱을 반환합니다.
    """
    return x ** 2
```

````python
def zegob(x):
    """
    이 함수는 입력 매개변수 x의 제곱을 반환합니다.
    """
    return x ** 2

help(zegob)
````

독스트링은 `"""`가 사용됨에 따라서, 여러 행에 걸친 함수 설명이 가능합니다.

독스트링에는 기본적으로 함수의 입력 변수와 반환값을 설명해야한다는 컨벤션이 있습니다.

```python
def zegob(x):
    """
    이 함수는 입력 매개변수 x의 제곱을 반환합니다.

    Arguments :
    x -- 제곱을 계산 할 입력 변수

    Returns :
    x의 제곱
    """
    return x ** 2
```

````python
def zegob(x):
    """
    이 함수는 입력 매개변수 x의 제곱을 반환합니다.

    Arguments :
    x -- 제곱을 계산 할 입력 변수

    Returns :
    x의 제곱
    """
    return x ** 2

help(zegob)
````

`help()`함수 대신에 `?`를 사용하셔도 됩니다.

````python
?zegob
````

# 함수의 변수

## 변수의 가시성

함수의 안에서 정의되는 변수들은 오로지 함수의 내부에서만 볼 수 있으며, 사용 가능합니다.

이런 변수들을 **지역 변수**(Local variables)라 합니다.

````python
def localvar():
    x = 10
    print(x)

x = 20
localvar()
print(x)
````

위 함수는 지역 변수를 설명하기 위한 함수인데요, `localvar()`의 내부에 `x`가 있고, 이를 반환하지만, 이 `x`는 오로지 저 함수 내부에서만 유효합니다.

따라서 `localvar()`의 `x`는 함수 밖에서는 따로 효력이 없어서, 위와 같이 함수의 밖에서 `x = 20`으로 값을 부여한 후 `localvar()`을 사용하더라도, 기존의 함수 밖에 있던 `x`의 값은 바뀌지 않습니다. 그래서 저렇게, 함수 내부의 `x`의 값인 10, 그리고 함수 밖의 `x`의 값인 20이 출력됩니다.

근데 항상 이런건 아니구요, 만약 우리가 함수 내부에서 일부러 전역적인 변수(Global variable)를 지정한다면, 이 놈은 함수 밖에도 영향을 미칩니다.

````python
def globalvar():
    global x
    x = 10
    print(x)
x = 20
print(x)
globalvar()
print(x)
````

`localvar()`과 다르게, `globalvar()`에서의 `x`는 함수의 밖에서도 힘을 잃지 않습니다. `globalvar()`의 내부의 `x`가 `global x`로 정의되어서 그래요.

`global x`라는 선언이 있었음에 따라서, 이 함수 안의 `x`는 함수 밖의 `x`와 같습니다.

참고로 `global x = 0`와 같이, 전역 변수 선언과 동시에 값의 변경은 파이썬에서 불가능합니다. 위처럼 따로 따로 해주셔야해요

사실 보통 이러한 전역 변수는 위와 같이 함수 내부에서 정의를 시작하는거보다는, 기존에 정의되어 있던 함수 밖의 변수를 함수 내부에서도 동일하게 작용시키기 위해서,

파이썬 인터프리터에게 이 변수가 함수 밖에 있던 그 함수랑 같은 놈임을 말해주기 위해 사용됩니다. 만약 기존의 함수 밖에 그 변수가 없다면, 그걸 생성하는 역할도 할 수 있구요.

## 스코프(Scope)

이렇게, 변수들은 범위 별로 유효성/가시성이 달라질 수 있는데요, 변수들이 유효한 범위를 스코프라고 합니다.

즉, 스코프는 프로그램의 특정 변수들이 어떤 범위에서 보이는지를 생각하기 위한 개념입니다.

위의 지역 변수 예시와 같은 예시를 한 번 더 보시죠.

````python
y = 0
def f(x):
    y = 2*x
    return y

print(f(1))
y
````

`f`는 수학의 함수 $y = 2x$를 표현하는 함수입니다.

여기서 기존에 함수의 바깥 스코프에서 `y`라는 변수가 있었는데, 갑자기 함수의 정의 안에 다시 `y`가 등장하죠.

만약 이 함수 내부의 `y`가 함수 바깥의 `y`를 덮어씌운다면 우리가 의도한대로 프로그램이 작동하지 않을 수 있습니다.

이게 뭐가 문제냐, 그냥 `y`말고 다른걸 사용하면 되지 않나 싶으실 수 있죠. 맞는 말입니다.

하지만 이렇게 사용자가 정의한 함수가 아닌, 파이썬에서 기본적으로 미리 정의되어 있는 표준 함수들이라면 어떨까요?

표준 함수들의 입력 변수는 우리가 다 알지 못합니다. 이걸 다 외울수도 없는 노릇이구요.

더 나아가 사용자가 정의한 함수더라도 경우에 따라서, 예를 들어 팀 단위로 프로그래밍을 하는 경우, 누군가 입력 변수를 기존의 변수와 같은 형태로 정의할 수도 있잖아요. 그 경우에는 만약 위와 같이, 함수 내부의 변수와 함수 밖의 변수를 구분하는 규칙이 없다면 큰 문제가 발생할 수 있습니다.

따라서 파이썬에서는 각 함수에게 함수만의 고유한 스코프를 부여하여, 그 스코프 내부의 변수는 전역 스코프에 영향을 주지 않도록 설정되어 있습니다.

이러한 스코프에 관한 규칙은 다음과 같이 정리해볼 수 있습니다.
- 함수 밖의 변수는 기본적으로 `global`이다.
- 함수 내부의 변수는 기본적으로  `local`이다.
- 함수 내부에 있더라도 `global`로 정의된 변수는 전역적이다.
- 함수 내부에 있는 `nonlocal`로 정의된 변수는 그 함수의 바로 바깥쪽 스코프까지 유효하다.(`nonlocal`은 함수 안에 함수를 정의하는 경우에만 사용 가능)
- 파이썬은 변수에 접근 시, 우선 지역 스코프의 변수에 먼저 접근하며, 만약 지역 스코프에 변수가 없다면 순서대로 바깥쪽 스코프로 가며 접근한다.

특히 마지막 규칙은 `LEGB(Local - Enclosing - Global - Built-in)`규칙이라고도 합니다. 이는 말 그대로, 프로그래머가 변수 `x`를 사용하는 경우 우선 지역 스코프에서 `x`가 있는지 확인하고, 거기 없다면 논로컬 스코프(그 지역이 중첩된 함수라면)에서 확인하고, 논로컬에서도 `x`가 없다면 전역 스코프, 심지어 전역에도 없다면 기본적으론 내장된 파이썬의 빌트인 스코프에서 `x`를 찾는 그런 규칙을 의미합니다. 만약 빌트인 스코프에도 없다면 없는 변수라는 에러가 발생됩니다.

파이썬에서는 C의 헤더 파일이나 자바의 패키지와 비슷한 *모둘*이라는 놈이 있습니다.(아래에서 더 다뤄집니다.) 우리는 프로그래밍의 용도에 따라서 모둘을 `import`라는 명령어를 이용하여 불러온 후, 그 모둘에 있는 기능들을 활용할 수 있는데요, 이 모둘은 자기 자신만의 고유한 전역 스코프를 가지고 있습니다.

즉, 모둘의 전역은 자기 자신의 전역을 의미하므로, 모둘은 해당 모둘을 로드한 프로그램 혹은 다른 모둘에는 직접적인 접근이 불가능합니다.

`import`라는 명령어는 사실 우선 `import`의 대상이 되는 모둘을 한번 *수행*하여 그 모둘 자기 자신의 스코프에 해당 모둘이 가지고 있는 변수 및 함수들을 채워놓고, 그리고 이들을 말 그대로 우리의 프로그램에 빌려오는 식으로 작용한다고 생각하시면 됩니다. 따라서 우리의 프로그램과 모둘의 프로그램의 스코프는 별개입니다.

쉽게 말해서, `module`에 `x`라는 변수가 있더라도, 그 `x`는 `module`에서만 전역 변수이고, 우리의 프로그램에서는 `x`가 따로 없다면 전역 변수가 아닙니다.

이제 이러한 스코프의 예시를 몇 개 보겠습니다.

### 예시

다음과 같은 경우,

````python
x = 4
def f():
    print(x)
f()
````

위에서 설명한 `LEGB` 규칙대로, 함수의 지역 스코프 내에 `x`라는 변수가 없으니, 이 `x`는 전역 스코프에서부터 불러와집니다.

따라서 전역 스코프의 `x`의 값이 출력됩니다. 이 경우에는 굳이 함수 내부에서 `global`로 `x`를 전역 변수로 지정하지 않아도 됩니다.

하지만 다음의 경우는 좀 달라요.

````python
x = 4
def f() :
    x = 5
    print(x)
f()
print(x)
````

이 경우에서는 보시다시피, 함수 `f`의 내부에 `x`라는 지역 변수가 있으므로, 함수를 사용할 시에는 이 지역 변수가 함수의 내부에서 사용됩니다.

따라서 `f()`는 바깥의 `x`와 상관없이 항상 같은 5라는 값을 출력하며, 함수 바깥의 전역 `x`는 기존의 값을 그대로 가지고 있습니다.

이래서 `global`을 사용하는거에요. 만약 `f`로 기존의 전역 변수의 값을 변경하고 싶다면 다음의 코드를 사용해야합니다.

````python
x = 4
def f() :
    global x
    x = 5
    print(x)
f()
print(x)
````

참고로 함수 내부의 지역 변수는 나중 순서에 지역 변수로 정의되더라도 지역 변수입니다. 순서에 상관없어요.

````python
x = 0
def f():
    print(x)
    x = 1  # 함수 내부에 이렇게 x를 지역 변수로 선언했으므로, 이 함수 내부에서 x는 지역 변수의 역할을 가짐.
f()
````

그 이유는 파이썬이 우선 함수를 소환하기 전에 먼저 함수를 전체적으로 확인하기 떄문입니다.

그럼에도 함수가 소환되는 경우에는 명령이 순차적으로 수행되므로, `print(x)`의 시점에서 아직 지역 변수 `x`에 값이 할당되지 않아서, 위와 같이 오류가 발생합니다.

기본적으로 지역 변수는 값이 할당되어야지 사용이 가능하다고 생각하시면 됩니다.

이런 스코프들은 실제 프로그래밍의 과정에서 혼동을 줄 수 있기 때문에, 일단 최대한 전역 변수를 건드리는 함수들은 정의하지 않는 게 맞습니다.

물론 전역 변수를 건드리는 게 유용하거나 타당한 근거가 있는 경우엔 상관없구요.

### 가변과 비가변 객체(Mutable Objects)

객체들은 가변성으로 카테고리를 구분할 수 있는데요, 가변성은 말 그대로 값의 변경이 가능한지를 의미합니다.

예를 들어, 숫자 `int`의 경우 값의 수정이 불가능합니다.

```python
a = 10
a = 20
```

어라? 위의 코드처럼하면 수정이 가능한거 아닌가? 라고 생각하실 수 있는데요, 그렇지가 않아요.

위 코드의 첫 줄에서 `a`에는 `10`이 저장되고, 두번째 줄에서 `a`에 `20`을 저장됩니다.

포인트는 우리는 위 코드에서 `a`에 저장되는 대상을 변경할 뿐이지, 숫자 `10` 그 자체를 변경하는 게 아니라는겁니다.

변수의 경우, 결국 특정 객체(값)를 가리키는 역할을 하는데요, 위에서는 `a`가 가리키던 대상인 `10`이 `20`으로 변하는 것이 아니고,

기존에 `10`을 가리키던 `a`가 이제 새롭게 생성된 `20`을 가리키는 것 뿐이에요.

즉, 위의 두 코드는 `a`에 값을 재할당하는거지, 객체 값 그 자체를 수정하는 게 아니라는거죠.

이러한 점은 숫자 대신에 스트링으로 생각하면 더 편합니다.

```python
s = "123123"
s[0] = "a" # 에러 발생
```

스트링은 위와 같이 첫번째 글자를 직접적으로 변경해줄 수 없었습니다. 대신에 `"a" + s[1:]`등의 슬라이싱을 이용해야했죠.

이런식으로 객체의 값 그 자체를 수정할 수 없는 친구들을 *비가변 객체(Immutable Objects)* 라 합니다.


반대로 내용물의 변경이 가능한 객체들도 있었죠. 바로 리스트, 딕셔너리 등이 이에 해당되는데요, 얘네를 *가변 객체(Mutable Objects)*라 합니다.

리스트의 경우

```python
x = [1, 2, 3]
x[0] = 12
```

이런식으로, 객체 `[1, 2, 3]`을 `[12, 2, 3]`로 수정이 가능하다 이런거죠. 새로운 리스트를 생성하는 게 아니라, 기존의 리스트에서 값만 바꾸는거니까요.

이 경우에는 객체의 내용물이 변경되더라도 변수 `x`는 그대로 같은 객체를 가리키고 있어서, 위의 비가변의 경우와는 완전 다릅니다.


### 함수의 중첩과 클로저

위에서 논로컬 스코프를 설명하며 함수는 다시 함수 안에 함수를 정의하며 중첩이 가능할 수 있음을 다뤘습니다.

대충 논로컬은 다음을 코드로 이해할 수 있습니다.

```python
x = 1
def f() :
    x = 10
    def g() :
        x = 100
        print(f"g() x = {x}")
    g()
    print(f"f() x = {x}")
```
우선 전역의 `x`는 1을 담고 있는 변수인데요, 함수 `f`의 내부에는 다시 `x = 10`을 가지고 있는 지역 변수가 있습니다.

또 다시 `f`의 내부의 함수 `g`에도 `x`가 등장하는데요, 이 `x`는 `g`를 기준으로 지역 스코프라 할 수 있습니다.

그리고 `g`의 논로컬 스코프는 이 `g`의 바깥에 있는 `f`의 스코프를 의미합니다.

따라서 위 코드에 등장하는 `x`들은 지역 변수(가장 안쪽)의 경우 `100`이라는 값을 가리키고 있고, 논로컬 변수(`f`의 스코프)에서는 `10`을 가리키며, 전역 스코프(가장 바깥쪽)에서는 `1`을 가리킵니다. 이는 다음과 같이 코드 셀을 실행하여 확인이 가능합니다.

````python
x = 1
def f() :
    x = 10
    def g() :
        x = 100
        print(f"g() x = {x}")
    g()
    print(f"f() x = {x}")
f()
print(f"global x = {x}")
````

여기서는 중첩이 딱 한 번 되기 때문에 헷갈릴 게 없는데요, 만약 한 번 이상 다중 중첩이 주어지는 경우에, 논로컬은 항상 그 위치를 기준으로 바로 바깥의 스코프까지입니다.

이 쯤되면 함수를 대체 복잡하게시리 왜 중첩시키는지 의문이 생기실 수 있는데요, 당연히 장점이 있기 때문입니다.

1.  불필요한 코드를 줄여서 가독성을 높일 수 있음.
2.  재귀적인 설계가 가능함.
3.  함수 내부에서 또 다른 함수를 정의한 후, 그 함수를 반환할 수 있음. 파이썬에서는 함수도 객체임!
4.  3번의 맥락에서 그 함수를 다시 또 다른 함수의 입력 변수로 돌려줄 수도 있음

1번은 그렇다치더라도 나머지는 아직 잘 이해되지 않으실 수 있는데요, 이는 예시를 보면서 생각해보기로 하겠습니다.

2번의 재귀를 보여주는 예시는 다음과 같습니다.

````python
def f(x,n, m):

    def g(a,b):      
        return a**b
    
        
    for i in range(m):
        x = g(x,n)
    return x
x = 2
n = 2
m = 4
print(f(x, n, m))
print(x**(n**m))
````

뭔가 쓸데없이 복잡해보이는데요, 간단합니다. 위 함수는 `x,n,m`를 받아서 $x^{n^m}$을 계산합니다.

그리고 `g(a,b)`는 $a^{b}$를 계산하는데요, 우리는 이 `g`를 이용하여, 다음의 과정으로 $x^{n^m}$을 계산하는겁니다.
$$x^{n^m} = \underset{m\text{ 번}}{\underline{x^{n\cdot...\cdot n}}} = \Big(\big((x^n)^n\big)^n...\Big)^n = g\Big(...g\big((g(x,n)\big)\Big)$$

이런 재귀적인 계산은 많은 복잡한 계산을 간단하게 할 수 있게 해줍니다. 예를 들어 피보나치 수열도 일반항을 구하기는 빡세지만,

재귀항으로 표현하기는 간단하잖아요? 이런 비슷한 경우에 중첩된 함수를 사용하면 유용합니다.(계산비용은 생각하지 않는다면 말이죠.)

그리고 3번, 4번은 다음과 같은 용도를 생각하시면 됩니다.

````python
def f(x,y) :
    def f1(x,y) : 
        return x**2+y
    def f2(x,y) :
        return x*y
    return [f1,f2]
g = f(x,y)
print(f"Value of f(x,y) = (x^2+y,y) in (3,4) is : ({g[0](3,4)},{g[1](3,4)})")
````

수학적인 맥락에서 우리는 위와 같은 $f(x,y) := (x^2+y,xy)$를 정의한다면, 위의 `f`를 통해서 각 성분 함수 $f_1(x,y) = x^2+y$, $f_2(x,y) = xy$를

반환하는 함수 `f`를 정의할 수 있습니다. 그러면 위와 같이 `g = f(x,y)`를 통해서 `g`에 각 성분 함수를 담고 있는 리스트 `[f1,f2]`를 정의할 수 있습니다.


또 중첩된 함수를 이용하여 우리는 **클로저**(Closure)라는 개념도 구현할 수 있습니다.

````python
def make_appender():
    lst = []
    def append(x):
        lst.append(x)
        return lst
    return append
````

이 함수 `make_appender()`는 빈 리스트를 만들어서, 그 리스트에 값을 추가하는 함수 `append()`를 반환하는 함수입니다.

만약 이 함수를 이용하여 다음과 같이 선언하면

````python
a = make_appender()
b = make_appender()
````

a와 b에는 결국 `make_appender()`로 인해 만들어진 `append()`라는 함수가 남아있는데요,

얼핏 보면 a와 b에 모두 같은 함수가 들어있는거 아닌가 하실 수 있으나, 안에 들어있는 친구들은 모두 다릅니다!

````python
a(1)
````

이렇게 a(1)을 한다면, 결국 이는 `append(1)`을 하는거니, `a(1)`d으로 인해 빈 리스트에 1이라는 값을 추가한 `[1]` 형태의 리스트가 반환됩니다.

만약 a,b에 들어있는 함수들이 모두 같다면, 이제 `b(2)`를 하면 반환되는 리스트가 `[1, 2]`여야겠죠?

````python
b(2)
````

하지만 그렇지 않습니다. `b(2)`는 마치 이전에 `a(1)`로 반환된 리스트 `[1]`은 뭔지 모른다는듯이, `[2]`를 반환합니다.

그 이유는 진짜 `b`는 `a(1)`을 모르기 떄문이에요.

a,b는 모두 `make_appender()`함수가 만들어내는 함수 `append()`를 가지고 있는데요, 이 함수 `append()`는 해당 함수의 논로컬인 `lst`라는 변수를 알고 있어요.

그리고 각각의 a,b에 `make_appender()`가 저장되는 시점에, a와 b는 이 `lst`를 각각 다른 변수로 알고 있는겁니다.

즉, `a(1)`은 `a`에 `make_appender()`가 사용되던 시점의 그 리스트 `lst`에다가 1을 추가하며,

`b(2)`는 `b`에 `make_appender()`가 사용되던 시점의 리스트 `lst`에다가 2를 추가합니다. 서로 다른 리스트를 추가할 대상으로 알고 있는거죠.

놀라운 점은, 이미 우리가 `make_appender()`라는 함수를 사용하고 끝냈는데도 불구하고 `a`와 `b`에는 해당 함수를 사용하면서 생성되었던

논로컬인 리스트 `lst`를 계속 기억하고 있다는겁니다. 이런식으로 `make_appender()`의 지역 변수로 사용되었던 놈을 함수 객체 `append()`안에 포함시키는걸 클로져라고 합니다.

# 클래스(Class)

클래서는 데이터와 기능들을 하나로 묶어서 새로운 객체를 만들기 위한 도구입니다.

정확히 말하면, 객체들이 어떤 형태로 설계되어 있는지를 작성하는 설계도로 이해할 수 있습니다.

즉, 지금까지 다룬 객체들의 경우, 정확히 말하면 **인스턴스(Instance)** 라는 놈입니다.

`10`이라는 숫자는 `int`의 데이터 타입을 가지는 객체라고 이해했었는데요, 클래스의 맥락에서 `10`은 `int`라는 클래스의 값이 `10`인 인스턴스가 되는겁니다.

클래스에는 각 객체들이 가질 수 있는 값들, 그리고 이 클래스의 객체들에 사용할 수 있는 메소드들이 들어있습니다.


클래스를 정의하는 키워드는 `class`입니다. 예를 들어서 은행 어플을 프로그래밍 하기 위해서 고객 정보의 객체를 설계한다고 해보죠.

만약 고객 정보로 고객의 이름, 계좌, 현재 잔고만을 고려하는 경우, 우리는 다음의 클래스를 정의할 수 있습니다.

````python
class Client :
    def __init__(self) :
        self.name = ""
        self.iban = 0
        self.balance = 0.0
````

뭔가 복잡해보이는데 간단합니다. 우선 클래스의 이름은 `client`로 했습니다. 말 그대로 고객이니까요.

그러면 이제 이 `client`의 인스턴스 객체들이 어떻게 구성되어 있는지를 정해야하는데요, 이는 위와 같이 `__init__()`라는 함수로 가능합니다.

`__init__()`은 **생성자(Constructor)** 라고 불리는 함수로, 말 그대로 해당 클래스의 객체의 생성하는 역할을 합니다.

생성자의 안에는 이 객체들이 가져야 할 데이터, 우리의 입장에서는 이름(name), 계좌(iban), 잔고(balance) 데이터들이 들어옵니다. 이러한 데이터들을 속성(Attribute)이라 합니다.

`__init__()`은 직접적으로 생성하는대에 사용되지는 않고, 생성 규칙을 정하는 느낌으로 이해하시면 됩니다.

그리고 `__init__(self)`의 `self`는 말 그대로 자기 자신을 의미합니다. 객체는 클래스로부터 생성되는데, 생성된 각 인스턴스들 자기 자신을 의미하는거죠.

위의 경우, `client`의 인스턴스가 생성될 경우, 그 인스턴스 자기 자신의 속성을 우선 이름은 `""`, 계좌 번호는 `0`, 잔고는 `0.0`인 상태로 두라는 의미입니다.

`self`가 잘 이해가 안되신다면 좀 더 아래에 메소드를 다루면서 이해하시면 되겠습니다.

만약 위의 형태로 클래스를 정의한다면, 해당 클래스의 인스턴스는 다음과 같이 정의 가능합니다.

````python
a = Client()
print(f"고객명 : {a.name}\n계좌번호 : DE{a.iban}\n잔고 : {a.balance} EUR")
````

이 시점에서 `a`는 `Client` 클래스의 인스턴스입니다. 보시다시피, `a`는 고객명, 계좌번호, 잔고로 이뤄져 있습니다.

근데 아직 값들이 하나도 제대로 들어있지 않죠. 값의 설정은 다음과 같이 가능합니다.

````python
a.name = "엄준식"
a.iban = 15578884844
a.balance = 1557.88848

print(f"고객명 : {a.name}\n계좌번호 : DE{a.iban}\n잔고 : {a.balance} EUR")
````

그냥 직접적으로 `a`를 참조하여 속성들의 값들을 변경해줄 수 있습니다.

이는 자바와 매우 다른 점입니다. 자바는 보통 속성을 `private`이라는 형태로 정의하여, 외부에서는 직접적으로 속성을 변경하는걸 막고, 속성 변경은 항상 메소드로 다뤄지는데요, 파이썬은 그에 비해 기본적으로 저렇게 직접 참조를 통한 속성의 변경이 가능하게 놔둡니다. 변경이 아니더라도 값을 그대로 불러오는거도 이렇게 참조로 가능합니다.

근데 보통은 저렇게 객체를 먼저 하나 생성하고, 따로 값을 설정하는거보다, 생성과 동시에 값을 부여해주는 게 편할 수 있습니다.

이를 위해서는 클래스를 다음과 같이 작성해야합니다.

````python
class Client :
    def __init__(self, name, iban, balance) :
        self.name = name
        self.iban = iban
        self.balance = balance
````

그러면 인스턴스 생성 시, `Client()`의 괄호 안에 `name, iban, balance`를 입력하여 바로 처음부터 속성값을 부여할 수 있습니다.

````python
a = Client("김찬호", 15578884844, 1557.88848)
print(f"고객명 : {a.name}\n계좌번호 : DE{a.iban}\n잔고 : {a.balance} EUR")
````

파이썬은 기본적으로 변수의 타입을 직접 지정하지 않아도 되기 떄문에 저렇게만 해도 괜찮은데요,

보통 우리는 `name`으로는 스트링, `iban`으로는 계좌번호이니 정수, `balance`로는 잔고이니 유리수임을 기대합니다.

이를 반영하여 우리는 속성들의 각 데이터 타입에 다음과 같이 힌트를 줄 수 있습니다.

````python
class Client :
    def __init__(self, name : str, iban : int, balance : float) :
        self.name = name
        self.iban = iban
        self.balance = balance
````

사실상 위의 클래스와 다른 점은 크게 없는데요, 저렇게 하면 IDE에서 속성값을 입력하는 경우에 뭐가 어떤 값인지 힌트가 나옵니다.

그렇다고 그 타입들을 강제하는건 아니라서요, `name`으로 예를 들어 `123`이라는 `int`값을 넣는거도 허용됩니다.

## 메소드

지금까지도 함수들을 다루면서 메소드라는 친구도 따로 구분하여 사용했었는데요, 메소드는 각 인스턴스이 직접적으로 접근하여 동작하는 함수의 일종이라고 할 수 있습니다.

따라서 기존의 함수들은 클래스와 별개로 그냥 항상 전역적으로 정의 및 사용될 수 있던 개념이었으나, 메소드는 그 메소드가 접근할 객체의 클래스에서 정의됩니다.

클래스 그 자체를 위한 클래스 메소드, 그리고 인스턴스들을 위한 인스턴스/오브젝트 메소드의 두 종류로 나눌 수 있습니다.

우리는 우선 그 중 인스턴스/오브젝트 메소드만을 다룹니다.

이 친구들은 인스턴스의 속성을 변경하거나, 속성을 이용하여 다른 작업을 수행하는 등의 역할을 하는 말 그대로 인스턴스들을 위한 함수입니다.

어쩄던간에 함수다보니, 정의하는 방법은 기존의 함수와 매우 비슷합니다 :

```python
def methode(self, a, b,...)
    #...
```

간단하죠. 똑같은데요, 차이점은 인수에는 항상 `self`가 오게 됩니다. 나머지 `a,b,...`등의 입력 매개변수들은 그 함수에 필요하다면 넣고, 아님 말고에요.

`self`는 말 그대로 이 메소드가 접근할 인스턴스 그 자체를 의미합니다. 우리가 메소드로 인스턴스의 속성을 변경한다고 한다면, 

코드 상 그 인스턴스를 불러야할거아니에요? 그걸 위한 대표적인 이름이 바로 `self`가 되겠습니다.

예를 들어 인스턴스 메소드는 `i`라는 인스턴스에 대해서 다음의 형태로 사용됩니다.

```python
i.methode(a,b,...)
```

`self`는 여기서 저 `i`를 지칭하는 대표적인 이름이라고 보시면 됩니다. 말 그대로 지금 참조되는 인스턴스 자기 자신을 의미하는거죠.

예를 들어서 위의 `Client`객체에 대해서, 고객 계좌에 잔고를 더하는 메소드는 다음과 같이 만들어볼 수 있습니다.

````python
class Client :
    def __init__(self, name : str, iban : int, balance : float) :
        self.name = name
        self.iban = iban
        self.balance = balance

    def add_balance(self, amount : float) :
        self.balance += amount

a = Client("김찬호", 15578884844, 0)
print(f"고객명 : {a.name}\n계좌번호 : DE{a.iban}\n잔고 : {a.balance} EUR \n")

a.add_balance(1557.0)
print(f"고객명 : {a.name}\n계좌번호 : DE{a.iban}\n잔고 : {a.balance} EUR")
````

별로 이해하시기 어려운건 없을겁니다. 그냥 함수랑 똑같아요. 인스턴스들에 참조하기 위한 `self`가 들어온다 그 차이죠.

근데 저는 지금 계속 `print()` 함수를 이용해서 속성값들을 출력하고 있는데요, 기존의 단순 데이터들에 했던대로 그냥 `a`를 출력하면 어떤 일이 벌어질까요?

````python
print(a)
````

그러면 보시다시피 우리가 기대한 속성값들이 아니라, 이 인스턴스 `a`의 실제 정체가 반환 보여집니다.

그래서 우리는 정보를 출력하려면 저렇게 각 속성을 직접 출력시켜야합니다.

아니면 그냥 이 출력하는거 자체를 담당하는 메소드를 정의해도 됩니다.

````python
class Client :
    def __init__(self, name : str, iban : int, balance : float) :
        self.name = name
        self.iban = iban
        self.balance = balance

    def add_balance(self, amount : float) :
        self.balance += amount

    def show(self) :
        print(f"고객명 : {self.name}\n계좌번호 : DE{self.iban}\n잔고 : {self.balance} EUR")

a = Client("김찬호", 15578884844, 0)
a.show()

a.add_balance(1557.0)
a.show()
````

이렇게 본인이 정의한 클래스의 인스턴스들에 대해서는 기존의 파이썬의 기본 표준 함수들이 작동하지 않을 가능성이 높습니다.

````python
abs(a)
````

당연한거겠죠. `abs()`는 절대값을 반환하는 함수인데, 고객들에 절대값이 어디있겠어요.

근데 이거도 덕 타이핑(Duck Typing)이라는걸 이용하면 해결이 가능합니다.

````python
class Client :
    def __init__(self, name : str, iban : int, balance : float) :
        self.name = name
        self.iban = iban
        self.balance = balance

    def add_balance(self, amount : float) :
        self.balance += self.balance + amount

    def show(self) :
        print(f"고객명 : {self.name}\n계좌번호 : DE{self.iban}\n잔고 : {self.balance} EUR")
    
    def __abs__(self) :
            return self.balance


a = Client("김찬호", 15578884844, 15.30)

    
abs(a)
````

덕 타이핑은 타입을 맞추는 방법을 의미하는데요, 오리(Duck)가 붙는 이유는 다음의 철학을 따르기 떄문입니다.
$$\textit{오리처럼 걷고, 오리처럼 꽥꽥대는 놈은 오리이다!}$$
`abs()`함수는 당연히 `Client` 객체가 들어오는걸 절대 기대하지 않습니다.

그럼에도 위와 같이, `__abs__()`메소드를 정의해줌을 통해서 `Client`에 대해서 `abs()`가 사용될 시, 어떻게 작동할지를 정해줄 수 있습니다.

즉, `Client`는 오리가 아니지만, `__abs()__`메소드를 정의하여 오리처럼 행동하는 방법을 알려준다면, `abs()`의 입장에서는 `CLient`도 오리라고 판단하고,

위와 같이 `Client`라는 오리는 아니지만 오리처럼 행동하는 놈에도 `abS()`를 쓸 수 있게 되는거죠.

저는 위에서 잔고를 그냥 그대로 반환하도록 설정했어요.

여튼 `abs()` 뿐 아니라 `len()`, `print()` 등의 많은 함수들이 덕 타이핑을 지원합니다.

## 클래스 이름의 규칙과 클래스의 독스트링

파이썬에서 클래스의 이름들은 캐멀 케이스(Camel Case)라는 컨벤션으로 이름을 짓습니다.

일단 위의 `Client`처럼 첫 글자는 대문자로, 그리고 여러 단어가 등장하는 경우, 각 단어의 첫번째 글자를 대문자로 적는 형태를 의미합니다.

언더스코어를 쓰지는 않습니다. 다음과 같이요.

```python
class ClientList : ...
```

또한 클래스의 이름에는 항상 그 클래스가 뭐하는 놈인지를 보여줄 수 있는 명사를 넣어주셔야합니다.

클래스에 대해서도 위에서의 독스트링을 똑같은 방법으로 넣어줄 수 있습니다. 함수의 경우와 마찬가지로 클래스 정의 후 바로 다음 스트링이 독스트링이 됩니다.

그리고 각 메소드들에 대해서도 독스트링을 넣어줄 수 있습니다.

# 모둘(Module)

모둘은 함수, 클래스, 변수들을 모은 집합체로, 파이썬이 기본적으로 제공하는 기능을 확장하기 위한 도구입니다. DLC죠.

## 모둘 로드하기

모둘은 다음과 같은 방법으로 여러분들이 만들고 있는 프로그램으로 가져올 수 있습니다.

~~~python
import module [as name]
~~~

`module`에는 가져오려는 모둘의 풀네임을 적어줘야 하고, `[as name]`의 `name`에는 이 모둘의 기능들을 사용할 때 사용할 또 다른 식별자를 정해둘 수 있습니다.

모둘의 함수들을 사용할 때에는 항상 그 모둘을 참조하는데, 이러한 경우의 편의성을 위한 기능이에요. 

예를 들어 `modul1`이라는 모둘의 함수 `f()`를 사용하는 경우, 우리는 다음의 명령어를 사용해야합니다.

```python
modul1.f()
```

항상 그 함수가 어떤 모둘껀지를 명시적으로 저렇게 적어줘야되는거죠. 근데 `modul1`이면 그래도 그나마 짧다쳐도, 막 10글자가 넘어가는 이름의 모둘에서 함수를 가져온다면, 작성하기도 귀찮고 무엇보다도 가독성이 떨어질 수 있어서, 이를 위해서 `[as name]`의 `name`에 그 모둘을 부르는 또 다른 이름을 적어서 줄이는거에요.

위에서 모둘은 모둘만의 스코프가 있어서, 모둘의 놈들을 직접 우리의 스코프에서 볼 수는 없음을 설명했었는데요,

`modul1.f()`처럼 `f`의 앞에 이 함수의 출신인 `modul1.`을 붙여주는거도 그 때문인겁니다.

근데 이렇게 계속 `modul1.`을 붙여주기 너무 귀찮아서, 그냥 마치 그 함수를 우리의 스코프의 친구로 사용하고싶을 수 있는데요,

이렇게 필요에 따라서 이 모둘이 포함하고 있는 클래스나 함수를 가져와야 할 경우, 우리는 다음의 명령어를 사용할 수 있습니다.

~~~python
from modulname import name1 [as othername], name2, ...
~~~

이 경우, `modulname`에 들어있는 클래스/함수/변수들인 `name1`, `name2`,... 들은 우리의 스코프에서 그대로 사용이 가능합니다.

다시 말해 `modul1.f()` 대신, `f()`를 바로 사용하실 수 있어요.

물론 이 경우, 이미 우리의 스코프에서 정의된 함수/클래스/변수의 이름과 모둘의 함수/클래스/변수의 이름이 충돌할 수 있어서, 

경우에 따라서는 `[as othername]`에 `othername`으로 모둘의 함수/클래스/변수의 이름을 바꿔주는 게 유용할 수 있습니다.


## 수학 모둘(Math)

모둘 `math` 는 수학에서 사용되는 많은 함수 및 상수들을 포함하고 있습니다.

예를 들어, `sin`, `cos`, `tan`, `exp`, `log`, `sqrt`, `pi`, `e` 등의 놈들이 있어요.

이 모둘을 사용하기 위해서는 다음과 같이 `math` 모둘을 `import`해야합니다 :

```python
import math
print("직각의 사인은 :", math.sin(math.pi/2))
print("직각의 코사인은 :", math.cos(math.pi/2))
print("직각의 탄젠트는 :", math.tan(math.pi/2))
print("2의 제곱근은 :", math.sqrt(2))
print("자연 상수는 :", math.e)
print("자연 상수의 자연 로그는 :", math.log(math.e))
print("지수 함수의 1에서의 함수값은 :", math.exp(1))
```

그 외에도 `math` 모둘에는 반올림/올림/내림 함수, 역삼각함수들, 쌍곡선 함수 등의 많은 놈들이 포함되어 있습니다.

````python
import math
print("직각의 사인은 :", math.sin(math.pi/2))
print("직각의 코사인은 :", math.cos(math.pi/2))
print("직각의 탄젠트는 :", math.tan(math.pi/2))
print("2의 제곱근은 :", math.sqrt(2))
print("자연 상수는 :", math.e)
print("자연 상수의 자연 로그는 :", math.log(math.e))
print("지수 함수의 1에서의 함수값은 :", math.exp(1))
````

그리고 위에서도 설명한대로, 우리는 `from [modulname] import [name1] , [name2],...`를 이용하여 특정 대상들만을 가져와서 `math`를 붙이지 않고도 사용 가능합니다.

````python
from math import sin, cos, tan, sqrt, e, log, exp

print("직각의 사인은 :", sin(math.pi/2))
print("직각의 코사인은 :", cos(math.pi/2))
print("직각의 탄젠트는 :", tan(math.pi/2))
print("2의 제곱근은 :", sqrt(2))
print("자연 상수는 :", e)
print("자연 상수의 자연 로그는 :", log(e))
print("지수 함수의 1에서의 함수값은 :", exp(1))
````

근데 프로그램의 진행 과정 중 저렇게 불러온 특정 함수나 변수, 상수, 클래스를 삭제하고 싶을 수 있는데요,

이 때는 키워드로 `del`을 사용하시면 됩니다.

````python
del cos
cos(pi)
````

그러면 이렇게 `cos`함수의 존재를 우리의 스코프에서 모르기 떄문에, `cos` 함수는 그대로 사용할 수 없게 됩니다.

만약 특정 모둘로부터 모든 함수/클래스/변수를 다 불러오고 싶다면, `from modulname import *`라고 작성하시면 되는데요,

사실상 이 명령어는 잘 사용되지 않습니다. 이름이 겹칠 위험성도 있으며,특정 함수가 어떤 모둘의 함수인지,

아니면 심지어 이 함수가 모둘의 함수가 아니라 우리가 직접 작성한 함수인지의 여부들까지 불명확해지거든요.

그냥 이런 명령어도 있는거만 아시면 됩니다.
