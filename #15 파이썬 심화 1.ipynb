# 클래스에 대해 더 자세히

우리는 이미 클래스라는 개념을 다뤄봤습니다. 파이썬의 모든 구성 요소들은 객체들이고, 이러한 객체들의 설계도가 클래스였죠.

이번 포스팅에서는 단순히 클래스를 정의하고 사용하는걸 넘어서서 조금 더 클래스에 관해 자세하게 다루려고 합니다.

## 네임 스페이스와 클래스 변수

클래스를 만든다는 행위는 그냥 하나의 객체를 만든다는 의미 뿐 아니라, 새로운 네임 스페이스(Name space)를 정의한다는 의미를 가집니다.

네임 스페이스는 간단하게 말해서, 구체적인 객체를 구현하기 위해서 특정 데이터와 기능을 묶어 관리하는 공간을 의미합니다.

예를 들어 다음의 예시를 보시죠.

```python
    class Account:
        acc_type = "Bank"
    
        def __init__(self, iban, balance):
            self.iban = iban # 계좌 번호
            self.balance = balance # 잔고
```

간단한 계좌를 구현하는 클래스입니다. 위와 같이 `Account`라는 클래스를 정의한다면,

이 클래스는 보시다시피 `acc_type = "Bank"`라는 변수, 그리고 생성자로 구성되는데요, 저걸 코드로 실행하면 파이썬은 

`Account`라는 클래스를 구성하는 요소가 `acc_type`이라는 변수와 생성자 `__init__()`라는 정보를 저장합니다. 

이 정보가 저장되는 공간을 네임 스페이스라고 하는거에요.

그리고 이  `Account`로 생성되는 인스턴스들은 모두 `iban`, `balance`라는 애들을 가지고 있으나, 값이 다른데요,

따라서 인스턴스들의 네임 스페이스에는 개별 인스턴스마다 다른 `iban`, `balance`값들의 정보를 저장하는 곳을 의미합니다.

근데 `acc_type`는 뭘까요? 저렇게 클래스에서 가장 상위 스코프(`class Account :` 에서 가장 상위)에서 정의되는 변수를 

**클래스 변수**(Class variables)라고 합니다. 클래스 변수는 그냥 간단히 말해 모든 인스턴스들이 다 가지고 있는 공통적인 변수에요.

클래스 변수는 그 클래스의 인스턴스들을 통일적으로 분류할 수 있는 하나의 고정된 상수같은 친구입니다.

네임 스페이스로 설명하자면, 이런 클래스 변수는 클래스의 네임 스페이스에 들어있지만, 인스턴스들은 저걸 직접 가지고 있지 않으며,

클래스의 네임 스페이스에 있는 클래스 변수를 공유하는 느낌으로 이해할 수 있습니다.

그에 비해서 생성자로부터 받을 수 있으며, 개별 인스턴스들의 네임 스페이스에 직접적으로 저장되는 친구들인

`iban`, `balance`는 인스턴스 변수(Instance varialbes) 혹은 속성 변수(Attribute variables)라고 구분합니다. 이들은 개별 인스턴스들의 상태를 표현합니다.

요약하면 모든 인스턴스는 속성 변수를 가지고 있으며, 클래스 변수는 클래스에 들어있으나, 접근 권한이 있는 정도인겁니다.

여기서 접근 권한은 오로지 그 클래스 변수가 어떤 클래스 변수인지를 읽을 수 있다는 정도의 접근 권한입니다.

따라서 개별 인스턴스에다가 클래스 변수를 따로 따로 설정해줄 순는 없으며, 기본적으로 클래스 변수는 모든 객체들이 동일하게 가져야 합니다. 

예를 들어 다음과 같이

`````python
class Account:
    acc_type = "Bank" # 계좌가 일반 은행 계좌임을 알려줌.

    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고

acc1 = Account(15571557, 30.0)
acc2 = Account(888484888484, 50.0)
`````

두 개의 Account 변수를 정의하는 경우, 이 놈들의 계좌 번호와 잔고는 서로 다릅니다. 하지만 결국 이들이 은행 계좌라는 `acc_type`은 변하지 않습니다.

`````python
print(acc1.acc_type)
print(acc2.acc_type)
`````

보시다시피 결국 모두 같은 `"Bank"`라는 `acc_type`를 가져올 수 있습니다. 

이는 저 `acc_type`이 개별 인스턴스가 아닌, 그들의 설계도인 클래스 `Account`에 들어있고, 이를 그냥 빌려와서 출력하는거기 떄문입니다.

어쨌든 어떤 인스턴스든 은행 계좌잖아요? 이런 공통적인 특징을 위한 변수가 클래스 변수인거에요.

재밌는 점은, 우리는 Account의 클래스 변수를 다음과 같이 바꿔줄 수도 있습니다.

`````python
Account.acc_type = "Stock"
`````

클래스 변수를 주식 계좌를 의미하는 `"Stock"`으로 바꿨는데요, 이렇게 클래스 변수를 바꿔버리면 개별 인스턴스들의 클래스 변수도 같이 변합니다 :

`````python
acc1.acc_type
`````

그리고 위에서는 변하지 않는다, 공통적인 특징을 위한 변수다 라고는 했으나, 억지로 변경하는 것 자체는 가능합니다.

`````python
acc2.acc_type = "Bank"

acc2.acc_type
`````

하지만 이건 엄밀히 말하면 "클래스 변수의 값을 바꾸는 것"은 아닙니다.

기본적으로 `acc1.acc_type`의 구문을 보면 파이썬은 저 인스턴스가 아니라, 저 인스턴스의 설계도인 클래스로 넘어가서

거기에 있는 `acc_type`이라는 이름의 변수를 읽어서, 그 값을 반환합니다. 위에서 설명한대로 인스턴스가 직접 이걸 가지고 있는게 아니라,

클래스에 정의되어 있는 클래스 변수에 접근할 권한을 가지고 있기 떄문입니다.

이는 위에서도 설명한 클래스 변수 속성 변수의 차이점입니다. 속성 변수는 그 자체로 인스턴스에 종속되지만, 클래스 변수는 클래스에 종속됩니다.

따라서 만약 `acc2.acc_type = "Bank"`의 형태로 마치 "인스턴스의 클래스 변수를 변경"하려는 시도를 하더라도,

실제로는 클래스 변수가 변경되는 게 아니라, `acc2`에 `acc_type`이라는 이름의 새로운 속성 변수를 정의하여, 거기에 값을 `"Bank"`로 넣는게 됩니다.

이 때문에 `acc2.acc_type`을 위와 같이 불러온다면, 우리가 보는건 실제로 변한 클래스 변수가 아니라

방금 생성된 `acc2`의 새로운 속성 변수 `acc_type`의 값인거에요. 이를 클래스 변수를 섀도잉(Shadowing)한다고도 합니다.

말 그대로 같은 이름의 속성 변수를 정의하여 기존의 클래스 변수를 가려버리는거죠.

클래스 변수는 그대로 있는데, `acc2.acc_type`를 한다면 반환되는건 클래스 변수가 아닌 새로 생성된 같은 이름의 속성 변수일 뿐입니다.


그리고 계속 설명한대로 종속 변수들은 그냥 평범하게 값을 읽어오거나 수정하셔도 좋습니다. 이건 이미 우리가 전에도 하던거였죠.

클래스 변수는 특히 리스트와 같이 변경 가능한(Mutables) 객체들을 사용할 때 혼란이 커집니다.

`````python
class Account:
    acc_type = "Bank" # 계좌가 일반 은행 계좌임을 알려줌.
    benefit = [] # 계좌별 혜택을 넣어놓기 위한 리스트

    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고

    def add_benefit(self, event) :
        self.benefit.append(event)

acc1 = Account(155788848, 0.0)
acc2 = Account(4420, 20.0)

acc1.add_benefit("우수 고객 혜택")
acc2.add_benefit("신규 고객 혜택")

acc1.benefit
`````

위에서 `acc_type`을 변경했을 때만 하더라도, 스트링이 Immutable이다보니까, 개별 인스턴스를 바꾼다고 클래스 변수가 바뀌지는 않았습니다.

그저 설명하대로 개별 인스턴스에서 `acc_type`라는 형태의 추가적인 속성 변수를 만들어냈을 뿐이죠.

하지만 만약 저렇게 `benefit`이라는 리스트같이 Mutable을 사용한다면, 개별 인스턴스에서의 클래스 변수 변경이

실제로 전체 클래스 변수의 변경을 야기할 수 있습니다.

그 이유는 간단해요. 우리의 클래스 변수 `benefit`는 빈 리스트를 하나 바라보고 있는데요,

`add_benefit`으로 개별 인스턴스에 혜택을 추가하더라도 이 개별 인스턴스의 클래스 변수는 실제 클래스의 클래스 변수를 바라보기 떄문에,

`benefit`이 바라보는 바로 그 빈 리스트에 위의 코드대로 차례대로 `"우수 고객 혜택"`, `"신규 고객 혜택"`을 채우게 됩니다.

원래 이런 혜택은 개별 고객마다 다 다른거여야 하는데, 저렇게 하는 경우 모든 고객에게 적용되어 버리죠.

따라서 이렇게 인스턴스마다 다른 개별적인 친구들은 무조건 속셩 변수로 사용하는 게 권장됩니다.

`````python
class Account:
    acc_type = "Bank" # 계좌가 일반 은행 계좌임을 알려줌.

    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고
        self.benefit = [] # 계좌별 혜택을 넣어놓기 위한 리스트

    def add_benefit(self, event) :
        self.benefit.append(event)

acc1 = Account(155788848, 0.0)
acc2 = Account(4420, 20.0)

acc1.add_benefit("우수 고객 혜택")
acc2.add_benefit("신규 고객 혜택")

acc1.benefit
`````

그러면 이렇게 의도한대로 각 고객에 맞는 혜택을 저렇게 저장시킬 수 있습니다.

또 메소드들은 클래스 밖에서도 정의될 수 있습니다.

````python

def add_benefit(self, event) :
        self.benefit.append(event)
class Account:
    acc_type = "Bank" # 계좌가 일반 은행 계좌임을 알려줌.
    ab = add_benefit

    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고
        self.benefit = [] # 계좌별 혜택을 넣어놓기 위한 리스트
````

간단합니다. 밖에다가 정의한 다음에, 클래스 변수로 그 메소드를 저장하면 됩니다.

그러면 파이썬은 알아서 저 `add_benefit`이 `Account`에 종속적인 메소드임을 알게 됩니다.

저게 지금 보면 굳이 왜 저러냐 싶으실 수 있는데요, 만약 외부에 있는 메소드를 우리의 클래스에 추가시키고 싶을 때

굳이 클래스를 건드리지 않고 매우 간단하게 그 메소드를 우리의 클래스에 플러그 인 할 수 있다는 장점이 있습니다.

## 상속

클래스에는 상속이라는 개념이 있습니다. 진짜 말 그대로 상속이에요.

클래스는 객체를 구현하며, 객체는 보통 역할이나 속성에 따라서 카테고리화가 가능합니다.

예를 들어서 "차량"이라는 개념을 사용 목적으로 나누는 경우에는 다음과 같이 카테고리화가 가능합니다.


$$
\text{세단, SUV, 해치백, 쿠페, 컨버터블 등등}
$$

아니면 차량의 동력원(엔진 종류)에 따라서는 다음과 같이 카테고리화가 가능하죠.

$$
\text{내연차, 하이브리드, 플러그인, 전기차, 수소전기차 등등}
$$

이렇게 하위 카테고리로 나눠진 차들은 구분되는 성질(사용 목적/엔진 종류 등)에 따라서 차이를 가지고 있으나,

얘네는 일단 모두 차량이라는 상위 개념으로부터 나눠져 모두 바퀴와 문, 창문이 달려있고, 동력 시스템이 존재하며, 

운전 및 화물적재가 가능한거처럼, 공통적인 특징들을 공유합니다. 대충 무슨 말을 하고 싶은지 감이 잡히실겁니다.

상위 개념(차량)들은, 더 구체적으로 들어가면 기준으로 잡는 속성(사용 목적/엔진 종류)에 따라서

더 아래의 하위 개념으로 나눌 수 있는데요, 이 하위 개념들은 모두 상위 개념을 특징들을 공유합니다.

따라서 우리는 이 하위 개념의 친구들을 각각 따로 따로 하나씩 구현하는 대신,

상위 개념을 먼저 정의하고, 그걸 토대로 더 특수한 차원에 있는 하위 개념을 구현하는 게 훨씬 안정적이고, 효율적입니다.

따라서 객체 지향 프로그래밍(클래스로 객체를 정의해서, 그 객체간의 관계를 짜서 프로그램을 만드는 것)에서는

이렇게 상위 객체로부터 하위 객체를 상속시켜 다루는 경우가 많습니다.

당장 지난 포스팅에서 다룬 SymPy의 `Add`, `Mul`, `Pow`라는 표현을 위한  객체 타입들은 `Expr`이라는 상위 객체로부터 상속된 개념입니다.

계속 다루던 계좌 클래스를 다시 보시죠

`````python
class Account:
    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고
`````

이번에는 `Account`객체를 저렇게 대충 하나 정해놓았다고 해보죠.

이 객체를 여러 특수한 객체로 나누고 싶을 떄, 우리는 상속의 개념을 이용하여 다음과 같이 구현합니다.

`````python
class Bank(Account) :
    acc_type = "Bank"
    
class Stock(Account) :
    acc_type = "Stock"
`````

간단합니다. 저렇게 클래스를 똑같이 정의하는데, 이제는 괄호 ()안에 어떤 놈으로부터 상속을 할 것인지를 지정할 수 있습니다.

그러면 저 `Bank`, `Stock`의 인스턴스들은 `Account`가 가지고 있는 생성자 및 속성 변수를 같이 공유합니다 :

`````python
a1 = Bank(1557,88848.4) # 같은 생성자 사용
print(a1.iban,a1.balance,a1.acc_type) # 속성 변수들을 출력
`````

그리고 `acc_type`는 `Account`의 네임스페이스에는 없고, `Bank` 및 `Stock` 클래스의 네임 스페이스에만 있으므로,

그냥 보통 방법대로 `Account` 인스턴스를 생성하면, 당연히 이 친구들은 `acc_type`도 없습니다 :

`````python
a = Account(123,1414.0)
a.acc_type
`````

이런 상속을 사용할 때 가장 주의할 부분은 상위 객체의 함수를 하위 객체 클래스에서 오버라이드, 즉 덮어씌울 떄 입니다.

위에서 언급한대로 하위 객체 `Bank`, `Stock`은 상위 객체의 생성자를 그대로 가지고 있는데요,

따라서 만약 `Bank`, `Stock`의 생성자에 그냥 나이브하게 추가하려는 속성 변수만을 추가로 받을 수 있도록 설정하면 오류가 생기게 됩니다.

예를 들어 `Bank`의 생성자에 관리 은행 지점을 표현하기 위한 `loc` 변수를 추가해보죠.

`````python
class Bank(Account) :
    acc_type = "Bank"

    def __init__(self, iban,balance,loc) :
        self.loc = loc
a1 = Bank(1557,88848.0,"Seoul")
`````

상식적으로는 뭐가 문제지 싶으실겁니다. 이미 우리는 생성자가 있으니, 추가적인 변수를 만들어 생성자를 수정하더라도

그냥 저렇게 추가 변수만 적어줘도 큰 문제가 없을거 같죠.

하지만 저렇게 생성자를 직접 `def`하는 순간, `Bank`는 기존의 `Account`가 가지고 있던 생성자를 무시하고,

새롭게 정의된 생성자만을 인식하게 됩니다. 따라서 만약 저렇게 한다면, `iban`과 `balance`의 값이 제대로 인스턴스에 들어가지 않습니다 :

`````python
a1.iban
`````

보시다시피 제대로 들어가지 않는걸 넘어서서, 상속받았던 `iban`과 `balance`의 인스턴스 변수 자체가 사라지게 됩니다.

따라서 만약 생성자를 덮어씌우는 경우에도 생성자의 내부에서 기존의 생성자를 불러와서 그걸 통해서 변수값들을 지정해야합니다.

`````python
class Bank(Account) :
    acc_type = "Bank"

    def __init__(self, iban,balance,loc) :
        Account.__init__(self,iban,balance) # 이 시점에서 Account의 생성자를 불러와서 우리의 인스턴스에 iban, balance 변수를 지정합니다.
        self.loc = loc

a1 = Bank(1557,88848.0,"Seoul")
print(a1.iban, a1.balance, a1.loc)
`````

그러면 이렇게 보시다시피 모든 인스턴스 변수들이 맞게 잘 구현되는걸 확인하실 수 있습니다.

물론 애초에 그냥 저렇게 덮어씌울 때

```python

def __init__(self,iban,balance,loc) :
    self.iban = iban
    self.balance = balance
    self.loc = loc
```

이런식으로 기존의 상위 객체의 속성 변수들을 복붙해와서 전부 한번씩 새로 다 적어줘도 문제가 없습니다.

하지만 이러면 사실상 상속을 사용하는 의미가 별로 없죠 ㅋㅋ 

특히나 지금은 간단해서 그렇지, 나중에 속성 변수가 엄청 많은 복잡한 프로그램을 다룰 떄에는 저렇게 복붙해오는 것도 다 비용이라서

그냥 위에서 했던거처럼 상위 객체의 생성자를 불러오는 게 더 낫습니다.

그리고 위에서처럼 `Account.__init__()`처럼 직접 상위의 생성자를 불러오는 대신, 다음과 같은 방법도 사용하실 수 있습니다.

```python
 def __init__(self, iban,balance,loc) :
        super().__init__(iban,balance) # 이 시점에서 Account의 생성자를 불러와서 우리의 인스턴스에 iban, balance 변수를 지정합니다.
        self.loc = loc
```

여기서 `super()`는 상위 생성자에 지금 우리의 인스턴스를 던져줘서, 상위 객체의 속성 변수를 채우는 역할을 합니다.

그렇게 상위에도 있던 속성을 채운 후 `Bank`에만 있는 `loc`라는 속성은 저렇게 `self.loc = loc`로 채우는거죠.

`super()`는 기본적으로 지금 객체의 바로 한 단계 상위 객체를 부릅니다.

이건 지금처럼 부모가 하나인 상속 관계에서 중요한 이야긴 아니지만, 부모가 여러개인 경우에는 좀 달라집니다.

이 떄는 MRO(Method Resolution Order)라는 규칙에 따라서 누구 부모로부터 이를 불러올지를 결정합니다. 지금은 그냥 그렇다는 점만 언급하고 넘어갑니다.

우리는 위의 이야기를 생성자를 예시로 설명했지만, 당연히 이건 예시일 뿐, 설명한 내용은 생성자 외에 다른 메소드들도 똑같이 적용됩니다.

하위 객체의 메소드를 오버라이드 할 경우 그 메소드의 상위 클래스 버젼은 잊혀지는건 똑같아요.

따라서 메소드를 완전 새로 구성하거나, 아니면 조금 확장하는 게 아닌 이상은 굳이 오버라이드 하지 않는 게 좋을 수도 있습니다.

물론 필요시에는 위에서 설명한 방법을 사용하셔서 요리조리 잘 원하는 결과를 만드셔야합니다.

## 프라이빗 변수(Private Variables)

다른 일반적인 프로그래밍 언어들과 달리, 파이썬에서 변수의 스코프는 오로지 코드 블록간의 우위 및 특정 명령어(`def`, `class` 등)에 따라서

LEGB라는 규칙에 의해서만 결정되며 변수의 유효 범위를 명시적으로 지정하는 직접적인 방법은 없습니다.

클래스의 맥락에서 보통 자바같은 언어에서는 각 인스턴스의 속성 변수값을 의도치 않게 변경하는걸 막기 위해서

속성 변수들을 `private`이라는 형태로 정의하여, 직접적인 속성값 변경을 차단하는데요, 이런 기능이 파이썬에는 없다는거에요.

떄문에 이런 기능을 어느 정도 구현하기 위해서 파이썬에서는 **네임 맹글링**(Name mangeling)을 사용합니다.

사용법은 심플합니다. 예를 들어 위의 은행 계좌의 맥락에서 추가로 고객들의 비밀번호 `password`를 구현 할 때,

보안상 문제를 해결하기 위해서 접근을 제한하고 싶다면, 우리는 이를 다음과 같은 형태로 정의합니다.

```python
__password
```
간단합니다. 그냥 언더스코어 `_`를 변수의 이름 앞에 2개 붙이면 됩니다.

저런식으로 변수를 명명한다면 코드 상 `__password` 혹은 `password`를 호출하더라도, 아무거도 발견되지 않습니다.
 
저렇게 언더스코어를 두 개 붙여줌만으로 변수 `__password`는 `_Account__password`라는 이름이 되기 떄문에, 

`_Account__password`라고 직접적으로 이름을 적지 않는다면 `__password`에는 접근이 불가능합니다.

다시 말해서 `__password`를 다른 이름(`_Account__password`)으로 숨기는거죠.

`````python
class Account:
    def __init__(self, iban, balance, password):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고
        self.__password = password # 비번

a = Account(1557,88848.4,'0000')

#a.password
a.__password
`````

보시다시피 `a.password`를 불러오든, `a.__password`를 불러오든 똑같이 그런건 없는뎁쇼? 라는 결과가 나타납니다.

하지만 `_Account__password` 인식이 정상적으로 됩니다.

`````python
a._Account__password
`````

보면 아시겠으나 네임 맹글링은 실제로 강력하게 저 변수에 접근하는걸 차단하는 것이 아니라, 그냥 딱 숨기는 역할 밖에 하지 않습니다.

그래서 자바랑 보안 성능도 다릅니다. 자바에서 `private`으로 지정된 변수들은 컴파일러 수준에서 차단되기 떄문에 만약 private인 변수를

누군가 불러온다고 하더라도, 컴파일러가 이걸 막습니다. 따라서 자바에서 `private` 변수는 해킹 시도도 어느 정도 막을 수 있으며,

변수 안의 값이 프로그래머의 의도대로 유지되도록 보호할 수 있습니다.

반면 파이썬의 네임 맹글링은 진짜 그냥 딱 코드 상에서 숨기는 역할 정도이구요, 프로그래머는 원한다면 언제든 저 변수에 접근이 가능합니다.

따라서 네임 맹글링은 코딩하는 과정에서 우연찮게 저 변수를 사용하거나 실수로 값을 바꾸는 경우를 예방하는 정도의 역할을 하는거지, 

완전한 차단은 아니란 점을 아셔야합니다.

## 매직 메소드(Magic Methods)

매직 메소드, 특수 메소드 혹은 던더 메소드(Dunder Method)라 불리는 메소드들은  

`+`나 `<` 등의 특정 객체들을 사용할 때 파이썬이 호출해오는 메소드를 의미합니다.

우리가 지금까지 사용했던 사칙연산자나 비교연산자들도 사실 말이 연산자지, 진짜 연산자는 아닙니다.

정확히 말하면 사용하는 형태는 연산자이지만, 결국 얘네들도 다른 메소드를 불러오기 위한 단축키같은 역할이었던거죠.

예를 들어 `+`로 객체를 더할 때에는 `__add__(obj)`라는 매직메소드가 호출됩니다.

따라서 `1 + 2`의 덧셈도 다음과 같이 직접적으로 `__add__()`를 호출하여 수행할 수도 있습니다.

`````python
(1).__add__(2)
`````

이거는 `int`나 `float`같은 객체 뿐 아니라, 모든 객체에 해당되는 이야기입니다.

예를 들어 위에서 정의한 `Account`객체도 얘네끼리 `<`를 사용하면, 이런 매직 메소드가 호출됩니다 :

`````python
class Account:
    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고

a1 = Account(2400,4420.0)
a2 = Account(1557,88848.0)
a1<a2
`````

하지만 당연히, 이 `<`가 호출하는 매직 메소드는 우리의 `Account` 버젼으로는 작성되어있지 않기 떄문에, 당장은 사용이 불가능합니다.

이런 매직 메소드는 예전에도 다뤘던 덕 타이핑의 아이디어로, 새로 생성한 객체에 대해서는 항상 새롭게 오버라이드해야합니다.

우리의 경우 `<`로 두 계좌의 잔고를 비교하여 반환하는 매직 메소드를 호출하도록 구현할 수 있습니다. `<`는 메소드 `__lt__()`를 호출합니다.

`````python
class Account:
    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고

    def __lt__(self, acc) :
        return self.balance < acc.balance

a1 = Account(2400,4420.0)
a2 = Account(1557,88848.0)
a1<a2
`````

그러면 우리는 위와 같이 `Account` 인스턴스끼리 `<`를 사용할 시 저렇게 잔고를 비교하는 매직 메소드가 호출되도록 설정할 수 있습니다.

이러한 매직 메소드들 중 대표적인 친구들은 다음과 같습니다.

- `__init__`: 인스턴스 생성 시 호출되는 생성자
- `__repr__`: 인스턴스를 읽을 수 있는 형태로 표현하여 반환함
- `__str__` : `print()` 혹은 `str()`을 사용했을 때, 출력되어야하는 스트링 모양을 반횐
- `__eq__`: 두 인스턴스가 같은지 확인하여 결과를 반환 (연산자 `==` 사용 시 호출)
- `__ne__`: 두 인스턴스가 다른지를 확인하여 결과로 반환 (연산자 `!=` 사용 시 호출)
- `__add__`: 덧셈 (`+` 사용 시 호출)
- `__sub__`: 뺄셈 (`-` 사용 시 호출)
- `__mul__`: 곱셈 (`*` 사용 시 호출)
- `__truediv__`: 나눗셈 (`/` 사용 시 호출)
- `__floordiv__`: 정수 나눗셈 (`//` 사용 시 호출)
- `__mod__`: 나머지 연산 (`%` 사용 시 호출)
- `__pow__`: 거듭제곱 연산 (`**` 사용 시 호출)
- `__lt__`: 좌변이 우변보다 작은지 비교 (`<` 사용 시 호출)
- `__le__`: 좌변이 우변보다 작거나 같은지 비교 (`<=` 사용 시 호출)
- `__gt__`: 우변이 좌변보다 큰지 비교 (`>` 사용 시 호출)
- `__ge__`: 우변이 좌변보다 크거나 같은지 비교 (`>=` 사용 시 호출)


물론 이 친구들 말고도 다양한 매직 메소드가 있습니다. 대표적인 예시들이에요.

### 이터레이터(Iterator)

클래스로 객체를 정의했다면, 우리는 많은 인스턴스를 여러 개 생성하여, 그걸 하나의 상자에 담고 싶을 수 있습니다.

예를 들어, 계좌인 `Account` 인스턴스를 여러 개 생성 후, `AccountList`라는 여러 계좌의 묶음에 담는거처럼요.


```python
class AccountList:
    def __init__(self, accs):
        self.accs = accs # AccountList가 담고 있는 Account들

GookMin = AccountList([a1,a2,a3]) # a1, a2, a3은 임의의 Account 인스턴스임을 가정
```

그런데 이런 `Account`의 컨테이너가 주어진다면, 우리는 다음과 같이 `for`문 등으로 반복 명령을 수행할 필요가 있을텐데요,

```python
for acc in GookMin:
    print(acc)
````

무작정 저렇게 `for`문을 사용하면 오류가 발생합니다. 그 이유는 `AccountList`의 `Iterator`가 정의되지 않았기 떄문입니다.

예전에도 다뤘듯이, `for`문이 돌 수 있는건 `Iterable` 객체이며, 이러한 `Iterable` 객체는 `for`문 등에 사용했을 때,

내부적으로는 `Iterator`라는 친구를 생성하고, 이를 통해서 `Iterable`의 모든 원소를 딱 한 번씩 돌며 찍먹하는데요,

우리는 따로 이러한 `Iterator`가 어떤지를 지정하지 않았기 떄문에, 사용조차 불가능한거죠.

이 때는 다음 매직 메소드들을 오버라이드 해주면 됩니다.

`__iter__` : `for`문 등에서 사용 시 어떤 `Iterator`가 반환되는지를 지정(대부분의 경우 인스턴스 자기 자신, 즉 자기 자신을 기준으로 훑고가는)

`__next__` : 그 `Iterator`가 어떻게 `AccountList`가 저장하고 있는 `Account`들을 훑고 갈지를 결정하여 다음 원소를 반환

이 둘은 다음과 같이 구현할 수 있습니다.

`````python
class AccountList:
    def __init__(self, accs):
        self.accs = accs # AccountList가 담고 있는 Account들
        self.index = -1 # 위에서 추가적으로, 지금 이터레이터가 어딜 읽고 있는지를 표시하기 위해서 속성 변수를 추가합니다. 
                        # 굳이 index가 0이 아니라 -1인 이유는 아래에서 확인 가능합니다.
    def __iter__(self) :     
        return self # 우리의 AccountList 자기 자신이 이터레이터가 됩니다. 자기 자신을 기준으로 값들을 하나씩 돌거라서요

    def __next__(self):

        self.index = self.index + 1        # 다음 칸을 출력해야하니, 우선 한 칸 앞으로 인덱스를 옮깁니다.
        
        if self.index == len(self.accs) : # 만약 이터레이터가 모든 성분들을 다 훑어 그 칸에 더 이상 성분이 없다면,
            self.index = -1               # 재사용을 위해 인덱스를 -1으로 돌림
            raise StopIteration           # 이제 더 돌 원소가 없으므로, StopIteration을 레이즈
            
        return self.accs[self.index]     # 아직 성분이 더 남아 있다면 값을 반환
`````

간단합니다. 설명한대로 `__iter__`는 그냥 보통 자기자신을 저렇게 반환하면 되구요,

`__next__`는 말 그대로 다음 원소를 반환해야하므로, 만약 우리의 인덱스가 가지고있는 데이터의 개수 범위를 넘지 않았다면

저렇게 다음 원소를 반환합니다. 만약 개수 범위를 넘었다면 `self.index`를 다시 -1로 초기화시키고, `StopIteration`을 레이즈합니다.

`for`문 등은 이러한 `StopIteration`이 레이즈될 시 반복문을 종료하기 떄문입니다.

이제 우리는 다음과 같이 `for`문을 자유롭게 사용 가능합니다.

`````python
class Account:
    def __init__(self, iban, balance):
        self.iban = iban # 계좌 번호
        self.balance = balance # 잔고

    def __str__(self) : # 계좌 정보를 스트링 형태로 반환하는 함수. print()나 str() 사용 시 출력되는 매직 메소드
        return f"계좌 번호 : {self.iban}\t 잔고 : {self.balance:.4f}"

a1 = Account(1,1557.1557)
a2 = Account(2,88848.4)
a3 = Account(3,2400.2400)
a4 = Account(4,44.20)

GookMin = AccountList([a1,a2,a3,a4])

for acc in GookMin :
    print(acc)
    
`````
