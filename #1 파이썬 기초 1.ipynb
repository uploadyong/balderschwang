## 파이썬의 명령어

파이썬의 프로그램은 컴퓨터가 수행할 미션들을 순서에 맞게 배치시킨 **명령어**들의 연속입니다.

명령어들은 크게 단순 명령과 복합 명령으로 나뉩니다.


### 단순 명령

단순 명령은 파이썬 프로그램의 초석입니다. 단순 명령은 단일의 단순한 명령들을 의미하며, 보통 단순 명령들은 한 줄에 작성됩니다.

단순 명령의 예시는 다음과 같습니다.

- 초기화 등 변수에 값을 부여하는 것(`x = 1`, `v = (1, 2)`)
- 표현 및 값의 계산 (`x + 2`)
- 프로그램 흐름을 제어하는 컨트롤 명령들 (`return`, `continue`, …)
- 기타 콘솔 출력 명령 (`print`) 혹은 모둘 임포트 (`import`) 등등

### 복합 명령

복합 명령은 여러 개의 단순 명령들이 하나로 모여서 생긴 복잡한 명령을 의미합니다. 이러한 단순 명령을 모은 그룹을 코드 블록이라고 합니다.

복합 명령은 코드 블록에 여러 단순 명령들을 수행하여, 복잡한 과정이나 논리를 실현시키는 역할을 합니다.

복합 명령의 예시는 다음과 같습니다.

- 함수의 정의 (`def`)
- 조건문 (`if`)
- 반복문 (`for`, `while`)
- 객체 및 클래스의 정의 (`class`)


## 표현(Expression)

표현은 말 그대로, 상징들이 모여서 만들어진 하나의 표현을 의미합니다.

표현의 예시는 다음과 같습니다.

- 수학적 표현 : `3 + 4` 혹은 `a * b`
- 논리 표현 : `a > b` 혹은 `a and b`
- 함수의 소환 :`f(x, y)`
- 기타 결국 어떤 값으로 귀결되는 표현들

표현은 **값**, **변수**, **연산**들이 조합된 어떤 형태를 의미합니다. 표현은 결국 하나의 값으로 귀결되는데요, 이를 반환값(return)이라 합니다. 변수가 등장하는 표현의 경우, 그 변수에 저장된 값에 따라서 반환값이 달라질 수 있고, 이에 따라서 정해진 값이 아닌, 예를 들어 임의의 `x` ,`y`에 대해서, `x+y`를 계산하는 등의 용도로 사용됩니다. 이러한 표현은 계산을 하거나, 데이터 연산을 위해 사용됩니다. 목적에 따라서 표현은 위 예시들과 같이 간단할 수 있고, 여러 값, 변수, 연산을 이용하여 매우 복잡할 수도 있습니다.

파이썬을 효율적으로 사용하고 복잡한 프로그램을 작성하기 위해서는 표현들을 이해하는 것은 매우 중요합니다.


## 명령의 순서

파이썬에서 명령이 수행되는 순서는 매우 중요합니다. 그렇다고 어려운건 아니에요. 엄청 직관적으로, 우리가 글을 읽는 방향대로 명령이 수행됩니다. 왼쪽부터 오른쪽으로, 위부터 아래로, 순서대로요. 파이썬의 경우, 소스 코드의 처음부터 끝까지 이런 순서로 명령이 수행됩니다.

따라서 명령어의 순서를 잘못 배치한다면, 이는 프로그램 자체에 영향을 미쳐서 우리가 원하는 결과를 얻어내지 못할 수 있습니다.

다만 주피터 노트북에서는 여러 코드 셀이 있고, 코드 셀을 독립적으로 실행시킬 수 있습니다. 따라서 다른 보통의 파이썬 IDE 환경과는 동작 방식이 조금 달라요. 예를 들어 아래의 코드 셀을 누르고 `Ctrl + Enter`로 실행한다면

````python
x = 3+4*7
x = 0
x + 1
````

곧 바로 위 코드 셀 왼쪽에 `[1]`이 표기되는걸 확인하실 수 있습니다. 이 `[1]`은 위 코드를 실행시킨 순번을 의미합니다. 즉, 지금 상태에서 노트북은 위의 코드가 이미 실행되어 있음을 알고 있습니다.

이제 다시 아래 코드를 살행한다면

````python
x
````

`[2]`가 생기는걸 알 수 있는데요, 이 `[2]`가 생기기 이전에 우리가 `[1]`에서 코드를 미리 실행시켰기 떄문에, 위 `[2]`의 코드에는 따로 x가 정의되지 않았으나, 이미 `[1]`에서 `x = 0`으로 선언 및 초기화되었으므로, 0이라는 값을 반환합니다.

이렇게 주피터 노트북은 사용자가 코드 셀을 실행한 순서대로 명령어를 다룹니다. 만약 이 순서가 좀 잘못되어서, 코드 셀 순서를 바꾸고 싶다면 상단의 새로 고침과 같은 표시를 누르시면 됩니다.

## 표현과 그의 반환값

표현은 위에서도 설명한대로 값, 변수, 연산의 조합이며, 항상 하나의 값을 가지며, 파이썬-인터프리터에 의해서 이 값은 계산됩니다.

값의 계산에는 순서가 있으며, 표현이 주어지는 경우엔 마지막에 남는 값이 나올 떄까지 파이썬에 의해 그 표현이 계산됩니다.

### 주피터 노트북의 자동 출력 기능

위에서 다룬 코드 셀 예시들에서 알 수 있는 점이 있는데요, 주피터 노트북은 코드의 가장 마지막에 저렇게 표현이 서있다면 자동으로 그 표현의 값을 계산하여 반환합니다. 파이썬은 그렇지 않고, 파이썬에서는 `print()`함수를 사용해야 합니다.

### print() 함수

주피터 노트북이 아닌 다른 파이썬 IDE 환경에서 표현의 계산값을 출력하고 싶다면 `print()` 함수를 사용해야 합니다. 이 함수는 복수의 입력 변수를 받아서, 그 입력 변수를 차례대로 콘솔에 출력합니다. 이 때, 각 입력 변수들은 출력 시 공백으로 구분됩니다.

```python
x = 10
y = 20
print(x,y,x + y)

```

위 코드에서 `print()`함수는 차례대로 `x` , `y`에 저장되어 있는 값과 `x+y`의 표현의 결과물을 콘솔에 출력합니다.

이는 아래 코드 셀에서 확인 가능합니다.

````python
x = 10
y = 20
print(x,y,x+y)
````

## 주석

프로그래밍에서는 소스 코드에 주석을 다는 게 매우 일반적입니다. 주석은 코드의 부분 부분에 추가적인 정보나, 그 코드의 원리의 설명을 적어놓기 위한 도구로, 팀 단위 프로그래밍을 하거나, 오랫만에 보는 코드를 이해하는대에 편리합니다.

소스 코드 내부에서 주석 표시가 있는 부분은 소스 코드가 아닌, 인간의 언어로 이해하여, 컴퓨터는 그 부분을 무시하고 명령을 수행합니다.

주석은 크게 두 가지 종류가 있습니다. 각 행의 소스 코드 옆에 부분적으로 주석을 달 수도 있으며, 특정 부분 전체를 주석화할수도 있습니다.

행 주석을 위한 명령어는 `#`입니다. `#` 이후에 등장하는 모든 문장들은 인터프리터 및 컴파일러에 의해 무시됩니다.

```python
# 한 줄 주석
x = 5  # 보통 이런식으로 그 코드 행에 어떤 일이 벌어지는지를 단편적으로 적을 때 사용

# 혹은 아래와 같이, 특정 코드행을 비활성화 하기 위해서도 사용됨.
# x = 20 
```

그리고 특정 부분 전체를 주석화하는 명령어는 `"""`입니다. 사용법은 다음과 같습니다.

```python
"""
asdasdasd
"""
```

아래 코드 셀을 실행시키면, 실제로 주석으로 지정된 놈들은 무시됨을 알 수 있습니다.

````python
"""
ㅁㅇㅁㄴㅇㅁㄴㅇ
"""
0
# print("sdasd")
````

### 마크다운 셀과 주석의 차이

지금 셀처럼 소스 코드가 아닌 텍스트가 들어오는 텍스트 셀은 주피터 노트북에서 마크다운 셀이라고 합니다.

주피터 노트북을 사용한다면 코드를 설명하기 위한 도구가 이렇게 마크다운 셀도 있고, 주석도 있는데요, 얘네는 각자 장단이 있어요.


**마크다운 셀**
- 문서를 전체적으로 더 자세하게 설명하기 용이함.
- 리스트, 표어(제목처럼 큰 글씨 적는거), 글씨의 강조(굵게, 기울임 등)가 가능함.
- 정보를 적는걸 목적으로 함.

**파이썬 소스 코드 내부의 주석**
- 코드 셀에 직접적으로 작성됨.
- `#` 및 `"""`로 구분되며, 프로그램 작동에 영향을 주지 않음.
- 코드를 그 위치에서 설명하거나, 위험성 설명(이런 결과가 나타날 수 있다.)에 적합함


### 마크다운 셀에서 포맷을 지정하는 법

**굵게**
- 특정 텍스트를 굵게 강조할 때에는 `**`를 사용합니다 : `**굵게**`
- 예시 : **굵게**

*기울임체*
- 기울임체를 위해서는 `*`를 사용합니다 : `*기울임체*`
- 예시 : *기울임체*

`인라인 코드`
- 마크다운 셀 내부에서 특정 명령어, 혹은 키보드 키를 표현할 때 사용합니다 : `` `인라인 코드` ``
- 예시: `인라인 코드`

코드 블록
- 마크다운 셀 내부에도 코드 셀에서처럼 파이썬 코드를 작성할 수 있습니다.
- 이 때, 코드는 실행가능하지 않으며, 그저 코드의 서식을 코드 셀의 서식으로 표현합니다. 
- 코드는 다음과 같이 명령어는 다음과 같습니다 :

\`\`\`python  
여기에 코드 작성  
\`\`\` 

- 예시 :
```python

```

- 순번 없는 리스트를 위해서는 `*` 혹은 `-`를 사용합니다 : `* 아이템` 혹은 `- 아이템`

- 예시 :
    1. 첫번째 아이템
    2. 두번째 아이템

- 예시 : $E=mc^2$

- 예시 :
  $$E=mc^2$$

## 변수

변수는 값을 대변하는 이름이라고 생각할 수 있습니다. 변수는 값을 저장하는 역할로 사용됩니다.

변수는 파이썬을 비롯한 많은 프로그래밍 언어에서 값 할당 연산자인 `=` 를 사용하여 값이 지정됩니다. `=`를 기준으로 왼쪽에는 변수의 이름, 오른쪽에는 그 변수에 저장할 값 혹은 표현이 들어옵니다. 사실 우리는 이를 이미 위에서 봤죠 :

```python
x = 10
```

`=`를 기준으로 왼쪽에는 표현이 올 수 없고, 오로지 변수의 이름에 해당되는 놈이 와야합니다. 다음은 비정상적인 코드입니다 :

```python
x + 10 = 20
```

위에서 다룬 표현은 변수로도 이뤄졌다고 했었는데요, 변수로 이뤄진 표현은 예를 들어 다음과 같습니다 :

```python
x = 10
x + 20  
```

이러한 표현은 결국에는 어떤 값이 되므로, 이 값을 자신의 값으로 하는 또 다른 변수도 만들 수 있습니다 :

```python
x = 10
y = 20
z = x + y
```

이 경우, `z`에는 `x+y`라는 표현의 반환값인 30이 저장됩니다.

다음과 같은 코드는 어떨까요?

```python
x = 10
x = x + 20
```
이 코드는 유효합니다. 문제가 될만하다고 생각하실 부분은 `x = x + 20`인데요, 이는 `x`라는 변수에, `x + 20`라는 표현의 반환값을 저장하라는 의미로, 이 때는 기존에 `x`가 저장하고 있던 10을 가져와서, `10 + 20`의 결과물인 `30`이 `x`에 저장되게 됩니다. 그니까 우선 우변에 있는 표현의 값을 먼저 계산하고, 그 이후에 `x`에 저장하기 떄문에 가능한거죠.

-   모든 파이썬의 값들은 **객체**(Object)입니다. 이에 대해선 나중에 더 자세히 다뤄집니다.

-   **변수**는 정확히 말하면 이 객체들을 참조하는 식별자/이름(Identifier)에 해당됩니다.

### 변수 이름의 규칙

변수의 작명에는 규칙들이 있습니다.

    -   소문자 `a-z` 혹은 숫자 `0-9` 및 언더스코어 `_`로 이뤄짐.
    -   대소구분을 해야됨.
    -   다른 특수 기호(한글 등)을 포함할 수 없음.
    -   숫자로 시작할 수 없음.
    -   *예약어*(Reserved Words/Keywords)는 사용할 수 없음. (`while`, `if`, `for` 등)

예약어는 이미 다른 무언가를 위해 배정된 말 그대로 예약된 언어로, 그 기능을 위해 사용되어야 하므로 변수 이름이 될 수 없습니다.
