## 객체와 타입들

파이썬에서는 모든게 **객체**입니다. 단순한 숫자 `10`, 문자열 `"ㅁㄴㅇ"` 등도 전부 객체(인스턴스)에요.

객체는 컴퓨터의 램(RAM)에 저장되며, 특정 타입을 하나씩 가지고 있습니다.

객체의 타입은 그 객체가 어떤 값을 가질 수 있으며, 그걸 토대로 어떤 연산이 가능한지를 결정합니다.

예를 들어, `int`라는 타입의 객체는 정수값만을 가질 수 있으며, `str`이라는 타입은 문자열을 데이터로 가집니다.

또한 연산에서도 차이가 있는데요, `int`인 두 객체를 `+`한다면, 그 표현은 결국 두 객체의 값을 더한 정수가 됩니다.

반면 `str`인 두 객체를 `+`한다면, 그 표현은 결국 두 `str`들을 이어 붙인 새로운 `str`의 형태가 됩니다. 

이를 순차(Concatenation)라고도 합니다.


파이썬의 놈들이 어떤 타입의 객체인지를 확인하기 위해서는 함수 `type()`를 사용할 수 있습니다 :


```python
type(10)  # <class 'int'>가 출력됨.
type("asdasd")  # <class 'str'>가 출력됨.
```

유의할 점은, 반환되는 값이 클래스 그 자체라는겁니다. `type()`함수는 입력 변수를 받아서, 그 입력 변수가 속한 타입의 클래스 그 자체를 반환합니다. 주피터 노트북에서는 `type(10)`을 한다면 int라는 결과가 나타나는데요, 이걸 보고 반환값이 문자열인 `str`의 `"int"`인가? 싶으실 수 있으나, 그렇지 않습니다 : 

````python
type(10)
````

우선 우리는 파이썬에서 지원하는 가장 기본적인 데이터 타입들을 공부하고, 나중에 좀 더 복잡한 친구들과 만나게 됩니다.

### 다양한 타입들

그냥 바로 어떤 친구들이 있는지 보시죠 :

- `int`: 정수, 예를 들어 `-1`, `2`, `3`
- `float`: 부동소수, 예를 들어 `1.0`, `2.0`, `3.0`
- `str`: 문자열, 예를 들어 `"ㅁㄴㅇ"`, `"ㅂㅈㄷㅂㅈㄷ"`
- `bool`: 부울값, 예를 들어 `True` und `False`
- `list`: 리스트, 예를 들어  `[1, 2, 3]`, `["ㅁㄴㅇ", "ㅂㅈㄷㅂㅈㄷ"]`
- `tuple`: 투펠, 예를 들어 `(1, 2, 3)`, `("ㅁㄴㅇ", "ㅂㅈㄷㅂㅈㄷ")`
- `NoneType`: 오로지 값 `None`을 가지는 특수한 타입
- `function`: 함수를 표현하는 특수한 타입
-   기타 등등 …

여기서 알 수 있는 점이, 파이썬에는 `char`라는 타입이 없다는겁니다. 파이썬에서는 단일 문자와 문자열을 통일하여 다루며, 모두 스트링에 해당되는 `str`의 타입을 가지게 됩니다.

이번 포스팅에서는 여기서 `int`, `float`, `bool`만을 다루며, 다음 포스팅에서 `list`,  `str`을 다루게 됩니다.

### 객체와 변수

지난 포스팅에서 다룬 객체는 변수를 가리키는 이름표라고 할 수 있습니다. 만약 변수가 생성된다면, 동시에 객체가 생성되어 그 객체를 부르는 이름(변수의 이름)이 그 객체를 가리키는 방식으로 정의되는거죠. 만약 변수에 다른 새로운 값이 주어진다면, 이전의 값은 삭제되며, 새로운 값이 그 자리에 서게 됩니다.

예를 들어 다음을 생각하시면 됩니다.

```python
x = 10  # x는 타입이 int이고, 값이 10인 객체를 가리키고 있음.
x = 20  # x는 이제 타입이 int이고, 값이 20인 객체를 가리키게 됨.
```

## 단순 데이터 타입들

단순 데이터 타입은 `int`, `float`, `str` und `bool`을 의미합니다. 이 타입들은 파이썬의 프로그램을 구성하는 가장 작은 단위인 객체들이며, 더 작은 놈들은 없습니다. 다시 말해서, 모든 더 복잡한 개념들의 초석인거죠.

다음의 예시를 보시죠 :

```python
x = 10  # int
y = 10.0  # float
w = True  # bool
z = "asdasd"  # str
```
다 간단합니다. C언어를 아신다면 이 데이터 타입을 표현하는 방식은 크게 다르지 않음을 알 수 있습니다. 차이가 있다면, C에서는 소문자로 시작하는 `true` 혹은 `false` 대신, 대문자로 시작하는 `True` 및 `False`를 참/거짓을 의미하는 값으로 표현합니다.

또한 C 등의 다른 언어와 달리, 변수에는 데이터 타입을 지정해줄 필요가 없습니다.


이제 각 데이터 타입들을 하나씩 보시죠.


### 정수(int)

정수는 소수점이 없는 수라고 생각하시면 됩니다. 

정수에는 다음과 같은 기본적인 수학 연산이 주어집니다 :

```python
x = 10 + 20  # 덧셈
y = 10 - 20  # 뺄셈
z = 10 * 20  # 곱셈
w = 101 // 20  # 정수 나눗셈(나머지는 버림)
v = 101 % 20  # 모둘로 연산(나머지를 계산하여 반환)
u = 10 ** 20  # 거듭제곱
```

간단합니다. 다 이미 아시는 연산들일텐데요, 딱 두 가지 C나 자바와는 다른 차이점이 있습니다.

첫번째로 정수 나눗셈 `//`라는 개념이 존재하는데요, 이 놈은 말 그대로 정수를 위한 나눗셈입니다. 물론 부동 소수에서도 사용이 가능한데요, 이 나눗셈은 연산 시 나머지 및 소수점은 모두 무시하고, 정수부만 반환하는 형태가 됩니다. 물론 피연산자의 한 놈이라도 부동 소수 `float`가 들어있다면, 그 결과물은 다시 `float`로 반환됩니다.

또한 거듭제곱이 보통 `^`로 표현되거나, C 혹은 자바에서처럼 따로 연산으로 제공되지 않고, 함수의 형태로 사용되는데요, 파이썬에서는 `**`가 사용됩니다. 이는 `^`가 다른 연산자(`xor`)로 사용되기 때문입니다.

위 코드는 아래 코드 셀에서 실행 가능합니다

````python
x = 10 + 20  
y = 10 - 20 
z = 10 * 20  
w = 101 // 20  
v = 101 % 20  
u = 10 ** 20  

x, y, z, w, v, u
````

정수의 경우, 다른 언어와 차별되는 점이 있습니다. 정수의 경우 보통 다른 언어들은 최대로 표현 가능한 크기 제한에 있는대에 비해 파이썬에서는 크기 제한이 없어요. 따라서 엄청 말도 안되게 큰 정수들도 파이썬에서는 표현이 가능합니다.

```python
x = 100**1000  # 100의 1000승
```

근데 공짜는 없다고, 이렇게 큰 수는 당연히 계산에 상당히 많은 시간이 소요됩니다.

```python
x = 100**10000000  
y = x + 1  
```
위와 같이 x라는 굉장히 큰 수에 오로지 1을 더하는 간단해보이는 연산도 엄청 오래 걸려요.

아니 심지어 오래 걸리는걸 넘어서서, 파이썬의 규칙상으로는 문제가 없으나, 우리의 컴퓨터가 버티지 못할 수 있습니다. 숫자가 큰 만큼 많은 저장 공간이 필요한데, 이에 따라서 저장 공간이 부족하여 프로그램이 다운될 수 있습니다. 따라서 이론상은 무한한 크기까지 가능하지만, 실제로는 우리의 컴퓨터가 버텨낼 수 있는 정도의 크기까지는 가능하다 정도로 생각하시면 됩니다.

혹은 저장 공간은 ㄱㅊ은데, 콘솔에 출력할 수 있는 문자 수 한계 떄문에 출력이 불가능할 수도 있습니다 :

````python
x = 100**10000000 
y = x + 1 
y
````

기대한대로, 일단 더럽게 오래 걸리고, 나중에 결과도 콘솔 한계를 넘어서서 출력 불가능 판정이 나옴을 알 수 있습니다.


### 부동소수 (float)

부동소수는 정수와 달리 소수점이 있는 숫자들입니다. 얘네들도 정수와 마찬가지로 다음의 연산들이 주어집니다 :

```python
x = 10.1 + 20.2  # 덧셈
y = 10.1 - 20.2  # 뺼셈
z = 10.1 * 20.2  # 곱셈
w = 101.1 / 20.2  # 나눗셈
v = 10.1 ** 20.2  # 거듭제곱
```

다 동일한데요, 나눗셈이 좀 다르죠? 저 나눗셈 `/`는 오로지 부동소수를 위한 나눗셈입니다. 따라서 결과도 소수점을 포함한 부동소수 형태의 결과가 나오며, 심지어 정수들끼리 `/`를 하더라도, 자동으로 파이썬은 연산의 대상이 `float`임을 기대하여, 결과물 또한 `float`로 반환합니다 :


````python
10/5 # 다 정수지만, 부동소수의 결과가 튀어나옴
````

이는 정수 나눗셈 `//`와는 다른 느낌이죠. `//`는 위에서 설명한대로 `float`에서도 사용 가능하며, 피연산자 중 하나라도 `float`라면 그 결과물을 자동으로 `float`로 뱉어줍니다 :

````python
10//3.0
````

하지만 보시다시피, `/`는 무조건 뭐가 들어와도 `float`가 반환됩니다. 자동으로 맞춰주는 그런거 없어요

````python
x = 10.1 + 20.2 
y = 10.1 - 20.2  
z = 10.1 * 20.2  
w = 101.1 / 20.2  
v = 10.1 ** 20.2  

x, y, z, w, v
````

부동소수의 고질적인 문제는 이 놈이 저장되는 방식에 따른 정밀도에 있습니다. 컴퓨터에 숫자는 이진법으로 저장되는데요, 이에 따라서 예를 들어서 `0.1`이라는 십진법의 간단한 소수도, 이진법으로 변환 시 무한 소수의 형태가 됩니다. 그런데 컴퓨터는 이 숫자를 저장하기 위해서 유한한 칸 밖에 없으므로, 적당한 소수점에서 절삭된 근사된 값만을 다룰 수 있는데요, 이런 이유로 간단한 유리수들을 더해주더라도, 그 값이 수학적으로 정확하지는 않게 됩니다 :

```python
x = 0.1 + 0.2  # 0.30000000000000004가 반환됨.
```

부동소수는 일반적인 유리수 표기법 외에도, 지수 표현을 이용해서도 표현이 가능합니다.

```python
x = 1e100 
y = 1e-100 
```
이 때, `1e100`은 $1\cdot 10^{100}$을 의미하며, `1e-100`은 $1\cdot 10^{-100}$을 의미합니다.

````python
x = 0.1 + 0.2 
x
````

````python
10.0**-100 + 1.0
````

부동소수는 유리수들을 근사적으로 표현하기 위해 사용됩니다. 근사적인 이유는 위에서 설명한대로, 저장할 수 있는 비트들의 개수가 한정되어있기 떄문에, 특정 소수점에서 절삭되기 떄문입니다. 예를 들어 다음을 보시죠 :

```python
x = 1.0 / 3.0  # 0.3333333333333333가 출력됨
x
```
이 `x`에는 이론적으로 $\frac{1}{3} = 0.333333...$의 무한 소수가 저장되어야 하지만, 저렇게 무한 소수가 아닌 유한 소수가 저장됩니다.


    


그리고 부동소수는 캔설레이션(Cancellation)이라는 문제도 가지고 있습니다.

캔셀레이션은 거의 같은 두 부동소수를 빼는 과정에서 생겨나는 문제로, 두 부동소수가 서로 빼질 경우, 그 오차는 매우 커집니다.

여기서 오차가 무슨 말일까요? 걍 두 수를 뺴는건데 왠 오차 드립?

우리가 실제로 수학적으로 생각하는 두 실수 $x,\;y$를 빼는 경우, 그 결과물인 $x-y$는 수학에서나 제대로 정의됩니다.

앞서 언급한대로 컴퓨터는 이진법으로 수를 저장하기에, 두 실수 $x,\;y$는 컴퓨터에서 적당한 오차를 가지고 있는 유한 소수 형태의 유리수 $x',y'$의 형태로 표현됩니다. 따라서 만약 컴퓨터에서 $x-y$를 한다면, 그 결과물은 $x'-y'$가 되는데요, 이 결과물 $x'-y'$과 실제 수학적인 결과물인 $x-y$간의 오차를 얘기하는거에요.

다른 연산은 괜찮은데요, $x,y$가 서로 $x\approx y$, 즉 거의 같은 실수일 경우, 실제 수학적 결과물 $x-y$에 비해 $x'-y'$는 매우 큰 오차를 발생시킵니다. 이런 일이 벌어지는대에는 결국 부동소수형의 데이터들이 한정된 정확도를 갖기 떄문입니다. 따라서 이 정확도를 벗어난 범위만큼 차이나는 숫자들에 대해서는 오차가 많이 발생할 수 있는거죠.

이에 대한 정확한 수치 수학적인 이론적 배경은 여기서 다루지 않도록 하겠습니다. 그냥 서로 가까운 숫자들끼리 뺴지는 경우에는 오류가 클 수 있다는 그 결과만 아시면 됩니다. 혹시 궁금하시다면 [블로그 포스팅](https://blog.naver.com/balderschwang/222966790585)을 참고해주세요. 이 포스팅도 근데 어느 정도 이론적 배경을 공유하셔야 이해할 수 있는 이야기라, 단편적으로 저 포스팅으로 정확히 이해하시기에는 어려울 수 있습니다. 다시 말하지만 걍 결과만 알고 넘어가도 무방합니다 ㅋㅋ

예를 들어 다음의 코드를 보시면

```python
x = 1.0000000000000001
y = 1.0
x - y
```

`x`에 저장된 숫자는 파이썬에서 충분히 다룰 수 있는 범위 내에 있는 숫자임에도 불구하고, `x-y`를 실행해버린다면 `0.0`이라는 값이 나옵니다.

````python
x = 1.0000000000000001
y = 1.0
x - y
````

이러한 캔셀레이션을 피하기 위해서 우리는 가능하다면 뺼셈은 없애는 과정을 선호합니다.

가장 대표적인 예시는 다음과 같은 경우입니다.

$$ \sqrt{x + \epsilon} - \sqrt{x},\quad \text{for } x\in \mathbb{R}, (0,\infty)\ni\epsilon\approx 0 $$

이 경우에는 $\sqrt{x+\epsilon}\approx \sqrt{x}$이므로, $\sqrt{x+\epsilon}-\sqrt{x}$에 캔셀레이션이 일어날 수 있습니다.

따라서 우리는 위와 같은 계산을 수행하기 위해서, 다음의 과정을 거칩니다.


$$ \sqrt{x + \epsilon} - \sqrt{x} =(\sqrt{x+\epsilon} -\sqrt{x})\cdot\frac{\sqrt{x+\epsilon} + \sqrt{x}}{\sqrt{x+\epsilon}+\sqrt{x}}=\frac{(x + \epsilon) - x}{\sqrt{x + \epsilon} + \sqrt{x}} = \frac{\epsilon}{\sqrt{x + \epsilon} + \sqrt{x}} .$$

고등학교에서 이걸 유리화였나 그렇게 불렀던 것 같은데, 합차 공식 $ (a+b)(a-b) = a^2 - b^2 $ 를 사용하는겁니다.

실제로 저렇게 변형한 전과 후에는 결과에 차이가 꽤 납니다.

````python
x = 10**6 
x + 0.00001
print(x, x + 0.00001)


notransf = (x + 0.00001)**0.5 - x**0.5

withtransf = 0.00001 / ((x + 0.00001)**0.5 + x**0.5)

print("Direct result : ", notransf)
print("Transformed result :", withtransf)
````

혹은 이거 외에도,

$$ax^2 + bx = c = 0,\quad a\neq 0$$

이런 이차 방정식을 푸는 프로그램을 작성하는 경우, 서로 다른 두 해가 존재하는 경우만 생각한다고 한다면 해는 다음으로 표현됩니다.

$$x_{\pm} = \frac{-b\pm\sqrt{b^2-4ac}}{2a}$$

여기서 해 $x_+ = \frac{-b+\sqrt{b^2-4ac}}{2a}$의 경우, 만약 $b\approx \sqrt{b^2-4ac}$인 경우, 캔셀레이션에 의해 오차가 커질 수 있는데요, 이를 피하기 위해 여기서는 $x_+\cdot x_- = \frac{c}{a}$임을 이용하여, 다음과 같이 두 해를 계산하면 됩니다 :
$$x_- = \frac{-b-\sqrt{b^2-4ac}}{2a},\quad x_+ = \frac{c}{a}\cdot \frac{1}{x_-}$$
그러면 $x_-$에서는 캔셀레이션이 일어나지 않고, $x_+$에서도 마찬가지로 캔셀레이션 없이 결과를 계산할 수 있습니다.

## 부울값 (bool)

부울값은 참 `True`과 거짓 `False`으로 된 값을 의미합니다. 얘네들은 특히나 특정 조건이 만족되는지를 판단하기 위해서 사용됩니다. 그리고 표현들 중, 이러한 부울값을 결과물로 가지는 표현들을 부울 표현이라고 합니다. 부울 표현들은 변수, `True`, `False`, 그리고 논리적 연산자들로 구성됩니다. 대표적인 논리적 연산자들은 `and`, `or`, `not`입니다. 다른 프로그래밍 언어에서는 이들을 기호들로 표현하나, 파이썬에선 걍 빠꾸없이 영어로 적습니다.

```python
True and False  # False가 반환
True or False  # True가 반환
not True  # False가 반환
```
이 연산자들은 모두 다 잘 아실거라 생각해서 설명 없이 넘어갑니다.

파이썬에서 부울값 `True`는 `1`, 그리고 `False`는 `0` 으로도 표현될 수 있습니다. 진짜로요 :

```python
True + True  # 진짜로 2를 반환 ㅋㅋ
```

부울값이 가장 많이 사용되는건 비교연산자의 결과로서입니다 :

```python
x = 10
y = 20

# 다양한 비교연산자들
```

y = 10.00000000000000001

x == y


비교 연산자에 의한 비교는 서로 다른 타입들끼리도 가능합니다 :

```python
x = 10
y = 20.0

x < y  # True 
x == 10.0 # True 
x != True # True 
1 == (x != True) # True 
(1 == 2) == 0 # True 
1 == 2 == 0 # False 
```

비교들 중에 두번째까지는 이해가 되실텐데, 그 이후는 조금 이상하게 보일 수 있습니다. 세번째 줄은 `x != True`라고 하는데, 그 결과가 `True`라고 되어있죠. 이 경우에는 `x`에 10이 저장되어 있고, 위에서 본대로 `True`는 정수로 생각했을 때 `1`이 되므로, 저 표현은 다시 `10 != 1`과 마찬가지여서 `True`가 반환됩니다.

그 다음 놈은 당연히 우선 괄호부터 계산 후, 그 결과로 괄호 바깥을 계산하게 됩니다. 결국 `x != True`는 `True`이므로, 이는 정수로 생각헀을 때 `1`이 되어 `1 == (x != True)`가 `True`가 되는거에요.

`(1 == 2) == 0` 도 마찬가지입니다.

`1 == 2 == 0` 는 연쇄된 비교 연산으로, 신기하게도 `(1 == 2) and (2 == 0)`와 같은 표현입니다. 이렇게 되는 이유는 수학적 등호의 성질(추이성)을 어느 정도 반영해주기 위함입니다. 


### `==` 와 `=`의 차이

`==`는 수학적 등호, `=`는 값을 부여하는 기능을 합니다. 이를 헷갈리시면 절대 안됩니다.

```python
x = 10  # x에 10이 저장됨
x == 10  # True가 반환됨
```


## 타입 변환(Type casting)

위에서도 계속 보다시피, 정수는 부동소수와 같은 경우도 있으며, 부울들은 정수와 같이 작용할 때도 있습니다.

이러한 타입의 자동 변환은 기본적으로 파이썬의 내부에서 암시적으로 행해지는 것들이 있는데요, 만약 명시적으로 타입 변환을 강제하고 싶다면 우리는 함수 `int()`, `float()`, `str()`, `bool()` 등을 이용할 수 있습니다.

```python
x = 10
y = float(x)
z = bool(x)  
```
간단합니다. 각 함수의 이름들은 변경하고자하는 타입이며, 괄호 안에는 변경하려는 객체가 들어옵니다.

````python
x = 10
y = float(x)
z = bool(x)

x, y, z
````

여기서 보시면, 다른건 다 이해가 될텐데, `z`가 1이 아닌데도 `True`가 됨이 의문이실 수 있습니다. 이는 `bool()`함수가 0이 아닌 모든 값을 `True`로 생각하기 떄문입니다.

아직 스트링을 다루지 않았지만, `"123123"`이라는 스트링도 다음과 같이 타입 변환이 가능합니다.

````python
int("123123")
````

또한 반대로 `123123`이라는 `int` 객체도 다시 스트링으로 변환이 가능합니다.

````python
str(123123)
````

하지만 만약 `"안녕하세요"`와 같이 수치형이 아닌 스트링을 변환하면 오류가 발생합니다.

````python
int("안녕하세요")
````
