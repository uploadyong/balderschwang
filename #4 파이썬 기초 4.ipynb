## 투펠(Tupel)



~~~python
(a, b, c, ...)
~~~

-   투펠은 객체들을 콤마로 구분하여 순서대로 나열한 놈입니다.
-   리스트의 `[`, `]`와 달리, 투펠은 소괄호 `(`, `)`가 사용합니다.
-   리스트와 투펠은 서로 매우 닮아있으나, 투펠은 **가변적이지 않다**는 성질이 있습니다.
-   즉, 투펠에 들어있는 성분들은 나중에 변경이 불가능합니다.

예를 들어,
```python
tupel = (1, 2, 3, 4, 5)
tupel[0] = 6 # 투펠에는 값을 부여하는 것이 지원되지 않으므로, 오류가 발생합니다.
```


-   투펠은 **임의 타입**의 **고정된 갯수** 의 객체들을 다루기 위해 사용합니다. 이를 **구조**를 위해 사용한다고도 합니다.

-   주의할 점은, 원소가 하나인 투펠은 항상 마지막에 콤마를 적어줘야한다는겁니다.

```python
tupel = (1,) # 홑원소의 투펠
t = (1)      # 이건 투펠이 아닙니다.
tupel = ()   # 물론 빈 투펠도 정의 가능합니다.
```
-   이는 위 코드의 2행처럼 `t=(1)` 이라고 사용할 경우, 함수에 매개변수와 혼용될 수 있기 떄문입니다. 위와 같이 `(1,)`을 사용하세요.

### 다중 할당(Multiple Assignment)

투펠은 언급한대로, *구조*를 위해 사용되는데요, 투펠은 여러 개의 변수에 값을 동시에 할당시킬 때 사용할 수 있습니다.

이를 다중 할당이라 합니다.

*예시* :

````python
tup = (5, 7)           
first, second = tup    # first에는 5, second에는 7이 할당됩니다.
first
````

또한 함수의 반환값이 여러 개인 경우에 우리는 투펠을 반환할 수 있습니다. 다음과 같이요.

````python
def pair(a,b):
    return a,b # 혹은 return (a,b)도 가능합니다.

first, second = pair(3,7)
print(first)
print(second)
````

보시다시피 반환 시에는 굳이 괄호 `()`없이 그냥 `a,b`를 적어줘도 됩니다. 자동으로 투플로 인식되거든요.

보통 다른 자바, C에선 투펠 같은게 기본적으로는 없어서 여러 값을 반환할 때 항상 배열(Array)을 사용하지만, 파이썬에선 이렇게 여러 값을 한 번에 반환 가능합니다. \[Python\] 포스팅에서는 파이썬의 명령어들만을 단순히 배우는 것을 넘어서서, 나중에 수학적인 프로그래밍을 위한 내용들도 같이 다룰 예정이기 떄문에 공역이 다변수인 경우의 함수를 표현할 때 이러한 투플을 자주 활용하게 됩니다. 

## 딕셔너리(Dictionaries)

~~~python
{key1: value1, key2: value2, ...}
~~~

-   딕셔너리는 사진이라는 이름을 가지나, 실질적으로는 전화번호부에 가까운 데이터 타입입니다. 과거의 상가로나 114같은거죠.
-   딕셔너리는 상가로의 상표명(가게 이름)에 해당되는 키(Key)와 그 상표의 전화번호인 값(Value)의 쌍을 여러 개 모은 구조입니다.
-   키(Key)로는 어떠한 **해쉬 가능한 객체**(Hashable objects)도 지정 가능하며, 값(Value)은 제약이 없어서 어떤 객체든 지정 가능합니다.
-   비유적으로 전화번호부에 기입되려면 사업자 등록(hashable)을 해야한다는 조건이 있으나, 전화번호는 그냥 전화번호(객체)기만 하면 뭐든 괜찮다는겁니다.
-   물론 상가로에 같은 업체가 여러 번 자신을 기입하면 안되는거처럼, Key는 중복되면 안됩니다.

-   이 때, 해쉬 가능한 객체는 불변의 성질을 가지는 객체로, 이를 테면 `int`, `float`, `str` 등의 단순 데이터 터입과 위에서 다룬 투펠이 이에 해당됩니다. 리스트 등과 같이, 스스로의 힘으로 값을 변경시킬 수 있는 친구들은 hashable이 아닙니다.(이 hashable들에 대해서는 나중에 더 다뤄집니다.)

예를 들어 딕셔너리는 다음과 같이 정의됩니다.

````python
d = {
    (1,) : 10, 
    "booleans" : [True, False], 
    (1, 2, 3) : 1
}
d
````

`d`에는 순서대로 `(1,)`, `"booleans"`, `(1,2,3)`이라는 key에 `10`, `[True, False]`, `1`의 value들이 저장되어 있습니다.

여기서 `d`에 저장되어 있는 정보들 중, 키가 `Key`인 놈은, `d[Key]`로 호출이 가능합니다 :

````python
print(d[(1,)])       # 위에서 정의한 d에 저장된 놈들 중, Key가 (1,)인 놈의 Value를 반환
print(d["booleans"]) # 위에서 정의한 d에 저장된 놈들 중, Key가 "boolean"인 놈의 Value를 반환
print([(1,2,3)])     # 위에서 정의한 d에 저장된 놈들 중, Key가 (1,2,3)인 놈의 Value를 반환
````

이 쯤에서 그러면 이게 리스트랑 뭐가 다른건지 싶으실 수 있는데요, 코딩을 하는 입장에서 주요한 차이점은 세 가지가 있습니다.

-    리스트는 내용물에 접근 시 인덱스로 접근하는 반면, 딕셔너리는 키로 접근함. 인덱싱이 전혀 불가능함.
-    리스트는 인덱스의 타입이 정수인 반면, 딕셔너리는 해시 가능한 모든 객체가 인덱스의 타입이 될 수 있음
-    리스트와 딕셔너리는 내용물에 접근하는 방식에 차이가 있어, 딕셔너리가 훨씬 빠른게 작동함.

다른거보다 첫번쨰 차이점은 실제로 지금 상태에서 `d[0]`을 입력하면 오류가 발생함을 통해 딕셔너리는 인덱싱을 지원하지 않음을 확인이 가능합니다.

````python
d[0]
````

`d[0]`은 의미상 `d`라는 딕셔너리에서 Key가 `0`인 놈의 Value을 읽겠다는거기 떄문에, `d[0]`은 `d`에 `0`이라는 Key가 없음에 따라서 오류가 발생됩니다.

인덱싱이 안된다기보다는, 오로지 Key로 값에 접근하는 방식으로 동작한다고 설명하는 게 정확하겠네요.

딕셔너리에 새로운 값을 추가하는 방법은 다음과 같습니다.

```python
d[new Key] = [new Value]
```

이 경우, 딕셔너리 `d`에는 `new Key`를 Key로 하고, `[new Value]`를 값으로 하는 형태가 저장되게 됩니다.

이 때, `new Key`가 이미 `d`에 존재하던 Key일 경우, 당연히 새로운 키가 생성되는 대신, 이전의 `new Key`에 있던 값을 `[new Value]`로 덮어씌워집니다.

예를 들어 다음과 같이 새로운 키와 값을 추가할 수 있습니다. 

````python
d[1] = 20
d
````

위에서 언급한대로 `d`에는 `1`이라는 Key가 없으므로, `d[1] = 20`은 Key가 `1`이고 Value가 `20`인 새 항목을 추가할 뿐입니다.

`d[1]`을 `d`의 인덱스가 1인, 즉 2번째 놈을 가져와라는 의미로 생각하시면 안돼요.


키에 저장되어 있는 값을 불러오면, 당연히 그 값의 데이터 타입에 따른 연산 및 함수들도 사용 가능합니다 :

````python
d["booleans"].append("new")
d
````

`d`의 Key가 `"boolenas"`인 Value는 `[True, False]` 형태의 리스트이고, 이에 따라서 위와 같이 `append()`메소드를 사용할 수 있습니다.

그러면 `d`에 있던 Key `"boolean"`의 리스트에도 위와 같이 새롭게 `"new"`가 추가된 형태가 됩니다.

위에서 설명한대로 딕셔너리에서는 해쉬 가능한 데이터 타입, 즉 불변성을 가지는 데이터 타입만을 키로 설정 가능한데요, 실제로 가변적인 형태의 데이터 타입인 리스트를 키로 지정하면 오류가 생김을 확인할 수 있습니다 :

````python
a = [1, 2, 3]
d[a] = 0
````


**딕셔너리의 확장**:
딕셔너리는 다음의 객체 함수(메소드)를 이용하여 다른 딕셔너리를 이어 붙일 수 있습니다.
~~~python
dict.update()
~~~

해당 메소드의 사용법은 다음과 같습니다.

~~~python
dict1.update(dict2)
~~~

이렇게 사용하는 경우, 기존에 있던 `dict1`에 새로운 `dict2`가 순서대로 이어 붙여집니다.

이 때, 만약 `dict1`과 `dict2`에 서로 겹치는 Key가 존재하는 경우, 이 함수는 말 그대로 **update**이기 떄문에, 기존의 `dict1`의 Key에 있던 값을 `dict2`의 값으로 덮어씌우게 됩니다. 기존의 값은 그저 업데이트가 되는거죠.


````python
print(d)
d1 = {'새로운 친구': 123}
d.update(d1)
print(d)
````

만약 여기에 `d2`를 다음과 같이 정의하여 `d`에 이어 붙인다면, 

````python
d2 ={(1,) : 123}
d.update(d2)
print(d)
````

다음과 같이 기존에 있던 Key `(1,)`에 저장되어 있는 값이 `10`에서 `123`으로 변경됩니다.

**키 삭제하기** :

딕셔너리에 저장되어 있는 키 및 그의 값을 삭제하는 메소드는 다음과 같습니다.

~~~python
dict.pop(Key_to_delete)
~~~

딕셔너리 `dict`에 저장되어 있는 키가 `Key_to_delete`인 항목을 제거하고, 거기 저장되어 있던 값을 반환합니다.

방금 업데이트시킨 `d`의 키가 `"새로운 친구"`인 놈을 삭제하면 다음과 같은 결과가 생겨납니다.`

````python
d.pop("새로운 친구")
````

이제 다시 `d`의 목록을 확인해보시면, `"새로운 친구"` 항목이 사라졌음을 알 수 있습니다.

````python
d
````

## 집합(Sets)

위에서 다룬 딕셔너리는 키라는 식별자에 따라 값을 저장하는 형태였는데요, 집합의 경우 식별자 없이 값만 저장하는 데이터 타입입니다.

집합은 다음과 형태로 정의됩니다.

~~~python
s = {value1, value2, ...}
~~~

-   수학의 집합을 표현하기 위한 데이터 타입입니다.
-   어떤 객체든 Value로 사용할 수 있었던 딕셔너리와 다르게, Value로는 해쉬 가능한 객체들만이 올 수 있습니다.
-   집합의 Value들은 정렬되지 않은 형태입니다.
-   중복되는 Value가 있다면 당연히 무시됩니다. 

````python
s = {1.0, 2, "asd",1,1,1.000000001}
s
````

물론 여기서는 마치 저 수치적인 Value들이 정렬되는것 처럼 보이나, 이는 집합의 내부 구현 과정에서 발생하는 현상일 뿐, 정렬이라 보긴 어렵습니다.

**집합의 원소에 접근** 

리스트나 딕셔너리와는 달리, 집합은 순서가 없기 떄문에 인덱싱 등 호출을 통한 원소 접근이 불가능합니다.

````python
s[1]
````

일반적으로 집합의 원소에 직접 접근하는 단순한 방법은 없고, 간접적으로 개별 원소가 해당 집합에 들어있는지 확인 가능합니다.

다음은 특정 `element`가 `s`이라는 집합에 있는지 확인하는 명령어입니다.

~~~python
element in s
~~~

이 명령어는 `element`가 `s`에 들어있는지를 확인 후, `True` 혹은 `False`를 반환합니다.

예를 들어 다음과 같이 사용 가능합니다.

````python
print(1 in s)
print(123 in s)
````

직접 집합의 원소를 확인하려면 위에서도 사용한 메소드 `pop()`을 사용하거나(이 경우, 어떤 원소가 튀어나올지는 모릅니다 ㅋㅋ 랜덤으로 하나 지워지고 하나가 반환됩니다.), 타입 변환 함수 `list()`를 사용할 수 있습니다.

```python
list(s)
```

이러면 집합 `s`가 리스트로 변환되는데요, 이 때 순서는 보장할 수 없습니다. 

따라서 만약 순서를 정렬하고 싶으시다면 `sort()`메소드를 사용하셔야합니다.

물론 `s`에 `int`형 객체와 투플이 동시에 들어있는 등, 서로 비교가 불가능한 타입들이 들어있다면, `sort()` 메소드는 에러를 발생시킵니다. 

참고로 투플들만이 들어있는 경우에는 정렬이 가능합니다. 이 때 `sort()`는 각 투플들의 성분들을 순서대로 비교하여 정렬시킵니다.


집합에 새로운 원소를 추가할 떄에도, 기존의 리스트나 딕셔너리처럼 인덱싱으로 값을 직접 부여하는 대신, 메소드 `add()`를 사용합니다 :

````python
s.add(123)
s
````

수학과 같이, `union(합집합), intersection(교집합), difference(차집합)` 등의 집합 연산이 가능합니다. 

````python
s1 = s.union({1234})
s1
````

````python
s2 = s.intersection({1,2})
s2
````

````python
s1.difference(s2)
````

만약 집합에서 특정 원소를 제거하고 싶다면, `revove()`메소드를 사용하시면 됩니다 :

````python
s.remove(2)
s
````

# 루프(Loops)

## 루프/반복문

루프/반복문은 말 그대로 명령을 여러 번 반복하기 위한 도구입니다.

파이썬에 이러한 루프는 크게 `for`와 `while` 두 가지가 있습니다. 하나씩 보시죠.

### `for`

~~~python
for index in iterable_object:
    # 반복 수행할 명령어
    ...
~~~

-   `for`는 `iterable_object`의 각 성분에 대해서, 코드 블록에 있는 명령을 수행합니다.

-   `if` 및 함수에서와 마찬가지로 `for`는 수행할 명령어를 작성할 코드 블록이 필요합니다. 따라서 들여쓰기를 필수적으로 해주셔야합니다. 

-   이 때, `iterable_object`의 어떤 성분을 기준으로 코드 블록이 수행되는지는 `index`가 결정합니다.

-   `iterable_obejct`는 말 그대로, Iteration(반복)이 가능한 객체로, 이를 테면 다음과 같은 놈들이 있습니다 :

    -   리스트(List)
    -   딕셔너리(Dictionary)
    -   스트링(str)
    -   투펠(Tupel)
    -   집합(Sets)
      

반복 가능한 객체에 대한 이야기는 아래에서 다시 다뤄집니다.

이 포스팅에서 자세히 다뤄지지 않는 반복 가능한 객체들도 아래의 놈들의 사용법을 숙지하신다면 문제없이 사용 가능하실겁니다.

이제 반복 가능한 객체들에 `for`를 어떻게 사용하는지 살펴보시죠.

#### 리스트

반복은 모든 리스트의 성분에 대해서 순서대로 진행됩니다.(왼쪽부터 오른쪽) 다음의 형태로 사용 가능합니다 :

```python
for val in [value_1,...,value_n] :
    # 수행할 명령어들
    # 블록 내부의 모든 명령어들을 각 value_1,...,value_n에 대해서 한 번씩 수행함.
    # 코드 블록 내부에서 만약 현재 수행중인 값에 접근하는건 val을 호출하여 가능.

```

예를 들어 다음과 같이 리스트의 모든 숫자를 출력하는 코드를 작성할 수 있습니다.

````python
for val in [1, 2, 3]:
    print(val,"\n")

print("끝")
````

#### 딕셔너리

리스트와 달리, 딕셔너리에서는 Key를 통해서 반복 작업이 수행됩니다. 다만 순서는 사용하시는 파이썬 버젼에 따라 다릅니다.

구버젼에서는 랜덤 순서로 반복 작업이 수행되었으나, 이제는 딕셔너리에 삽입된 순서대로 작업이 수행됩니다.

````python
d = {1: "하나", (1,2) : "얘는 뭐야" , 2: "둘"}
for key in d:
    print(key, " = ", d[key])
````

구버젼이라면 위와 다른 순서가 나타날 수 있습니다.

위에서는 딕셔너리의 Key에만 접근하여 Key와 그 Key에 저장되어 있는 Value를 출력하는데요,

이에 따라서 `d[key]`와 같이 직접 Key를 이용하여 Value에 접근 후 값을 반환하는 방법을 사용합니다.

위는 간단해서 괜찮지만, 복잡한 프로그램이면 이런 `d[key]`같은 직접 찹조는 가독성을 해쳐서, 다음과 같이 `items()`메소드를 사용하기도 합니다.

```python
for key,val in d.items()
    # 수행할 반복 명령
```

`items()` 메소드는 딕셔너리에 저장되어 있는 Key와 Value를 `(Key, Value)` 형태의 투플로 묶은 놈들을 모두 모은 객체(`dict.items`)의 형태로 반환합니다.

이를 이용하면 우리는 위의 프로그램을 다시 다음으로도 적어줄 수 있습니다.

````python
for key, val in d.items():
    print(key, '=', val)
````

`items()`함수는 딕셔너리의 각 성분들을 `(Key,Value)`의 쌍으로 묶어서, 이들을 마치 리스트와 비슷한 객체인 `dict.item`이라는 이름의 객체의 형태로 내보냅니다. 얘는 파이썬의 딕셔너리 전용 객체로, 더 넓은 관점에서는 `View`라는 놈의 일종으로 여겨집니다.

참고로 `for key, val in d.item()`라는 구문은 투플에서 다룬 다중 할당입니다. `key`라는 이름의 변수와 `val`이라는 이름의 변수에다가 `d.items()`에 들어있는 각 투플 `(Key, Value)`에 대해서, `key,val = (Key, Value)`의 할당을 한다는겁니다.

````python
for kv in d.items():
    print(kv[0], '=>', kv[1])
````

이 `d.items()`로 반환되는 뷰 객체도 저렇게 반복이 가능하며, 각 성분들은 `(Key, Value)` 형태의 투플이므로, 저렇게 인덱스로 접근이 가능합니다.

`enumerate()`함수는 파이썬에 기본적으로 내장되어 있는 함수로, 리스트 등의 iterable 객체의 내용물에 순서대로 순번을 부여하여, `(number,value)`의 형태의 투플로 묶어서 모은 enumerate라는 객체를 내보내는 함수입니다.

````python
for idx, val in enumerate('몇번째글자일까용'):
    print('{0} -> {1}'.format(idx, val)) # 여기서 print() 내부에 있는 구문은 나중에 fstring을 다루며 더 다뤄집니다.
````

#### `for`의 예시 : 숫자들을 합하기

`for`는 반복적인 작업을 위해 사용되고, 우리는 예를 들어 아래와 같이 반복 가능한 객체에 들어있는 숫자들을 모두 더하는 함수를 정의할 수 있습니다.

````python
def calculate_sum(numbers):
    print('Start')
    result = 0
    for idx, n in enumerate(numbers):
        print(f'index = {idx}') # print() 안의 f''꼴의 놈은 f스트링이라는 객체로, 나중에 다시 다뤄짐.
        print(f'added number = {n}')
        result = result + n
        print(f'result after adding = {result} \n')
    print('End')
    return result
````

````python
calculate_sum({-1, -4, 3, 8})
````

### `while`

`while`은 `for`와 같은 반복문인데요, 반복 가능한 객체를 따라서 반복하는 `for`와 달리 `while`은 (`if`처럼) 조건에 따라 반복됩니다.

~~~python
while 조건:
    ...
    ...
~~~

-   먼저 `조건`의 값이 계산된 후, 이 놈이 참인 경우 아래의 코드 블록이 수행됩니다.
-   코드 블록을 한 사이클 수행한 다음에는 다시 그 상태에서 `조건`이 참인지 여부에 따라서 코드 블록을 반복합니다.
-   이러한 반복은 조건이 더 이상 참이 아니고, 거짓이 될 때까지 이어집니다.

경우에 따라서는 코드 블록이 전혀 수행되지 않거나, 끝없이 코드 블록이 반복 수행 되는 경우도 있습니다.


#### 예시 : 최대공약수

두 수 `n`과 `m`의 최대공약수를 계산하는 유클리드 알고리즘은 다음과 같습니다 :

-  `n != m`이 성립하는 경우,:
    -   `n`과 `m` 중 더 큰 수를 두 수의 차이 $\vert n-m\vert$로 대체하는걸 `n == m`이 될 때까지 반복한다.
-  `n == m`이 성립한다면, `n`은 최대공약수이다.

이는 `while`을 사용하여 다음과 같이 구현할 수 있습니다.

````python
def gcd(n, m):
    while m != n:
        print('Loop: n =', n, 'm =', m)
        if n > m:
            n = n - m
        else:
            m = m - n
            
    print('GCD : ', n)
    return n

gcd(24, 18)
````

### `break`
`break`는 반복문에 사용되는 컨트롤 명령어로, 반복문 내의 `break`의 위치에서 곧 바로 반복문을 종료할 수 있습니다.

~~~python
break
~~~

`while`은 기본적으로 조건이 참이면 계속 반복되고, `for`도 모든 iterable의 성분을 돌 떄까지 반복되나, 이를 강제적으로 종료하는 명령어입니다.

종료 후에는 당연히 반복문의 아래에 있는 명령들이 다시 파이썬의 순서에 따라 순차적으로 이어집니다.

````python
for val in range(-3, 3):
    if val > 0:
        break
    print(val)
print('끝')
````

C나 자바 등의 다른 언어에서는 대부분 이 `for` 등의 반복문에서의 인덱스들은 그 반복문 내부에서만 유효합니다.

하지만 파이썬에서는 반복문의 밖에서도 인덱스가 유효합니다 :

````python
def first_positive_of(numbers):
    for n in numbers:
        if n > 0:
            break
    return n # 들여쓰기를 보면 아시다시피, n은 for 반복문 밖에 있습니다. C나 자바에서는 n이 유효하지 않지만, 파이썬에서는 이렇게 n을 리턴 할 수 있습니다.

first_positive_of([-3, -4, 1, 5, -2])
````

특히나 저런 경우 `n`에는 반복문을 종료할 시점에 저장되어 있던 값이 그대로 저장됩니다. 

원래 C나 자바에서 반복문 안은 반복문 밖과 완전 다른 취급이라, 특별한 조치를 취하지 않는 이상 반복문을 도는 변수를 포함하여 반복문 내부에서 새롭게 정의되는 변수들은 그 반복문에서만 유효하고, 반복문을 벗어난다면 더 이상 그 변수들에는 접근이 불가능합니다.

하지만 파이썬은 그러지 않고, 반복문 내부의 이야기도 끝까지 들고가요. 이는 실사용에서 많은 경우 편리한 경우도 많지만, 동시에 일반적인 프로그래밍 언어와는 많이 다른 부분이기 떄문에, 주의가 필요합니다. 이 떄문에 보통 반복문 내부의 변수들의 이름은 간결하고 누가봐도 반복문 전용인 이름들을 사용하는 게 바람직합니다.

#### `break`의 `while`에서의 활용

`while`은 기본적으로 조건이 만족되는 경우에 특정 작업을 반복하기 위한 구조이고, 이에 따라 매 반복 단계에서 조건이 우선적으로 확인됩니다.

근데 떄로는 조건같은거 신경쓰지 않고 반복하다가, 특정 단계에서 조건을 확인하여 반복문을 종료하는 게 더 도움이 되는 경우도 있습니다.

이 때는 조건으로 `True`를 사용합니다.

**예시 : 고정점 반복법**

함수 $f$에 대해, 다음을 만족하는 점 $a$를 고정점이라고 합니다.

$$f(a) = a$$

특정 수학적 조건을 만족하는 함수 $f$에 대해서는 시작점 $x_0$을 골라서, $x_{n+1} = f(x_n)$의 형태로 수열 $(x_n)_{n\in\mathbb{N}}$을 구성시키면

$x_n\underset{n\to\infty}{\to}a$인 $a$가 존재하며, $f(a) = a$가 성립하여, 결국 고정점으로 수렴하게 됩니다.

따라서 우리는 $a = \lim\limits_{n\to\infty}x_n = \lim\limits_{n\to\infty}f(x_n)$임을 알 수 있고, 이를 이용하면 충분히 큰 $N$에 대해 $f(x_N)\approx a$라고 할 수 있습니다.

이를 이용하여 우리는 $x_1,x_2,...$를 구성시켜 $x_N \approx a$인 $N$을 구해, $x_N$을 고정점의 근사로 계산할 수 있습니다.

이 때,  $x_N \approx a$임을 판단하기 위한 근거는 당연히 $x_N$과 $a$의 오차입니다. 우리는 이 오차가 충분히 작으면, $x_N$을 $a$의 근사로 생각할 수 있습니다.

따라서 우리가 생각하는 오차의 허용치 $\epsilon>0$을 골라서, $n = 1,2,3,...$에 대해 다음이 만족될 때까지 $x_n$을 계산합니다.

$$\vert f(x_n) - a\vert < \epsilon$$

그리고 이걸 만족하는 $N$이 발견된다면, $x_N$을 $a$의 근사값으로 반환합니다.

기본적인 아이디어는 이건데요, 여기는 꽤나 치명적인 모순이 있습니다. 우리는 $a$를 구하고 싶은데, $a$를 사용해야한다는거죠.

이건 말이 안되죠. 우리가 알고싶은 정답을 그 정답을 구하기 위한 과정에서 사용한다는게요.

따라서 우리는 이 대신에, 다음의 조건을 사용합니다.

$$\vert f(x_n) - x_n\vert <\epsilon$$

직관적으로 만약 충분히 큰 $N$에 대해서 $x_N \approx a$라는건 $f(x_N) \approx f(a) = a \approx x_N$라는 의미이니, $\vert f(x_N)-x_N\vert \approx 0$이겠죠?

그니까 만약 각 $n$번째 반복 단계에서 $x_n$과 $f(x_n)$의 오차가 충분히 적다면, 우리는 이미 이를 고정점의 근사로 생각할 수 있습니다.

\[해석학 2\](다변수 미적분학)를 공부하신 분들은 이 과정이 결국 바나흐의 고정점 정리를 간략히 설명하는 것임을 알 수 있으실겁니다.

여튼 이 과정은 다음과 같이 파이썬으로 구현할 수 있습니다.

````python
def fixpoint(f, x0, eps = 1e-5) :
    """
    함수의 고정점을 구하는 함수입니다.

    Args:
        f : 함수
        x0 : 시작점
        eps 오차 허용치, 디폴트로 1e-5값이 들어갑니다.

    Returns :
        float : 계산된 결과
    """
    x = x0
    while True:
        # 이 경우 우리는 while의 조건을 판단하기 전에 우선 f(x)를 구해야합니다.
        if abs(f(x) - x) < eps: # 만약 오차가 엡실론보다 작다면 우리는 이 반복 과정을 멈춥니다.
            break
        x = f(x)
    return x
````

간단합니다. 전부 설명한 그대로에요. 우리는 각 $n$번째 계산 과정에서

$$\vert f(x_n)-x_n\vert <\epsilon$$

이 성립하는지를 판단하여, 만약 성립하지 않는다면 $x_{n+1} := x_n$을 정의하여 $n+1$번째 과정을 반복합니다.

만약 위 조건이 만족되는 경우, 우리는 `break`로 반복문을 탈출하고 저장된 값을 반환합니다.

예를 들어 다음의 함수는

$$f(x) = \frac{1}{x} +\frac{x}{2}$$

고정점 $a$가 존재하는 경우, 다음이 만족됩니다 :
$$a= f(a) \quad \Leftrightarrow \quad a = \frac{1}{a} + \frac{a}{2} \quad \Leftrightarrow \quad a^2 = 1 + \frac{a^2}{2} \quad \Leftrightarrow \quad a^2 = 2\quad \Leftrightarrow \quad a = \pm \sqrt{2}$$

따라서 이 $f$ 에 `fixpoint()`를 사용한다면, 우리는 결국 근사적으로 $\sqrt{2}$의 값을 얻어낼 수 있습니다 : 


````python
def f(x):
    return 1/x + x/2
fixpoint(f, 10)
````

### `continue`

반복문은 결국 다음과 같이 동작합니다.

$$\text{반복 과정 1} \quad \Rightarrow\quad \text{반복 과정 2} \quad \Rightarrow  \quad... \quad\text{반복 과정 n} $$

`for`와 `while`은 $n$번째 반복 과정, 즉 종료의 시점이 어떻게 결정되는지에 따라서 차이가 날 뿐이구요.

위에서 다룬 `break`는 마지막 단계까지 이어가지 않고, 중간 단계에서 반복문을 미리 종료하는걸 의미했습니다.

`continue`는 각 반복 과정들의 중간에서 그 반복 과정을 건너뛰고, 다음 반복 과정으로 넘어가는걸 강제하는 명령어입니다.


~~~python
continue
~~~

예를 들어 홀수를 출력하는 함수를 우리는 다음과 같이 구현할 수 있습니다.

````python
def print_odds(numbers):
    for n in numbers:
        if n % 2 == 0: # 짝수면 continue로 다음 과정으로 넘어감
            continue
        print(n)

print_odds(range(10))
````

## 리스트 컴프리헨션(List Comprehensions)

파이썬에는 리스트 컴프리헨션이라는 리스트를 구성하기 위한 매우 강력한 도구가 존재합니다.

리스트를 구성할 때, 다음의 문제 상황이 주어진다고 생각해보시죠.

우리는 집합 $M\subset \mathbb{N}$에 대해, 다음과 같이 $M$의 모든 짝수를 제곱시킨 놈들을 모은 집합을 리스트로 표현하고자 한다.

$$
\left\{ n^2 : n \in M,\, n \text{은 짝수} \right\}
$$

지금까지 배운 내용을 활용하면 우리는 이 집합을 다음과 같이 반복문으로 구성시킬 수 있습니다.

````python
def squares_of_evens(numbers):
    result = []                    # 일단 빈 리스트를 선언
    for n in numbers:              # numbers의 각 원소 n에 대해서
        if n % 2 == 0:             # n이 짝수이면
            result.append(n ** 2)  # result에 n**2를 새로 추가함
    return result                  # 반복문의 결과물을 반환

squares_of_evens(range(10))
````

파이썬은 이러한 과정을 그냥 한 번에 할 수 있는데요, 그걸 리스트 컴프리헨션이라고 합니다.

리스트 컴프리헨션을 이용하면 위 `squares_of_evens()`함수를 사용할 필요 없이, 다음의 표현으로 리스트 구성이 가능합니다.

```python
[n**2 for n in numbers if n%2 == 0]
```

이걸 이용하면 `squares_of_evens()`를 다음과 같이 세련되게 정의할 수도 있습니다.

````python
def squares_of_evens(numbers) : 
    return [n**2 for n in numbers if n%2 == 0]

squares_of_evens(range(10))
````

대충 느낌이 오죠. 우선 가장 왼쪽에는 어떤 놈(`n**2`)들이 저장될건지, 그리고 그 놈은 어느 출신인지(`numbers`의 각 `n`들), 

그리고 `n**2`를 저장하기 위한 조건이 있는지(`n%2 == 0`, 즉 짝수 n에 대해서)가 차례대로 옵니다.

이를 일반적으로 다음과 같은 공식으로 구문화할 수 있습니다.

~~~python
[표현 for 변수 in Iterable]
[표현 for 변수 in Iterable if 조건]
~~~

간단합니다. 표현에는 리스트에 결과적으로 어떤 놈이 저장될 것인지, 위 예시의 맥락에서는 $n^2$에 해당되는 친구가 옵니다.

그리고 우리는 이 표현을 주어진 Iterable객체(위 예시 맥락에서 집합 $M$)에 있는 각 변수($n$)들을 가져와서 완성시키구요.

또한 만약 이 생성에서 조건이 있는 경우(위 예시 맥락에서 $n$이 짝수라는 조건), 우리는 저렇게 마지막에 `if` 이후에 조건을 붙여줍니다.

조건이 없다면 `if`는 생략하구요.


이러한 리스트 컴프리헨션는 심지어 복수의 `for`와 `if`를 사용할 수도 있다는 점에서 강점이 있습니다.

````python
[(m, n) for m in range(11) if m % 2 == 0 
        for n in range(m) if n % 2 != 0]
````

방법은 간단합니다. 위 예시에서도 보시다시피, `for-if`의 쌍을 순차적으로 배치하시면 됩니다. `for-if`는 왼쪽에서 오른쪽으로 동작합니다

위 예시는 결국 첫번째 성분 $m$으로는 ${0,...,10}$중 짝수, 두번째 성분으로는 ${1,...,m}$ 중 홀수(즉, $m$보다 작거나 같은 홀수)를 골라서 만들어지는

모든 투펠들을 모은 리스트를 생성합니다. 이 때, $m = 0$인 경우엔 당연히 $n$을 구성할 수 없으므로, $(0,n)$형태의 투플은 저장되지 않습니다.

## 제너레이터(Generator Expressions)

오늘 다룬 컨테이너 타입들인 리스트, 딕셔너리, 집합은 모두 컴프리헨션이 가능하나, 투펠은 컴프리헨션이 없습니다.

물론 투펠에도 컴프리헨션과 비슷한 역할을 하는 도구가 있는데요, 이를 우리는 **제너레이터**라고 합니다.

````python
tupel = (n ** 2 for n in range(10) if n% 2==0)
````

제네레이터는 말 그대로 뭔가를 생성하는 친구로, 위의 코드는 결국 $n^2$를 모든 $\{0,....,9\}$범위의 짝수에 대해서 제곱하는 놈을 생성한다는 의미를 가집니다.

결국 컴프리헨션이랑 구문 문법도 똑같고, 똑같은거 아닌가? 하실 수 있는데요, 제너레이터와 컴프리헨션은 그 작동 방식에서 차이가 있습니다.

컴프리헨션은 그 때 그 때 해당되는 값들을 직접 생성하여 리스트/딕셔너리/집합에 저장합니다.

그에 비해서 제너레이터는 값을 직접 생성하는 대신, 그 값을 생성하는 규칙을 들고있다가, 필요할 시 그 규칙을 이용하여 값들을 즉시 생성합니다.

이 덕분에 제너레이터 표현을 사용한다면 값들을 저장할 공간이 필요 없고, 각 값들에 접근하는 것도 훨씬 빠릅니다.

주의할 점은 기존의 컴프리헨션은 결국 그 결과물이 그대로 리스트면 리스트, 딕셔너리면 딕셔너리, 집합이면 집합이었는데요,

제너레이터를 사용하여 투펠을 정의하는 경우, 거기 들어있는건 사실 투펠이 아니라 제네레이터라는 다른 객체입니다.

````python
tupel
````

따라서 만약 이 제네레이터를 투펠화시키고 싶다면, 다음과 같이 형변환을 해주셔야 합니다.

````python
tuple(tupel)
````

엄밀히 말하면 제네레이터 표현은 오로지 투펠만을 위한 개념은 아닙니다.

그냥 맥락상 컴프리헨션을 생각한다면 투펠 입장에서는 컴프리헨션이 없는 대신, 제네레이터를 사용할 수 있다 그 정도인거죠.

여튼 이 `tupel`에 들어있는건 실제로 투펠이 아닌 제네레이터이기 떄문에, 기존의 투펠과는 다른 방식으로 값들에 접근해야 합니다.

투펠은 인덱싱을 지원하여 `tupel[1]`이런식으로 인덱스가 1인, 즉 두번째 성분을 직접 읽는게 가능했으나

제네레이터 표현식은 이런 직접적인 인덱싱이 불가능합니다.

하지만 제네레이터는 iterable이기 떄문에, `for`를 이용하여 예를 들어 다음과 같은 방식으로 값들에 접근할 수 있습니다.

````python
tupel = (n ** 2 for n in range(10) if n% 2==0)
for i in tupel:
    print(i)
````

여기서 주의할 점이 있어요. 지금 저 `for` 문을 다시 한 번 해보죠.

````python
for i in tupel:
    print(i)
````

그러면 아무런 반응이 없을겁니다. 그게 정상이에요. 제네레이터는 특수한 객체로, 계속 사용이 불가능하고, 딱 1회용이거든요. 

우리는 위에서 이미 `for`로 제네레이터를 끝까지 다 사용했으므로, 두 번 다시 이를 사용할 수 없습니다.

따라서 제네레이터를 사용한 이후에는 재사용을 위해서는 다시 초기화(새로 제네레이터를 생성)를 해줘야 합니다.

이 때, "사용"이란 말에 주의하셔야합니다. 조금이라도 사용했다면, 이미 사용한 애들은 다시 사용이 불가능합니다.

그리고 아직 사용되지 않은 친구들은 그대로 남아있어요.

위의 `tupel`에는 결국 `0, 4, 16, 36, 64`라는 수를 생성하는 제네레이터가 들어있는데요,

만약 그 제네레이터의 첫번째 성분인 `0`을 읽어 사용한다면, 제네레이터는 `4, 16, 36, 64`만을 기억하고 있습니다.

````python
tupel = (n ** 2 for n in range(10) if n% 2==0)
next(tupel)
````

이렇게 우리는 `next()`함수를 이용하여 제네레이터가 생성할 수 있는 다음 값을 불러올 수 있습니다.

여기서 또 `next()`를 사용하면, 설명한대로 제네레이터는 `4, 16, 36, 64`를 기억하므로 다음과 같이 4가 출력됩니다.

````python
next(tupel)
````

제네레이터는 지금 값을 어디까지 읽었는지만을 기억합니다. 한 마디로 한 방향으로만 가는거죠. 뒤로가는건 불가능하구요.

따라서 제네레이터는 그 전체를 모두 사용하지 않고 일부만을 사용했더라도,

사용되기 이전의 정보는 전혀 기억하고 있지 않으므로, 사용되지 않은 그 이후의 생성 값들만을 가지고 있습니다.

위의 `for`문의 경우 `tupel`의 모든 성분을 다 읽었으므로, 아무런 값도 생성하지 못하여 반응이 없는거에요.

만약 뒤로가고 싶다면 마찬가지로 새로 제네레이터를 생성해야합니다.

위와 같은 컴팩트한 모양으로는 제네레이터 표현을 사용하기 어려운 경우도 있습니다.

좀 더 복잡한 규칙을 사용해야하는 경우인데요, 이 때는 제네레이터 함수(Generator function)라는 개념을 이용하여 이를 대신합니다.

제네레이터 함수는 함수와 똑같이 정의되며, 예를 들어 다음과 같습니다.

````python
def generator_even_squares():
    n = 0
    while True:
        n += 1
        yield n**2

square = generator_even_squares()
````

이 `generator_even_squares()`를 정의하여, 위와 같이 `square`에 이 함수를 소환한다면

결국 `square`에 들어오는 놈은 $(1,4,9,...)$이런식으로 $n^2$인 무한 수열을 투펠에 넣은 모양이 됩니다.

이렇게 마치 무한 투펠과 같은 놈을 생성하거나, 투펠에 들어가게 만들 값이 복잡할 경우에는 위의 제네레이터 함수를 이용하여 제네레이터를 생성합니다.

제네레이터 함수는 위와 같이 함수와 거의 동일하게 사용할 수 있지만, 기존의 함수는 `return`으로 값을 내보낸 반면, 제네레이터 함수는 `yield`를 사용합니다.

## Iterable과 Iterator

Iterable이라는 객체는 말 그대로 iterate를 할 수 있는, 다시 말해서 반복을 할 수 있는 객체들을 의미합니다.

뭔가 말이 이상한데요, 그냥 간단히 얘기해서 여러 객체들로 구성되며, 그 구성 객체들을 하나 하나씩 뽑아낼 수 있는 놈을 의미합니다.

리스트, 투펠, 딕셔너리, 스트링, 집합 등은 모두 그를 구성하는 성분들에 일일이 접근하여 하나씩 뽑을 수 있잖아요? 이런걸 의미하는거에요.

마찬가지로 제네레이터도 생성되는 값들에 반복적으로 접근 가능하므로, iterable에 해당됩니다.

각 Iterable은 구조에 따라서 성분들에 어떤 식으로 접근할 수 있는지가 달라지는데요, 이러한 규칙을 정하는 객체를 Iterator라합니다.

만약 우리가 Iterable을 반복문 등에서 사용하는 경우, 파이썬은 암시적으로 이 Iterable을 위한 Iterator를 생성합니다.

Iterator는 `next()`라는 함수를 통해서, Iterator가 해당 Iterable에서 뽑아낼 다음 값을 출력하게 할 수 있습니다.

예를 들어 다음과 같은 리스트가 주어진다면

````python
l = [1,2,3]
````

이 리스트의 Iterator는 다음의 형변환으로 불러올 수 있습니다.

````python
it = iter(l)
it
````

그러면 우리는 이 때 `next()`함수를 사용하여, 리스트의 Iterator가 읽을 다음 숫자를 불러올 수 있습니다.

````python
print(next(it))
print(next(it))
print(next(it))
````

여기서 만약 한 번 더 `print(next(it))`를 한다면, `StopIteration`이라는 이름의 익셉션이 발생합니다.

````python
print(next(it))
````

눈치채셨을 수 있는데요, 제네레이터는 Iterable이자 동시에 자기 자신이 Iterator가 되는 특이한 친구입니다.

애초에 제네레이터는 자신을 어떻게 반복할지 규칙이 정해져있거든요.

따라서 반복문 등에서 제네레이터가 사용될 때는 따로 Iterator가 생성되지 않습니다.


## Iterable과 Iterator들의 특수한 함수들

### `range()`

`range()` 는 특수한 함수(정확히는 생성자, 나중에 더 다뤄짐)로, 수들의 열을 생성하기 위한 함수입니다. 이 함수로 생성되는 수들의 열은

반복문에서도 사용 가능합니다. 이 놈은 제네레이터처럼 생성 규칙 그 자체를 보유하는 놈으로, 리스트 등보다 접근 속도가 빠르고 효율적입니다.

### `zip()`

`zip()` 함수는 복수의 Iterable 객체들을 받아서, 공통 Iterator를 반환하는 함수입니다. 이 공통된 Iterator는 그냥 각각의 Iterable들에서 

값을 뽑는 순서를 그대로, 투플에 넣은 형태라고 생각하시면 됩니다.

예를 들어 투펠과 리스트가 주어진다면, `zip()`함수가 반환하는 Iterator는 투펠의 첫번째 성분, 리스트의 첫번쨰 성분을 순서대로 투펠에 넣은 놈을 먼저 고르고, 

다음에는 다시 각각의 두번째 성분을 골라서 투펠에 넣은놈을 고르는 이런 방식을 반복하는 형태의 Iterator가 됩니다.

예를 들어 아래와 같이

```python
zip(A,B,C) 
```

 `A`, `B`, `C`라는 Iterable을 `zip()`함수에 넣으면, 그 결과로 반환되는 Iterator는 순서대로

$$(\text{A[0], B[0], C[0]}) \quad \overset{\text{next()}}{\Rightarrow}\quad (\text{A[1], B[1], C[1]})\quad \overset{\text{next()}}{\Rightarrow}\quad...$$

이런 형태로 데이터들을 반복하는 Iterator가 됩니다.

만약 이 Iterator에 `next()`를 반복하여 사용하는데, 만약 Iterable 중 먼저 반복이 끝나는 친구가 있다면, 이 Iterator()도 반복을 멈춥니다.

### `enumerate()`

`enumerate()`는 Iterable을 하나 받아서, Iterable의 Iterator가 어떤 순서대로 데이터를 읽는지를 알려주는 순번이 적혀있는 Iterator를 반환합니다.

예를 들어, 다음과 같이 `A`라는 Iterable을 집어넣는다면,

```python

enumerate(A)
```

그 결과로 나오는 Iterator는 다음의 형태로 데이터들을 읽습니다.
$$(\text{0, A[0]}) \quad \overset{\text{next()}}{\Rightarrow}\quad (\text{1, A[1]})\quad \overset{\text{next()}}{\Rightarrow}\quad...$$
